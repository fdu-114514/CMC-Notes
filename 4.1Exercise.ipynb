> 1. 椭球面方程为$\frac{x^2}{9}+\frac{y^2}{16}+\frac{z^2}{4}=1$，试求过$x$轴且与椭球面的交线是圆的平面方程，并求出该圆的半径。

**思路：** 过$x$轴的平面方程只含$y, z$，知道这个就不难了。

**解答：** 设平面方程为$z=ky$，与椭球面方程联立得到交线$l$的方程
$$\begin{cases}
    x^2&=9(1-\frac{y^2}{16}-\frac{k^2y^2}{4}) \\
    z&=ky
\end{cases}$$
$l$上的点$(x, y, z)$到原点的距离为$\rho=\sqrt{x^2+y^2+z^2}$，即
$$\begin{aligned}
    \rho^2&=x^2+y^2+z^2 \\
    &=9-\frac{9}{16}y^2-\frac{9}{4}k^2y^2+y^2+k^2y^2 \\
    &=9+(\frac{7}{16}-\frac{5}{4}k^2)y^2
\end{aligned}$$
交线是圆的话，$\rho$应该是常数，因此有
$$\begin{aligned}
    \frac{7}{16}&=\frac{5}{4}k^2 \\
    k&=\pm\sqrt{\frac{7}{20}}
\end{aligned}$$
因此，平面方程为$z=\pm\sqrt{\frac{7}{20}}y$，圆的半径$\rho=3$。

> 2. 求过原点的平面使得它与椭圆柱面$\frac{x^2}{4}+\frac{y^2}{9}=1$截线为一个圆。

**思路：** 长轴是$y$轴，因此平面应该与$y$无关。

**解答：** 设平面方程为$z=kx$，交线上任一点$(x, y, z)$到原点的距离满足$\rho=\sqrt{x^2+y^2+z^2}$为定值。那么有
$$\begin{aligned}
    \rho^2&=x^2+y^2+z^2 \\
    &=x^2+9(1-\frac{x^2}{4})+k^2x^2 \\
    &=9+(k^2-\frac{5}{4})x^2
\end{aligned}$$
它与$x$无关，所以
$$\begin{aligned}
    k^2-\frac{5}{4}&=0 \\
    k&=\pm\frac{\sqrt{5}}{2}
\end{aligned}$$
因此平面方程为$z=\pm\frac{\sqrt{5}}{2}x$。

> 3. 证明空间曲线
> $$\begin{cases}
    x^2+y^2+z^2&=8 \\
    xy+xz+yz&=x+y+z
\end{cases}$$
> 是两个圆，并求出它们各自的半径。

**思路：** 形式上非常像$(x+y+z)^2$，那就用吧。

**解答：** 注意到
$$\begin{aligned}
    (x^2+y^2+z^2)+2(xy+xz+yz)&=(x+y+z)^2 \\
    (x+y+z)^2-2(x+y+z)-8&=0 \\
    (x+y+z+2)(x+y+z-4)&=0
\end{aligned}$$
因此曲线方程又可以写成
$$\begin{cases}
    x^2+y^2+z^2&=8 \\
    x+y+z+2&=0
\end{cases}\text{或}\begin{cases}
    x^2+y^2+z^2&=8 \\
    x+y+z-4&=0
\end{cases}$$
这都是球心在原点的球面与平面的交线，因此必定为圆。在第一种情况，球心到平面的距离为$d_1=\frac{2}{\sqrt{3}}$，由勾股定理可知圆的半径$r_1=\sqrt{8-\frac{4}{3}}=\sqrt{\frac{20}{3}}$。类似地，可求出$d_2=\frac{4}{\sqrt{3}}, r_2=\sqrt{8-\frac{16}{3}}=\sqrt{\frac{8}{3}}$。

> 4. 设$z=2x^2+y^2$截出平面$4x+2y+z=1$上的一个椭圆，求这个椭圆的面积$A$。

**思路：** 先算出$xy$平面上投影出的椭圆面积，再还原为真实的面积。

**解答：** 联立得到
$$\begin{aligned}
    4x+2y+2x^2+y^2&=1 \\
    2(x^2+2x+1)+(y^2+2y+1)&=4 \\
    \frac{(x+1)^2}{2}+\frac{(y+1)^2}{4}&=1
\end{aligned}$$
因此$xy$平面上投影椭圆的面积为
$$\begin{aligned}
    |D_{xy}|&=\pi\cdot\sqrt{2}\cdot2 \\
    &=2\sqrt{2}\pi
\end{aligned}$$
平面的法向量$\vec{\mathbf{n}}=(4, 2, 1)$，而$xy$平面的法向量就是$z$轴，其方向向量为$\vec{\mathbf{k}}=(0, 0, 1)$。因此，平面与$xy$平面的夹角$\gamma$满足
$$\begin{aligned}
    \cos\gamma&=\frac{\vec{\mathbf{n}}\cdot\vec{\mathbf{k}}}{|\vec{\mathbf{n}}||\vec{\mathbf{k}}|} \\
    &=\frac{1}{\sqrt{21}}
\end{aligned}$$
因此实际面积为
$$\begin{aligned}
    |D_{xy}|&=|\cos\gamma|\cdot A \\
    A&=\sqrt{21}\times 2\sqrt{2}\pi \\
    &=2\sqrt{42}\pi
\end{aligned}$$

**注：** 本题的关键点是如何通过投影面积得到实际面积，因此需要掌握以下两个公式。
> $$\cos\gamma=\frac{\vec{\mathbf{n}}\cdot\vec{\mathbf{k}}}{|\vec{\mathbf{n}}||\vec{\mathbf{k}}|}$$
> $$|D_{xy}|=|\cos\gamma|\cdot A$$

> 6. 设$a>0, b>0, c>0$，证明：单叶双曲面$\Sigma:\frac{x^2}{a^2}+\frac{y^2}{b^2}-\frac{z^2}{c^2}=1$上任意一点，都有经过该点并完全位于$\Sigma$上的直线。

**思路：** 使用设直线的方法显然很麻烦，不过这种题还是要先看一遍答案，再遇见就会了。

**解答：** 对$\Sigma$的方程进行变形
$$\begin{aligned}
    \frac{x^2}{a^2}+\frac{y^2}{b^2}-\frac{z^2}{c^2}&=1 \\
    \frac{x^2}{a^2}-\frac{z^2}{c^2}&=1-\frac{y^2}{b^2} \\
    (\frac{x}{a}+\frac{z}{c})(\frac{x}{a}-\frac{z}{c})&=(1+\frac{y}{b})(1-\frac{y}{b})
\end{aligned}$$
于是有比例式
$$\frac{\frac{x}{a}+\frac{z}{c}}{1+\frac{y}{b}}=\frac{\frac{x}{a}-\frac{z}{c}}{1-\frac{y}{b}}=\lambda$$
由此得到
$$\begin{cases}
    \frac{x}{a}+\frac{z}{c}&=\lambda(1+\frac{y}{b}) \\
    \frac{x}{a}-\frac{z}{c}&=\lambda(1-\frac{y}{b})
\end{cases}$$
对任意$\lambda$，方程组都代表一条直线。且由上述变形可知直线完全位于$\Sigma$上。这也表明，对于$\Sigma$上的任意点$(x_0, y_0, z_0)$，总存在$\lambda_0$，使得线性方程组成立。也就是说，$(x_0, y_0, z_0)$就位于直线上，且这条直线完全位于$\Sigma$上。

**注：** 如果使用另一种比例式
$$\frac{\frac{x}{a}+\frac{z}{c}}{1-\frac{y}{b}}=\frac{\frac{x}{a}-\frac{z}{c}}{1+\frac{y}{b}}=\mu$$
就能推出代表另一种直线的方程组。于是，单叶双曲面实际上也是直纹面，即其上每一点都有两条经过该点并完全位于曲面上的直线。

> 7. 求直线$l_1:\begin{cases} x-y&=0 \\ z&=0 \end{cases}$与直线$l_2:\frac{x-2}{4}=\frac{y-1}{-2}=\frac{z-3}{-1}$的距离。

**思路：** 直接使用两直线之间距离公式即可。

**解答：** $l_1$过点$P_1(0, 0, 0)$，方向向量为$\vec{d}_1=(1,1,0)$。$l_2$过点$P_2:(2, 1, 3)$，方向向量为$\vec{d}_2=(4, -2, -1)$。首先计算两个向量的外积

    &=(-1, 1, -6)
\end{aligned}$$
且$\vec{P_1P_2}=(2, 1, 3)$，因此由两直线之间距离公式得到

    &=\frac{19}{\sqrt{38}} \\
    &=\sqrt{\frac{19}{2}}
\end{aligned}$$

> 8. 已知四面体的六条棱长，计算它的体积称之为欧拉六面体问题。设四面体$OABC$，它的六个棱长为$l, m, n, p, q, r$。证明：该四面体的体积为
> $$V=\frac{1}{6}\begin{vmatrix}
    p^2 & \frac{p^2+q^2-n^2}{2} & \frac{p^2+r^2-m^2}{2} \\
    \frac{p^2+q^2-n^2}{2} & q^2 & \frac{q^2+r^2-l^2}{2} \\
    \frac{p^2+r^2-m^2}{2} & \frac{q^2+r^2-l^2}{2} & r^2
\end{vmatrix}^{\frac{1}{2}}$$

**思路：** 四面体体积与向量的混合积有关。另外，行列式内的项形式有点像余弦定理。

**解答：** 设$O(0, 0, 0), A(a_1, a_2, a_3), B(b_1, b_2, b_3), C(c_1, c_2, c_3)$。设$|OA|=p, |OB|=q, |OC|=r, |BC|=l, |AC|=m, |AB|=n$。由向量混合积与四面体体积之间的关系
$$\begin{aligned}
    V&=\frac{1}{6}\begin{vmatrix}
        a_1 & a_2 & a_3 \\
        b_1 & b_2 & b_3 \\
        c_ 1& c_2 & c_3
    \end{vmatrix} \\
    V^2&=\frac{1}{36}\begin{vmatrix}
        a_1 & a_2 & a_3 \\
        b_1 & b_2 & b_3 \\
        c_ 1& c_2 & c_3 
    \end{vmatrix}\cdot\begin{vmatrix}
        a_1 & b_1 & c_1 \\
        a_2 & b_2 & c_2 \\
        a_3 & b_3 & c_3
    \end{vmatrix} \\
    &=\frac{1}{36}\begin{vmatrix}
        a_1^2+a_2^2+a_3^2 & a_1b_1+a_2b_2+a_3b_3 & a_1c_1+a_2c_2+a_3c_3 \\
        a_1b_1+a_2b_2+a_3b_3 & b_1^2+b_2^2+b_3^2 & b_1c_1+b_2c_2+b_3c_3 \\
        a_1c_1+a_2c_2+a_3c_3 & b_1c_1+b_2c_2+b_3c_3 & c_1^2+c_2^2+c_3^2
    \end{vmatrix}
\end{aligned}$$
注意到
$$\begin{aligned}
    a_1^2+a_2^2+a_3^2&=|OA|^2=p^2 \\
    b_1^2+b_2^2+b_3^2&=|OB|^2=q^2 \\
    c_1^2+c_2^2+c_3^2&=|OC|^2=r^2 \\
    a_1b_1+a_2b_2+a_3b_3&=\vec{OA}\cdot \vec{OB} \\
    &=|OA|\cdot|OB|\cos\theta \\
    &=|OA|\cdot|OB|\cdot\frac{|OA|^2+|OB|^2-|AB|^2}{2\cdot|OA|\cdot|OB|} \\
    &=\frac{p^2+q^2-n^2}{2} \\
    a_1c_1+a_2c_2+a_3c_3&=\vec{OA}\cdot\vec{OC} \\
    &=|OA|\cdot|OC|\cdot\cos\theta \\
    &=\frac{p^2+r^2-m^2}{2} \\
    b_1c_1+b_2c_2+b_3c_3&=\vec{OB}\cdot\vec{OC} \\
    &=|OB|\cdot|OC|\cdot\cos\theta \\
    &=\frac{q^2+r^2-l^2}{2}
\end{aligned}$$
代入上式，并取平方根可得结论。

> 12. 设有两条异面直线，从其中一条直线的任意点向另一条直线作垂线，据此写出这些垂线生成的曲面的参数方程，并判断此曲面的类型。

**思路：** 不妨令其中一条直线是一条特殊的直线，例如$x$轴，然后从另一条直线的任意点向$x$轴作垂线。

**解答：** 不妨设其中一条直线是$x$轴，另一条直线$L$向$x$轴作垂线，其中$L$的方向向量为$(l, m, n)\neq \mathbf{0}$，且经过点$(x_0, y_0, z_0)$。由此可以写出$L$的参数方程
$$L:\begin{cases}
    x&=x_0+ls \\
    y&=y_0+ms \\
    z&=z_0+ns
\end{cases}, s\in\R$$
过$L$上某点$(\bar{x}, \bar{y}, \bar{z})$作垂直于$x$轴的平面，它与$x$轴的交点为$(\bar{x}, 0, 0)$。因此，连接$(\bar{x}, \bar{y}, \bar{z})$与$(\bar{x}, 0, 0)$的直线$l_0$就是所作的垂线。可以写出$l_0$的参数方程为
$$l_0:\begin{cases}
    x&=\bar{x} \\
    y&=-\bar{y}t \\
    z&=-\bar{z}t
\end{cases}, t\in\R$$
将$L$的方程代入，可以得到曲面$\Sigma$的参数方程为
$$\Sigma:\begin{cases}
    x&=x_0+ls \\
    y&=-(y_0+ms)t \\
    z&=-(z_0+ns)t
\end{cases}, s, t\in\R$$
现在将$s, t$消掉，就能得到直角坐标系下$\Sigma$的方程。若$l=0$，则$L$与$x$轴垂直。此时，$L$上任意点到$x$轴的垂线都交于一点$(u, 0, 0)$，此时曲面即为$x=u$。若$l\neq 0$，则由后两式可得
$$\frac{y}{z}=\frac{y_0+ms}{z_0+ns}$$
而$s=\frac{x-x_0}{l}$，代入可得
$$\begin{aligned}
    \frac{y}{z}&=\frac{y_0+\frac{m}{l}(x-x_0)}{z_0+\frac{n}{l}(x-x_0)} \\
    y(lz_0+nx-nx_0)&=z(ly_0+mx-mx_0)
\end{aligned}$$
该方程实际上代表了一个双曲抛物面。例如，当$l=m=1, n=0, (x_0, y_0, z_0)=(0,0,1)$时，可得方程$y=zx$。

> 13. 半顶角为$\alpha$的圆锥面的方程可写为$z=b\sqrt{x^2+y^2}$，其中$b=\cot\alpha(0<\alpha<\frac{\pi}{2})$。设$P$是圆锥面上的一个点，它的球坐标为$(r, \alpha, \theta)$。在圆锥面与$xOz$坐标面的交线上取点$A$，沿过点$A$的母线将圆锥面割开后铺平，并建立极坐标系$\rho-\tau$，这时点$A, P$分别对应点$A', P'$。
>
> $(1)$求圆锥面铺平后扇形的圆心角$\beta$。
>
> $(2)$若$P'$的极坐标为$(\rho, \tau)$，求它与$P$的球坐标$(r, \alpha, \theta)$的关系。

**思路：** 注意到平面$z=a$与圆锥面的交线为一个圆，然后就非常好处理了。

**解答：** $(1)$以平面$z=a$截该圆锥面，得到的曲线是一个圆，方程为
$$\begin{aligned}
    a&=\cot\alpha(\sqrt{x^2+y^2}) \\
    x^2+y^2&=\frac{a^2}{\cot^2\alpha}
\end{aligned}$$
即圆的半径$r=\frac{a}{\cot\alpha}$，由此得出周长为$C=\frac{2\pi a}{\cot\alpha}$。剪开后，扇形的半径
$$\begin{aligned}
    R&=\sqrt{x^2+y^2+a^2} \\
    &=\sqrt{a^2(1+\frac{1}{\cot^2\alpha})} \\
    &=a\sec\alpha
\end{aligned}$$
因此，圆心角$\beta$为
$$\begin{aligned}
    \beta&=\frac{C}{R} \\
    &=\frac{2\pi a}{\cot\alpha}\cdot\frac{1}{a\sec\alpha} \\
    &=2\pi a\cdot\frac{\sin\alpha}{\cos\alpha}\cdot\frac{1}{a}\cdot\cos\alpha \\
    &=2\pi\sin\alpha
\end{aligned}$$

$(2)$ 可以确定的是$\rho=r$，因为是沿着母线割开的。然后探讨角度之间的关系。为求方便，直接假设$A, B$两点都在被截出的圆上，两者在圆上形成的圆心角为$\theta$。因此，$|\stackrel\frown{AB}|=\frac{\theta a}{\cot\alpha}$。在扇形上两者形成的圆心角就是$\tau$，因此有关系式
$$\begin{aligned}
    R\tau&=\frac{\theta a}{\cot\alpha} \\
    \tau&=\theta a\cdot\frac{\sin\alpha}{\cos\alpha}\cdot\frac{\cos\alpha}{a} \\
    &=\theta\sin\alpha
\end{aligned}$$

> 14. 一只蚂蚁在圆锥面$z=\sqrt{3x^2+3y^2}$上从点$A(1,0,\sqrt{3})$爬到点$B(-1,0,\sqrt{3})$。问它沿何种路径可使得爬行的路程最短，并求最短路程。

**思路：** 还是先剪开，求出平面上的路径，再转换到圆锥面上。但是比起例题的圆柱面，圆锥面的计算更麻烦。

**解答：** 设原点为$O(0,0,0)$，将圆锥沿着母线$OA$剪开，并铺平到以$O'$为原点，以$O'A'$为极轴的极坐标系平面上。显然$|O'A'|=|OA|=2$，因此$A$的极坐标为$(2, 0)$。由上题结论，扇形的圆心角$\beta=2\pi\sin\alpha$。其中
$$\begin{aligned}
    \cot\alpha&=\sqrt{3} \\
    \alpha&=\frac{\pi}{6} \\
    \sin\alpha&=\frac{1}{2}
\end{aligned}$$
因此圆心角$\beta=\pi$。注意到在圆锥面上时，$\angle AOB=\pi$，因此$\angle A'OB'=\frac{\pi}{2}$，$B$的极坐标为$(2,\frac{\pi}{2})$。将极坐标转换为直角坐标就有$A(2, 0), B(0, 2)$，这两点的最短路径是
$$l:\begin{cases}
    x&=2-t \\
    y&=t
\end{cases},t\in[0, 2]$$
路程为$s=\sqrt{2^2+2^2}=2\sqrt{2}$。容易将线段转换为极坐标形式
$$\begin{aligned}
    r\cos\tau&=2-r\sin\tau \\
    \sqrt{2}r\sin(\tau+\frac{\pi}{4})&=1 \\
    r&=\frac{1}{\sqrt{2}\sin(\tau+\frac{\pi}{4})}
\end{aligned}$$
其中$\tau\in[0,\frac{\pi}{2}]$。转换回圆锥面时，我们有
$$x=r\sin\alpha\cos\theta,y=r\sin\alpha\sin\theta, z=r\cos\alpha$$
且$\alpha=\arccos(\frac{\sqrt{3}}{2})=\frac{\pi}{6},\tau=\theta\sin\alpha=\frac{\theta}{2}$。由此得到
$$r=\frac{2}{\sin\tau+\cos\tau}=\frac{2}{\sin\frac{\theta}{2}+\cos\frac{\theta}{2}}$$
最后得到
$$x=\frac{\cos\theta}{\sin\frac{\theta}{2}+\cos\frac{\theta}{2}}, y=\frac{\sin\theta}{\sin\frac{\theta}{2}+\cos\frac{\theta}{2}},z=\frac{\sqrt{3}}{\sin\frac{\theta}{2}+\cos\frac{\theta}{2}}(\theta\in[0,\pi])$$

> 15. 证明：参数方程
> $$\begin{cases}
    x=a+r\cos\varphi\cos\theta\cos t-r\sin\theta\sin t \\
    y=b+r\cos\varphi\sin\theta\cos t-r\cos\theta\sin t \\
    z=c-r\sin\varphi\cos t
\end{cases},t\in[0,2\pi]$$
> 表示的曲线是空间中圆心在$(a, b, c)$半径为$r$的圆$L$，并指出常数$\varphi,\theta$的几何意义。

**提示：** 需要证明曲线是球面与平面的交线。

**解答：** 注意到
$$\begin{aligned}
    &(x-a)^2+(y-b)^2+(z-c)^2 \\
    &=[r(\cos\varphi\cos\theta\cos t-\sin\theta\sin t)]^2+[r(\cos\varphi\sin\theta\cos t+\cos\theta\sin t)]^2+(r\sin\varphi\cos t)^2 \\
    &=r^2(\cos^2\varphi\cos^2\theta\cos^2t+\sin^2\theta\sin^2t+\cos^2\varphi\sin^2\theta\cos^2t+\cos^2\theta\sin^2t+\sin^2\varphi\cos^2t) \\
    &=r^2(\cos^2\varphi\cos^2t+\sin^2t+\sin^2\varphi\cos^2t) \\
    &=r^2
\end{aligned}$$
这表明曲线上的点都位于球心为$(a, b, c)$，半径为$r$的球面上。然后注意到
$$\begin{aligned}
    &(x-a)\sin\varphi\cos\theta+(y-b)\sin\varphi\sin\theta+(z-c)\cos\varphi \\
    &=r(\cos\varphi\sin\varphi\cos^2\theta\cos t-\sin\varphi\sin\theta\cos\theta\sin t+\cos\varphi\sin\varphi\sin^2\theta\cos t+\sin\varphi\sin\theta\cos\theta\sin t-\sin\varphi\cos\varphi\cos t) \\
    &=r(\cos\varphi\sin\varphi\cos t-\cos\varphi\sin\varphi\cos t) \\
    &=0
\end{aligned}$$
这表明曲线上的点也都位于平面$(x-a)\sin\varphi\cos\theta+(y-b)\sin\varphi\sin\theta+(z-c)\cos\varphi$上，而平面与球面的交线就是一个圆。平面的单位法向量为$\mathbf{e}=(\sin\varphi\cos\theta,\sin\varphi\sin\theta,\cos\varphi)$。将圆心移到原点，观察到平面法向量$\mathbf{e}$就是圆的朝向，$\varphi$是$\mathbf{e}$与$z$轴的夹角，$\theta$是$\mathbf{e}$在$xOy$平面上的投影与$x$轴的夹角。

> 17. $(1)$在两条异面的线段$AB, CD$上随意各取一个点，作连接这两个点的线段。判断所有这些线段上的点构成的几何形状（假定两条线段互相垂直，且都被它们的公垂线平分）。
> 
> $(2)$问如果将$AB, CD$换成两条异面直线，图形又如何。

**思路：** 如果某个点$P$在几何形状上，则必有一条过点$P$且与$AB, CD$都相交的直线。

**解答：** $(1)$ 不失一般性，设$A(0, b, h), B(0, -b, h),C(-a, 0, 0), D(a, 0, 0)$，其中$h\neq 0$。记构成的几何形状为$\Omega$，设点$M(x, y, z)$在$\Omega$上，就必有一条通过$M$且与$AB, CD$都相交的直线。设交点分别为$P_1(0, u, h), P_2(v, 0, 0)$，就有$-b\le u\le b, -a\le v\le a$。显然$\vec{P_1M}$平行于$\vec{P_2M}$，因此
$$\frac{x}{x-v}=\frac{y-u}{y}=\frac{z-h}{z}$$
于是有方程组
$$\begin{cases}
    x&=\frac{z-h}{z}(x-v) \\
    y-u&=\frac{z-h}{z}y
\end{cases}$$
化简得到
$$\begin{cases}
    u&=\frac{h}{z}y \\
    v&=-\frac{h}{z-h}x
\end{cases}$$
由此得到不等式组
$$\begin{cases}
    -b\le\frac{h}{z}y\le b \\
    a\le \frac{h}{z-h}x\le -a
\end{cases}$$
注意到$0\le z\le h$，于是可以解得
$$\begin{cases}
    -bz\le hy\le bz \\
    a(z-h)\le hx\le -a(z-h)
\end{cases}$$
将其转写成四条等式，就得到四个平面方程：$hy-bz=0,hy+bz=0,hx-a(z-h)=0, hx+a(z-h)=0$。$\Omega$就是这四个平面所界定的四面体。

$(2)$ 将等式依次改写为$z=\frac{h}{b}y, z=-\frac{h}{b}y, z=\frac{h}{a}x+h, z=-\frac{h}{a}x+h$，令$a, b\to+\infty$，就知道$\Omega$是由$z=0$和$z=h$所界定的空间区域。

> 18. 给定二次曲面$(x+y)^2+(y+z)^2+(z+x)^2-1=0$，求它在$xOy$坐标面上的投影区域$D$。

**思路：** 把曲面方程看作$z$的一元二次方程，利用判别式即可。

**解答：** 对曲面方程进行整理得到
$$\begin{aligned}
    x^2+2xy+y^2+y^2+2yz+z^2+z^2+2xz+x^2-1&=0 \\
    2z^2+2(x+y)z+(2x^2+2y^2+2xy-1)&=0
\end{aligned}$$
这可看作一个关于$z$的一元二次方程。要使得$z$有解，必须有判别式$\Delta\ge 0$，也就是
$$\begin{aligned}
    [2(x+y)]^2-4\times 2(2x^2+2y^2+2xy-1)&\ge 0 \\
    (x^2+2xy+y^2)-(4x^2+4y^2+4xy-2)&\ge 0 \\
    3x^2+3y^2+2xy&\le 2
\end{aligned}$$
因此，投影区域$D$就是$3x^2+2xy+3y^2\le 2$。

> 19. 设有限长的光滑曲线$L$在$x, y$轴上的投影区间的长度分别为$a, b$。证明：$|L|^2\ge a^2+b^2$，等号成立当且仅当$L$是直线段，其中$|L|$是$L$的长度。

**提示：** 研究连接$L$的上、下、左、右端点所形成的折线段，并利用勾股定理。

**解答：** 设$L$以$P, Q$为端点，$A, B, C, D$分别是$L$的左、上、右、下端点，不妨设$A$的横坐标是0，$C$的横坐标是$a$，$B$的纵坐标是$b$，$D$的纵坐标是0。那么有
$$\begin{aligned}
    |L|&=|\stackrel{\frown}{PQ}| \\
    &\ge |\stackrel{\frown}{AD}| \\
    &\ge |AB|+|BC|+|CD|
\end{aligned}$$
设$|AB|=u=\sqrt{u_x^2+u_y^2}$，其中$u_x,u_y$分别为$AB$在$x, y$轴上的投影长度，类似定义$|BC|=v, |CD|=w$。那么就有
$$\begin{aligned}
    |L|^2&\ge (u+v+w)^2 \\
    &=u^2+v^2+w^2+2(uv+vw+wu) \\
    &\ge u_x^2+v_x^2+v_y^2+w_y^2+2u_xv_x+2v_yw_y \\
    &=(u_x+v_x)^2+(v_y+w_y)^2 \\
    &=a^2+b^2
\end{aligned}$$
若$L$正好是连接$P, Q$的线段，则等号显然成立，此时$|L|$取得最小值。由于连接$P, Q$两点最短的曲线是线段$PQ$，因此$L$是直线段。
$$\begin{aligned}
    |L|^2&\ge a^2+b^2 \\
    &=(u_x+v_x)^2+(v_y+w_y)^2 \\
    &=u_x^2+v_x^2+v_y^2+w_y^2+2u_xv_x+2v_yw_y
\end{aligned}$$

> 20. 将函数$z(x, y)=3|2x-y|+2|x+y-3|$表达为不带绝对值的函数形式。

**思路：** 只需讨论绝对值内部什么时候取正，什么时候取负即可。

```python
%matplotlib inline
x = np.linspace(-5, 5, 1000)
y1 = 2*x 
y2 = 3-x
plt.plot(x, y1, label=r'$2x-y=0$')
plt.plot(x, y2, label=r'$x+y-3=0$')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

**解答：** 设$f(x, y)=2x-y, g(x, y)=x+y-3$，则$l_1:f(x, y)=0, l_2:g(x, y)=0$这两条直线将$xy$平面分成了四部分。在第一部分，$f, g>0$，此时有
$$\begin{cases}
    2x-y&\ge 0 \\
    x+y-3&\ge 0
\end{cases}$$
而且
$$\begin{aligned}
    z(x, y)&=3(2x-y)+2(x+y-3) \\
    &=8x-y-6
\end{aligned}$$
类似地，可以写出其他区域上$z(x, y)$的表达式和限制条件。最终结果为
$$z(x, y)=\begin{cases}
    8x-y-6, &f, g>0 \\
    4x-5y+6, &f>0, g<0 \\
    -4x+5y-6, &f<0, g>0 \\
    -8x+y+6, &f, g<0
\end{cases}$$
另外，在$l_1$上，我们有
$$z(x, y)=\begin{cases}
    6x-6, &x\ge 1 \\
    6-6x, &x<1
\end{cases}$$
在$l_2$上，我们有
$$z(x, y)=\begin{cases}
    9x-9, &x\ge 1 \\
    9-9x, &x<1
\end{cases}$$
