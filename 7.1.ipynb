## 7.1 初等积分法与线性方程

### 一阶微分方程

1. 可变量分离的方程：指可整理为以下形式的方程，两边积分可得通解。
$$f(x)\text{d}x=g(y)\text{d}y$$

2. 齐次方程：标准形式为$y'=f(\frac{y}{x})$。令$u=\frac{y}{x}$，则$y'=xu'+u$，因此原式变为以下形式(可分离变量)
$$xu'+u=f(u)$$

3. 一阶线性方程：有下列标准形式
$$\frac{\text{d}y}{\text{d}x}+P(x)y=Q(x)$$
其中$P(x), Q(x)$为`连续函数`，方程的通解为
$$y=\text{e}^{-\int P(x)\text{d}x}(\int Q(x)\text{e}^{\int P(x)\text{d}x}\text{d}x+C)$$

4. 伯努利方程：有下列标准形式
$$\frac{\text{d}y}{\text{d}x}+P(x)=Q(x)y^n$$
其中$P(x), Q(x)$为`连续函数`。令$z=y^{1-n}$可计算得
$$\frac{\text{d}z}{\text{d}x}+(1-n)P(x)z=(1-n)Q(x)$$

5. 全微分方程：若一阶方程$M(x, y)\text{d}x+N(x, y)\text{d}y=0$的左端为某二元函数$u(x, y)$的全微分，即$\text{d}u=M(x, y)\text{d}x+N(x, y)\text{d}y$，则称该方程为全微分方程或恰当方程。若$M(x, y)\text{d}x+N(x, y)\text{d}y=0$不是恰当方程，但可找到一个连续可微函数$\mu(x, y)\neq 0$使得$\mu(x, y)M(x, y)\text{d}x+\mu(x, y)N(x, y)\text{d}y=0$为恰当方程，则称$\mu(x, y)$为该方程的积分因子。若$\mu M\text{d}x+\mu N\text{d}y=\text{d}U$，则$U(x, y)=C$是该方程的通解。

### 可降阶的高阶方程

1. 形如$y^{(n)}=f(x)$的方程，只需求$n$次积分即可。

2. 形如$F(x, y', y'')=0$的方程，要求方程不含未知函数$y$。令$y'=p(x)$，则该方程化为一阶方程$F(x, p, p')=0$。

3. 形如$F(y, y', y'')=0$的方程，要求方程不含自变量$x$。将$y$视为自变量，令$y'=p(y)$，则
$$y''=\frac{\text{d}p}{\text{d}x}=\frac{\text{d}p}{\text{d}y}\cdot \frac{\text{d}y}{\text{d}x}=p\frac{\text{d}p}{\text{d}y}$$
则该方程化为一阶方程$F(y, p, p\frac{\text{d}p}{\text{d}y})=0$

### 线性微分方程

形如
$$\frac{\text{d}^ny}{\text{d}x^n}+a_1(x)\frac{\text{d}^{n-1}y}{\text{d}x^{n-1}}+\cdots+a_{n-1}(x)\frac{\text{d}y}{\text{d}x}+a_n(x)y=f(x)$$
的方程称为$n$阶线性方程，其中系数$a_i(x)(i=1, 2, \cdots, n)$及自由项$f(x)$均为定义在区间$I$上的连续函数。此时对任意$x_0\in I$及任意$\bar{y}_0, \bar{y}_1, \cdots, \bar{y}_{n-1}$，上面的线性微分方程在$I$上存在唯一解$y=\varphi(x)$，满足初始条件
$$\varphi(x_0)=\bar{y}_0, \varphi'(x_0)=\bar{y}_1, \cdots, \varphi^{(n-1)}(x_0)=\bar{y}_{n-1}$$
当$f(x)\equiv 0$时，称为$n$阶齐次线性微分方程，简称齐次线性方程。否则称为非齐次线性微分方程。

1. 设$y_1(x), y_2(x), \cdots, y_k(x)$均为齐次线性方程的解，则
$$y=C_1y_1(x)+C_2y_2(x)+\cdots+C_ky_k(x)$$
也是该齐次线性方程的解，其中$C_1, C_2, \cdots, C_k$为任意常数。

2. $n$阶齐次线性方程一定存在$n$个线性无关的解$y_1(x), y_2(x), \cdots, y_n(x)$，且通解可表示为
$$y=C_1y_1(x)+C_2y_2(x)+\cdots+C_ny_n(x)$$
其中$C_1, C_2, \cdots, C_n$为任意常数。

3. 对于非齐次线性方程，若$\tilde{y}(x)$为非齐次线性方程的某个特解，$y(x)$是对应齐次线性方程的通解，则$Y(x)=\tilde{y}(x)+y(x)$是非齐次线性方程的通解。非齐次线性方程的任意两个解之差是对应的齐次线性方程的解。这两点从定义来看很直观。

### 常系数齐次线性微分方程

$$\frac{\text{d}^ny}{\text{d}x^n}+a_1\frac{\text{d}^{n-1}y}{\text{d}x^{n-1}}+\cdots+a_{n-1}\frac{\text{d}y}{\text{d}x}+a_n(x)y=0$$
其中$a_i(i=1, 2, \cdots, n)$为实常数。令$n$次多项式$I(\lambda)=\lambda^n+a_1\lambda^{n-1}+\cdots+a_{n-1}\lambda+a_n=0$，称为上述方程的特征方程，它的根称为特征根。每一个特征根对应着若干个线性无关的解，所以不同的特征根对应的线性无关的解就可以组成常系数齐次线性微分方程的$n$个线性无关的解。

| 特征方程的根 | 齐次方程对应的解 |
| --- | --- |
| 单实根$\lambda$ | 有一个解$\text{e}^{\lambda x}$ |
| 一对单复根$\lambda_{1, 2}=\alpha\pm\beta\text{i}$ | 有两个线性无关的解$y_1=\text{e}^{\alpha x}\cos\beta x, y_2=\text{e}^{\alpha x}\sin\beta x$ |
| $k$重实根$\lambda$ | 有$k$个线性无关的解$y_1=\text{e}^{\lambda x}, y_2=x=\text{e}^{\lambda x}, \cdots, y_k=x^{k-1}=\text{e}^{\lambda x}$ |
| 一对$k$重复根$\lambda_{1, 2}=\alpha\pm\beta\text{i}$ | 有$2k$个线性无关的解$$y_1=\text{e}^{\alpha x}\cos\beta x, y_2=x\text{e}^{\alpha x}\cos\beta x, \cdots, y_k=x^{k-1}\text{e}^{\alpha x}\cos\beta x \\ \bar{y}_1=\text{e}^{\alpha x}\sin\beta x, \bar{y}_2=x\text{e}^{\alpha x}\sin\beta x, \cdots, \bar{y}_k=x^{k-1}\text{e}^{\alpha x}\sin\beta x$$ |

> 注：只需要记忆实根的情况，因为复根可运用欧拉公式$\text{e}^{\alpha+\beta\text{i}}=\text{e}^{\alpha}(\cos\beta+\text{i}\sin\beta)$推出。

### 常系数非齐次线性微分方程的算子解法

设常系数非齐次线性微分方程为
$$y^{(n)}+a_1y^{(n-1)}+\cdots+a_{n-1}y'+a_ny=q(x)$$
作如下规定：
$$D^k=\frac{\text{d}^k}{\text{d}x^k}, D^ky=\frac{\text{d}^ky}{\text{d}x^k}=y^{(k)}$$
注意$D^0=1, D^0y=y$。上式可表达为如下形式
$$f(D)y=D^ny+a_1D^{n-1}y+\cdots+a_{n-1}Dy+a_ny=q(x)$$
其中
$$f(D)=D^n+a_1D^{n-1}+\cdots+a_{n-1}D+a_n$$
称为形式的算子多项式。例如，微分方程$y''+y=\cos x$可写为$(D^2+1)y=\cos x$。算子和算子多项式可理解为函数集合到函数集合到映射，有时候也称$f(D)y$为算子$f(D)$作用到函数$y=y(x)$上。这是一种形式上的乘法，具有线性性质，满足通常多项式的一切运算规则，甚至可以进行因式分解。

设$f(D)$是算子多项式，若$f(D)\varphi(x)=q(x)$，则记$\varphi(x)=\frac{1}{f(D)}q(x)$，称算子$\frac{1}{f(D)}$为$f(D)$的逆算子，这是解方程时作的形式除法。例如，因为$D(x^2)=2x$，所以$x^2=\frac{1}{D}(2x)=\int 2x\text{d}x$。一般来说，我们有
$$\frac{1}{D}q(x)=\int q(x)\text{d}x, \frac{1}{D^k}[q(x)]=\int\cdots\int q(x)(\text{d}x)^k$$
逆算子也具有相同的线性运算性质。而且，逆算子还有与实、虚部的如下交换公式：设$u(x), v(x)$为实函数，$\text{i}=\sqrt{-1}$，分别用$\text{Re}z, \text{Im}z$表示复数$z$的实部与虚部，则有
$$\frac{1}{f(D)}\text{Im}[u(x)+\text{i}v(x)]=\text{Im}(\frac{1}{f(D)}[u(x)+\text{i}v(x)])$$
$$\frac{1}{f(D)}\text{Re}[u(x)+\text{i}v(x)]=\text{Re}(\frac{1}{f(D)}[u(x)+\text{i}v(x)])$$

将非齐次方程写为$f(D)y=q(x)$后，特解可表示为$y=\frac{1}{D}q(x)$
1. 若$q(x)=P_n(x)$为$n$次多项式，可将$\frac{1}{D}$展开为形式的泰勒级数$\frac{1}{D}=b_0+b_1D+\cdots+b_nD^n$(取到$D^n$项为止)，此时特解$y^*=(b_0+b_1D+\cdots+b_nD^n)q(x)$

2. 若$q(x)=\text{e}^{\lambda x}v(x)$，其中$\lambda$可以是复数，$v(x)$是实函数时，我们有公式
$$\frac{1}{f(D)}\text{e}^{\lambda x}v(x)=\text{e}^{\lambda x}\frac{1}{f(D+\lambda)}v(x)$$

当$q(x)$是三角函数时，可以通过欧拉公式来转化为指数形式求解。

### 欧拉方程

形式为
$$x^ny^{(n)}+p_1x^{n-1}y^{(n-1)}+\cdots+p_{n-1}xy'+p_ny=f(x)$$
其中$p_1, p_2, \cdots, p_n$都是常数。作代换$x=\text{e}^t$，记$D=\frac{\text{d}}{\text{d}t}$，则有
$$x^ky^{(k)}=D(D-1)\cdots(D-k+1)y$$
于是欧拉方程可化为常系数线性方程。

### 微分方程组

由数个微分方程联立得到的方程组称为微分方程组。若方程组中每个方程都是常系数线性微分方程，则称为常系数线性微分方程组，可以通过消元法转换为高阶线性微分方程来求解。(主要还是用算子法比较方便)

### 例题

> 1. 解微分方程
> $$y(y+1)\text{d}x+[x(y+1)+x^2y^2]\text{d}y=0$$

**解答：** 尽量尝试凑微分
$$\begin{aligned}
(y+1)(y\text{d}x+x\text{d}y)+(x^2y^2)\text{d}y&=0 \\
(y+1)\text{d}(xy)+(x^2y^2)\text{d}y&=0 \\
\frac{\text{d}(xy)}{(xy)^2}+\frac{\text{d}y}{y+1}&=0 \\
\text{d}(-\frac{1}{xy}+\ln|y+1|)&=0 \\
\end{aligned}$$ 
因此，通解为
$$-\frac{1}{xy}+\ln|y+1|=C_1$$
$$y+1=C\text{e}^{\frac{1}{xy}}$$
其中$C=\pm \text{e}^{C_1}$

**注解：** 此题并非恰当方程，于是推导过程中乘了积分因子$\frac{1}{x^2y^2(y+1)}$。选择这个积分因子是为了凑微分，这个方法还是比较看感觉和熟练度的。

> 2. 求微分方程
> $$(1+y^2)\text{d}x+(x-\arctan y)\text{d}y=0$$
> 的通解。

**解答：** 微分方程可变形为
$$\begin{aligned}
\frac{\text{d}x}{\text{d}y}&=\frac{\arctan y-x}{1+y^2} \\
\frac{\text{d}x}{\text{d}y}+\frac{1}{1+y^2}x&=\frac{\arctan y}{1+y^2}
\end{aligned}$$
这是关于未知函数$x$的一阶线性微分方程，直接套公式得到
$$\begin{aligned}
x&=\text{e}^{-\int \frac{1}{1+y^2}\text{d}y}(\int \frac{\arctan y}{1+y^2}\text{e}^{\int \frac{1}{1+y^2}\text{d}y}\text{d}y+C) \\
&=\text{e}^{-\arctan y}(\int \arctan y\text{e}^{\arctan y}\text{d}(\arctan y)+C) \\
&=\text{e}^{-\arctan y}(\arctan y\text{e}^{\arctan y}-\text{e}^{\arctan y}+C) \\
&=C\text{e}^{-\arctan y}+\arctan y-1
\end{aligned}$$

**注解：** 什么时候会将$x$看成未知函数，而并非$y$呢？就是像本题一样，关于$x$的项形式很简单(甚至就只有$x$本身)，但关于$y$的项则很复杂的时候。

> 3. 求方程$yy''-2(y')^2=0$的通解。

**解法1:** 对方程两边同乘$1/y^3$，则方程变为
$$\begin{aligned}
\frac{yy''-2(y')^2}{y^3}&=0 \\
\frac{\text{d}}{\text{d}x}(\frac{y'}{y^2})&=0
\end{aligned}$$
因此
$$\begin{aligned}
\frac{\text{d}y}{y^2}&=-C\text{d}x \\
-\frac{1}{y}&=-Cx-D \\
y&=\frac{1}{Cx+D}
\end{aligned}$$
其中$C, D$为常数且不同时为零。

**解法2:** 令$y'=p(y)$，则$y''=\frac{\text{d}p}{\text{d}x}=\frac{\text{d}p}{\text{d}y}\cdot\frac{\text{d}y}{\text{d}x}=p\frac{\text{d}p}{\text{d}y}$，方程变为
$$\begin{aligned}
yp\frac{\text{d}p}{\text{d}y}-2p^2&=0
\end{aligned}$$
若$p\neq 0$，则两边同除以$p$
$$\begin{aligned}
y\frac{\text{d}p}{\text{d}y}&=2p \\
\frac{1}{2}\frac{\text{d}p}{p}&=\frac{\text{d}y}{y} \\
p&=C'y^2
\end{aligned}$$
从而可以解得和上面一样的结果。

**注解：** 解法2是较为通用的方法。解法1在我看来，凑微分的一步过于具有技巧性。另外，$y=0$也是方程的解，但未包含在通解之中。解题过程中，必须假定$y\neq 0$。

> 4. 求二阶非齐次线性方程
> $$y''-3y'+2y=2\text{e}^{-x}\cos x+\text{e}^{2x}(4x+5)$$
> 的通解。

**解答：** 非齐次线性方程的通解就是其特解加上对应齐次线性方程的通解。先求出对应齐次线性方程的通解，特征方程为
$$r^2-3r+2=0$$
容易知道$r_1=1, r_2=2$，从而对应齐次方程的通解为
$$\tilde{y}=C_1\text{e}^{x}+C_2\text{e}^{2x}$$
再求本方程的特解。为此，将方程拆成两个方程
$$y''-3y'+2y=2\text{e}^{-x}\cos x$$
$$y''-3y'+2y=\text{e}^{2x}(4x+5)$$
对第一个方程，设特解为$y=\text{e}^{-x}(A\cos x+B\sin x)$，则$y'=\text{e}^{-x}((B-A)\cos x+(-A-B)\sin x), y''=\text{e}^{-x}(-2B\cos x+2A\sin x)$。因此
$$\text{e}^{-x}((-2B-3B+3A+2A)\cos x+(2A+3A+3B+2B)\sin x)=2\text{e}^{-x}\cos x$$
$$\begin{cases}
5A-5B&=2 \\
5A+5B&=0
\end{cases}$$
解得$A=\frac{1}{5}, B=-\frac{1}{5}$，特解$y_1=\frac{1}{5}\text{e}^{-x}(\cos x-\sin x)$。对第二个方程，设特解为$y=\text{e}^{2x}x(ax+b)$，则$y'=\text{e}^{2x}(2ax^2+(2a+2b)x+b), y''=\text{e}^{2x}(4ax^2+(8a+4b)x+(2a+4b))$。因此
$$\text{e}^{2x}((4a-6a+2a)x^2+(8a+4b-6a-6b+2b)x+(2a+4b-3b))=\text{e}^{2x}(4x+5)$$
$$\begin{cases}
2a&=4 \\
2a+b&=5
\end{cases}$$
解得$a=2, b=1$，特解$y_2=\text{e}^{2x}(2x^2+x)$。最后，得到原方程的通解为
$$y=C_1\text{e}^{x}+C_2\text{e}^{2x}+\frac{1}{5}\text{e}^{-x}(\cos x-\sin x)+\text{e}^{2x}(2x^2+x)$$

**注解：** 当$q(x)=\text{e}^{\lambda x}(A\cos x+B\sin x)$时，可设$y^*=\text{e}^{\lambda x}(C\cos x+D\sin x)$。当$q(x)=\text{e}^{ax}v(x)$时，若$a$是特征方程的$m$重根，则设$y^*=\text{e}^{\lambda x}x^ms(x)$，其中是$s(x)$与$v(x)$同次的多项式。可与下面的算子解法比较。

**算子解法：** 求对应齐次线性方程的通解步骤同上，先求第一个方程的特解。因为$2\text{e}^{-x}\cos x=2\text{e}^{-x}\text{Re}(\text{e}^{\text{i}x})=\text{Re}(2\text{e}^{(-1+\text{i})x})$，因此转而求解
$$y''-3y'+2y=2\text{e}^{(-1+\text{i})x}$$
注意到算子多项式$f(D)$在$r=-1+\text{i}$处的值为
$$p(-1+\text{i})=(-1+\text{i})^2-3(-1+\text{i})+2=-2\text{i}+3-3\text{i}+2=5-5\text{i}\neq 0$$
因此
$$\tilde{y}=\frac{2\text{e}^{(-1+\text{i})x}}{5-5\text{i}}=\frac{1}{5}(1+\text{i})\text{e}^{(-1+\text{i})x}=\frac{1}{5}(1+\text{i})\text{e}^{-x}(\cos x+\text{i}\sin x)=\frac{1}{5}\text{e}^{-x}((\cos x-\sin x)+\text{i}(\cos x+\sin x))$$
取实部，得到
$$y_1=\frac{1}{5}\text{e}^{-x}(\cos x-\sin x)$$
对第二个方程，可以计算得到
$$\begin{aligned}
    y_2&=\frac{1}{(D-1)(D-2)}\text{e}^{2x}(4x+5) \\
    &=\text{e}^{2x}\frac{1}{D(D+1)}(4x+5) \\
    &=\text{e}^{2x}(\frac{1}{D}\frac{1}{D+1}(4x+5)) \\
    &=\text{e}^{2x}(\frac{1}{D}(1-D)(4x+5)) \\
    &=\text{e}^{2x}(\frac{1}{D}(4x+1)) \\
    &=\text{e}^{2x}(2x^2+x)
\end{aligned}$$

> 5. 用算子解法求下列常系数非齐次线性方程的特解：
> $(1):2y''+2y'+y=x^2+2x-1; (2)y''+y'=x^2+1; (3)y^{(n)}+a_1y^{(n-1)}+\cdots+a_{n-1}y'+a_ny=b$，其中$a_n\neq 0$

**解答：** $(1)$算子多项式
$$f(D)=2D^2+2D+1$$
因此特解为
$$y^*=\frac{1}{2D^2+2D+1}(x^2+2x-1)$$
将上面的逆算子展开为泰勒级数，由于$\deg q(x)=2$，因此展开到二阶即可。具体展开时可用长除法，注意应该从低次项除到高次项。结果为
$$y^*=(1-2D+2D^2)(x^2+2x-1)=x^2+2x-1-4x-4+4=x^2-2x-1$$

$(2)$算子多项式
$$f(D)=D^2+D$$
因此特解为
$$\begin{aligned}
    y^*&=\frac{1}{D(D+1)}(x^2+1) \\
    &=\frac{1}{D}[\frac{1}{1+D}(x^2+1)] \\
    &=\frac{1}{D}[(1-D+D^2)(x^2+1)] \\
    &=\frac{1}{D}(x^2+1-2x+2) \\
    &=\frac{1}{D}(x^2-2x+3) \\
    &=\frac{1}{3}x^3-x^2+3x
\end{aligned}$$
或者
$$\begin{aligned}
    y^*&=\frac{1}{D(D+1)}(x^2+1) \\
    &=\frac{1}{D}[\frac{1}{1+D}(x^2+1)] \\
    &=[\frac{1}{D}(1-D+D^2)](x^2+1) \\
    &=(\frac{1}{D}-1+D)(x^2+1) \\
    &=\frac{1}{3}x^3+3x-x^2-1
\end{aligned}$$
这里可以看出特解不是唯一的。

$(3)$算子多项式
$$f(D)=D^n+a_1D^{n-1}+\cdots+a_{n-1}D+a_n$$
因此特解为
$$y^*=\frac{1}{D^n+a_1D^{n-1}+\cdots+a_{n-1}D+a_n}(b)$$
将上面的逆算子展开为泰勒级数，由于$\deg q(x)=0$，因此展开到常数项即可。这又称为“算子常数”，结果为
$$y^*=\frac{1}{a_n}b=\frac{b}{a_n}$$

> 6. 用算子解法求下列常系数非齐次线性方程的特解：$(1):y''-2y'+y=x\text{e}^x; (2)y''-6y'+13y=\text{e}^{3x}\sin 2x$。

**解答：** $(1)$算子多项式
$$f(D)=D^2-2D+1=(D-1)^2$$
运用公式$f(D)(\text{e}^{\lambda x}q(x))=\text{e}^{\lambda x}\frac{1}{f(D+\lambda)}q(x)$，可以得到
$$y^*=\text{e}^{x}\frac{1}{(D-1+1)^2}(x)=\text{e}^{x}\frac{1}{D^2}(x)=\frac{1}{6}x^3\text{e}^{x}$$

$(2)$算子多项式
$$f(D)=D^2-6D+13$$
因为$\text{e}^{3x}\sin 2x=\text{Im}(\text{e}^{(3+2\text{i})x})$，所以转而求解
$$y''-6y'+13y=\text{e}^{(3+2\text{i})x}$$
计算$f(D)$在$r=3+2\text{i}$处的值
$$p(3+2\text{i})=(3+2\text{i})^2-6(3+2\text{i})+13=0$$
此路不通，只能按书上的方法求解
$$\begin{aligned}
y^*&=\text{e}^{3x}\frac{1}{(D+3)^2-6(D+3)+13}(\sin 2x) \\
&=\text{e}^{3x}\frac{1}{D^2+4}(\text{Im}(\text{e}^{2\text{i}x})) \\
&=\text{e}^{3x}\text{Im}(\frac{1}{D^2+4}(\text{e}^{2\text{i}x})) \\
&=\text{e}^{3x}\text{Im}(\text{e}^{2\text{i}x}\frac{1}{(D+2\text{i})^2+4}(1)) \\
&=\text{e}^{3x}\text{Im}(\text{e}^{2\text{i}x}\frac{1}{D(D+4\text{i})}(1)) \\
&=\text{e}^{3x}\text{Im}(\text{e}^{2\text{i}x}\frac{1}{D}(\frac{1}{4\text{i}})) \\
&=\text{e}^{3x}\text{Im}(\text{e}^{2\text{i}x}\frac{x}{4\text{i}}) \\
&=\text{e}^{3x}\text{Im}((\cos 2x+\text{i}\sin 2x)\frac{x}{4\text{i}}) \\
&=-\frac{x}{4}\text{e}^{3x}\cos 2x
\end{aligned}$$

> 7. 求常系数线性微分方程组
> $$\begin{cases}
x'-y'&=2t \\
x'+2y'&=3t
\end{cases}$$
> 的特解，其中$x, y$都是$t$的函数。

**解答：** 将方程组写成算子形式
$$\begin{cases}
Dx-Dy&=2t \\
Dx+2Dy&=3t
\end{cases}$$
计算得到
$$Dx=\frac{7}{3}t, Dy=\frac{1}{3}t$$
因此
$$x=\frac{1}{D}(\frac{7}{3}t)=\frac{7}{6}t^2$$
$$y=\frac{1}{D}(\frac{1}{3}t)=\frac{1}{6}t^2$$

> 8. 设$p_1(x), p_2(x)$连续，$y_1(x), y_2(x)$是齐次线性方程$y''+p_1(x)y'+p_2(x)y=0$的两个线性无关的解。
> $$W(x)=\begin{vmatrix}
y_1(x) & y_2(x) \\
y'_1(x)  & y'_2(x)
\end{vmatrix}$$
> 称为该方程的朗斯基行列式。证明在任何区间$I$上$W(x)\neq 0$。

**解答：** 容易计算得到
$$W'(x)=y'_1y'_2+y_1y''_2-y'_1y'_2-y''_1y_2=y_1y''_2-y''_1y_2=y_1(-p_1y'_2-p_2y_2)+y_2(p_1y'_1+p_2y_1)=-p_1W(x)$$
因此$W(x)$满足微分方程$y'=-p_1y$。解此方程得到
$$W(x)=C\text{e}^{P(x)}$$
其中$P(x)=-\int p_1(x)\text{d}x$。若有$W(x_0)=0$，则因为$C\text{e}^{P(x)}>0$，必有$C=0$，从而$W(x)\equiv 0$。因此，我们得到
$$\frac{y'_1}{y_1}=\frac{y'_2}{y_2}$$
解此方程，得到
$$\ln|y_1|=\ln|y_2|+C'$$
$$y_1=Cy_2$$
这与$y_1(x), y_2(x)$线性无关矛盾！
