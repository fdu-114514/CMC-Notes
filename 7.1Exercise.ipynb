### 用初等积分法解微分方程

> 1. 下列表达式是某个微分方程的通解，求方程的形式：
> 
> $(1)x^2+Cy^2=1$；
>
> $(2)(y-C_2)^2=4C_1x$。

**思路：** 先求出$C$的表达式，然后直接对$x$求导就行了。

**解答：** **解答：** $(1)$ 由原方程可得
$$C=\frac{1-x^2}{y^2}$$
此时两边对$x$求导并代入上式可得
$$\begin{aligned}
    2x+C\cdot 2yy'&=0 \\
    2x+\frac{2(1-x^2)}{y}y'&=0 \\
    xy+(1-x^2)y'&=0
\end{aligned}$$

$(2)$ 由原方程可得
$$C_1=\frac{(y-C_2)^2}{4x}$$
此时两边对$x$求导并代入上式可得
$$\begin{aligned}
    2y'(y-C_2)&=4C_1 \\
    2y'(y-C_2)&=4\cdot\frac{(y-C_2)^2}{4x} \\
    2xy'&=y-C_2 \\
    C_2&=y-2xy'
\end{aligned}$$
再对$x$求导并代入上式可得
$$\begin{aligned}
    yy''+(y')^2-C_2y''&=0 \\
    2xy'y''+(y')^2&=0 \\
    2xy''+y'&=0
\end{aligned}$$

> 2. 设$f(s), g(s)$可导，$w=yf(xy)\mathrm{d}x+xg(xy)\mathrm{d}y$。若存在$u$使得$\mathrm{d}u=w$，求$f(s)-g(s)$。

**思路：** 回想全微分的定义。

**解答：** 由全微分的定义可知
$$\frac{\partial u}{\partial x}=yf(xy), \frac{\partial u}{\partial y}=xg(xy)$$
于是有
$$\begin{aligned}
    \frac{\partial}{\partial y}[yf(xy)]&=\frac{\partial}{\partial x}[xg(xy)] \\
    f(xy)+xyf'(xy)&=g(xy)+xyg'(xy)
\end{aligned}$$
此时令$s=xy$就有
$$\begin{aligned}
    f(s)+sf'(s)&=g(s)+sg'(s) \\
    [f(s)-g(s)]'&=-\frac{1}{s}[f(s)-g(s)] \\
    \frac{[f(s)-g(s)]'}{f(s)-g(s)}&=-\frac{1}{s} \\
    \ln|f(s)-g(s)|&=-\ln|s|+C_1 \\
    f(s)-g(s)&=\frac{C}{s}
\end{aligned}$$
其中$C$是任意常数。

3. 设$f(x)$为可微函数，满足方程$f(x)=\mathrm{e}^x+\mathrm{e}^x\int_0^xf^2(t)\mathrm{d}t$。求$f(x)$的表达式。

**思路：** 直接开导。

**解答：** 方程两边对$x$求导可得
$$\begin{aligned}
    f'(x)&=\mathrm{e}^x+\mathrm{e}^x\left(\int_0^xf^2(t)\mathrm{d}t+f^2(x)\right) \\
    \int_0^xf^2(t)\mathrm{d}t&=\mathrm{e}^{-x}f'(x)-1-f^2(x)
\end{aligned}$$
代入原方程可得
$$\begin{aligned}
    f(x)&=\mathrm{e}^x+\mathrm{e}^x[\mathrm{e}^{-x}f'(x)-1-f^2(x)] \\
    &=\mathrm{e}^x+f'(x)-\mathrm{e}^x-\mathrm{e}^xf^2(x) \\
    &=f'(x)-\mathrm{e}^xf^2(x)
\end{aligned}$$
将$f(x)$换成$y$，移项可得
$$\begin{aligned}
    y'&=y+\mathrm{e}^xy^2 \\
    \frac{y'}{y^2}-\frac{1}{y}&=\mathrm{e}^x
\end{aligned}$$
令$u=\frac{1}{y}$，则$u'=\frac{-y'}{y^2}$，代入可得
$$\begin{aligned}
    u'+u&=-\mathrm{e}^x \\
    u&=\mathrm{e}^{-x}\left(\int-\mathrm{e}^x\cdot\mathrm{e}^x\mathrm{d}x+C\right) \\
    &=\mathrm{e}^{-x}(-\frac{1}{2}\mathrm{e}^{2x}+C) \\
    &=C\mathrm{e}^{-x}-\frac{1}{2}\mathrm{e}^x
\end{aligned}$$
令$x=0$可得条件$f(0)=1$，因此
$$\begin{aligned}
    f(x)&=\frac{1}{C\mathrm{e}^{-x}-\frac{1}{2}\mathrm{e}^x} \\
    1&=\frac{1}{C-\frac{1}{2}} \\
    C&=\frac{3}{2}
\end{aligned}$$
最后得到
$$f(x)=\frac{2}{3\mathrm{e}^{-x}-\mathrm{e}^x}$$

**注：** 此题计算中遇到了伯努利方程，需要注意。

> 4. 试确定函数$u(t)$使得$\frac{\mathrm{d}u(t)}{\mathrm{d}t}=u(t)+\int_0^1u(s)\mathrm{d}s,u(0)=1$。

**思路：** $\int_0^1u(s)\mathrm{d}s$也不过是个常数罢了，再求一次导就能消掉。

**解答：** 方程两边对$t$求导可得
$$\begin{aligned}
    u''&=u' \\
    \frac{\mathrm{d}(u')}{\mathrm{d}t}&=u' \\
    u'&=C_1\mathrm{e}^t
\end{aligned}$$
再次积分可得
$$\begin{aligned}
    \frac{\mathrm{d}u}{\mathrm{d}t}&=C_1\mathrm{e}^t \\
    u&=C_1\mathrm{e}^t+C_2
\end{aligned}$$
得到表达式后，首先先将那个讨厌的积分去掉
$$\begin{aligned}
    \int_0^1u(s)\mathrm{d}s&=\int_0^1(C_1\mathrm{e}^s+C_2)\mathrm{d}s \\
    &=C_1(\mathrm{e}-1)+C_2
\end{aligned}$$
现在确定常数$C_1, C_2$。在原方程代入$t=0$就得到
$$\begin{aligned}
    u'(0)&=C_1\cdot\mathrm{e}^0 \\
    C_1&=1+\int_0^1u(s)\mathrm{d}s \\
    C_1&=(\mathrm{e}-1)C_1+C_2+1
\end{aligned}$$
再将$t=0$代入$u(t)$的表达式可得
$$\begin{aligned}
    u(0)&=C_1+C_2 \\
    C_2&=1-C_1
\end{aligned}$$
代回上式可得
$$\begin{aligned}
    C_1&=(\mathrm{e}-1)C_1+2-C_1 \\
    C_1&=\frac{2}{3-\mathrm{e}} \\
    C_2&=1-\frac{2}{3-\mathrm{e}} \\
    &=\frac{2-\mathrm{e}}{1-\mathrm{e}}
\end{aligned}$$
最后得到
$$u(t)=\frac{2\mathrm{e}^t+1-\mathrm{e}}{3-\mathrm{e}}$$

> 5. 求方程$\frac{\mathrm{d}x}{\sqrt{xy}}+(\frac{2}{y}-\sqrt{\frac{x}{y^3}})\mathrm{d}y=0(x>0)$的通解。

**思路：** 作变形$\sqrt{\frac{x}{y^3}}=\frac{1}{y}\sqrt{\frac{x}{y}}$后发现$\frac{x}{y}$非常难处理，因此考虑换元。

**解答：** 令$\frac{x}{y}=u$，则$x=uy, \mathrm{d}x=y\mathrm{d}u+u\mathrm{d}y$，代入原式可得
$$\begin{aligned}
    \frac{1}{y\sqrt{u}}(y\mathrm{d}u+u\mathrm{d}y)+\left(\frac{2}{y}-\frac{1}{y}\sqrt{u}\right)\mathrm{d}y&=0 \\
    \frac{1}{\sqrt{u}}\mathrm{d}u&=-\frac{2}{y}\mathrm{d}y \\
\end{aligned}$$
此时两边积分可得
$$\begin{aligned}
    2\sqrt{u}+C_1&=-2\ln|y| \\
    y&=C\mathrm{e}^{-\sqrt{u}}
\end{aligned}$$
也就是书上写的
$$y\mathrm{e}^{\sqrt{\frac{x}{y}}}=C$$
其中$C$是任意常数。

> 6. 求微分方程$y'=\sin^2(x-y+1)$的通解。

**思路：** $\sin$里面的玩意太难看了吧，直接换元。

**解答：** 令$u=x-y+1$，则$y'=1-u'$，代入方程可得
$$\begin{aligned}
    1-\frac{\mathrm{d}u}{\mathrm{d}x}&=\sin^2u \\
    \frac{\mathrm{d}u}{1-\sin^2u}&=\mathrm{d}x \\
    \int\sec^2u\mathrm{d}u&=\int\mathrm{d}x \\
    \tan u&=x+C
\end{aligned}$$
即$\tan(x-y+1)=x+C$。

> 7. 设实数$a\neq 0$，求微分方程初值问题$\begin{cases}
    y''-ay'^2=0 \\
    y(0)=0, y'(0)=-1
\end{cases}$的解。

**思路：** 问题不大，先求通解再确定常数就好了。

**解答：** 首先我们容易求出
$$\begin{aligned}
    \frac{\mathrm{d}(y')}{\mathrm{d}x}&=ay'^2 \\
    \frac{\mathrm{d}(y')}{y'^2}&=a\mathrm{d}x \\
    -\frac{1}{y'}&=ax+C_1 \\
    y'&=-\frac{1}{ax+C_1} \\
    \mathrm{d}y&=-\frac{1}{a}\cdot\frac{\mathrm{d}(ax+C_1)}{ax+C_1} \\
    y&=-\frac{1}{a}\ln|ax+C_1|+C_2
\end{aligned}$$
然后我们先代入$y'(0)=-1$的条件，就得到
$$\begin{aligned}
    -\frac{1}{C_1}&=-1 \\
    C_1&=1
\end{aligned}$$
然后代入$y(0)=0$的条件得到
$$\begin{aligned}
    -\frac{1}{a}\ln|1|+C_2&=0 \\
    C_2&=0
\end{aligned}$$
因此问题的解为$y=-\frac{1}{a}\ln|ax+1|$。

> 8. 设$u=u(x, y)$是方程$P\mathrm{d}x+Q\mathrm{d}y=0$的积分因子且$uP\mathrm{d}x+uQ\mathrm{d}y=\mathrm{d}V$，证明：$u\phi(V)$也是该方程的积分因子，其中$\phi(V)$连续。

> 9. 设$\mu_1(x, y)\neq 0, \mu_2(x, y)\neq 0$，且$\mu_1/\mu_2\neq k$(常数)。它们都是方程$M(x, y)\mathrm{d}x+N(x, y)\mathrm{d}y=0$的积分因子。求证：$\frac{\mu_1}{\mu_2}=C$是该方程的通解，其中$C$为任意非零常数。

**思路：** 看见全微分的形式，首先要想到它与偏微分的联系。

**解答：** 由全微分的定义可知
$$\begin{aligned}
    \frac{\partial}{\partial y}(\mu_1M)&=\frac{\partial}{\partial x}(\mu_1N) \\
    M\frac{\partial\mu_1}{\partial y}+\mu_1\frac{\partial M}{\partial y}&=N\frac{\partial\mu_1}{\partial x}+\mu_1\frac{\partial N}{\partial x} \\
    \frac{\partial N}{\partial x}-\frac{\partial M}{\partial y}&=\frac{1}{\mu_1}\left[M\frac{\partial\mu_1}{\partial y}-N\frac{\partial\mu_1}{\partial x}\right] \\
    \frac{\partial N}{\partial x}-\frac{\partial M}{\partial y}&=M\frac{\partial\ln|\mu_1|}{\partial y}-N\frac{\partial\ln|\mu_1|}{\partial x}
\end{aligned}$$
同理得到
$$\frac{\partial N}{\partial x}-\frac{\partial M}{\partial y}=M\frac{\partial\ln|\mu_2|}{\partial y}-N\frac{\partial\ln|\mu_2|}{\partial x}$$
两式相减就可以推出
$$\begin{aligned}
    M\left(\frac{\partial\ln|\mu_1|}{\partial y}-\frac{\partial\ln|\mu_2|}{\partial y}\right)-N\left(\frac{\partial\ln|\mu_1|}{\partial x}-\frac{\partial\ln|\mu_2|}{\partial x}\right)&=0 \\
    M\frac{\partial\ln|\frac{\mu_1}{\mu_2}|}{\partial y}-N\frac{\partial\ln|\frac{\mu_1}{\mu_2}|}{\partial x}&=0 \\
    \frac{N}{M}&=\frac{\frac{\partial\ln|\frac{\mu_1}{\mu_2}|}{\partial y}}{\frac{\partial\ln|\frac{\mu_1}{\mu_2}|}{\partial x}}
\end{aligned}$$
而对于方程$\frac{\mu_1}{\mu_2}=C$，有$\mathrm{d}(\ln|\frac{\mu_1}{\mu_2}|)=0$，也就是
$$\begin{aligned}
    \frac{\partial\ln|\frac{\mu_1}{\mu_2}|}{\partial x}\mathrm{d}x+\frac{\partial\ln|\frac{\mu_1}{\mu_2}|}{\partial y}\mathrm{d}y&=0 \\
    \frac{\frac{\partial\ln|\frac{\mu_1}{\mu_2}|}{\partial y}}{\frac{\partial\ln|\frac{\mu_1}{\mu_2}|}{\partial x}}&=-\frac{\mathrm{d}x}{\mathrm{d}y}
\end{aligned}$$
与上式比较就可得
$$\begin{aligned}
    \frac{N}{M}&=-\frac{\mathrm{d}x}{\mathrm{d}y} \\
    M\mathrm{d}x+N\mathrm{d}y&=0
\end{aligned}$$
这说明由方程$\frac{\mu_1}{\mu_2}=C$所确定的隐函数是微分方程$M\mathrm{d}x+N\mathrm{d}y=0$的通解。

> 10. 求解方程组
$$\begin{cases}
    \frac{\mathrm{d}y}{\mathrm{d}x}=z(y+z)^n \\
    \frac{\mathrm{d}z}{\mathrm{d}x}=y(y+z)^n
\end{cases}$$
初始条件为：当$x=0$时，$y=1, z=0$。

**思路：** $y+z$是一个很特殊的结构，可以换元，并将两方程相加或相减。

**解答：** 先设$n\neq 0$，则令$u=y+z$，将两方程相加可得
$$\begin{aligned}
    \frac{\mathrm{d}y}{\mathrm{d}x}+\frac{\mathrm{d}z}{\mathrm{d}x}&=(y+z)^{n+1} \\
    \frac{\mathrm{d}u}{\mathrm{d}x}&=u^{n+1} \\
    -\frac{1}{n}u^{-n}&=x+C \\
    u^n&=-\frac{1}{n(x+C)}
\end{aligned}$$
由条件可得
$$\begin{aligned}
    1&=-\frac{1}{nC} \\
    C&=-\frac{1}{n}
\end{aligned}$$
因此
$$\begin{aligned}
    u^n&=\frac{1}{1-nx} \\
    u&=(1-nx)^{-\frac{1}{n}}
\end{aligned}$$
令$v=y-z$并将两方程相减可得
$$\begin{aligned}
    \frac{\mathrm{d}v}{\mathrm{d}x}&=-vu^n \\
    \frac{\mathrm{d}v}{v}&=-\frac{\mathrm{d}x}{1-nx} \\
    \ln|v|&=\frac{1}{n}\ln|1-nx|+C \\
    v&=C(1-nx)^{\frac{1}{n}}
\end{aligned}$$
由条件可得$C=1$，因此
$$v=(1-nx)^{\frac{1}{n}}$$
最后就可得到
$$\begin{aligned}
    y&=\frac{u+v}{2}=\frac{1}{2}[(1-nx)^{-\frac{1}{n}}+(1-nx)^{\frac{1}{n}}] \\
    z&=\frac{u-v}{2}=\frac{1}{2}[(1-nx)^{-\frac{1}{n}}-(1-nx)^{\frac{1}{n}}]
\end{aligned}$$
若$n=0$，则用同样的方法可得
$$\begin{aligned}
    \frac{\mathrm{d}u}{\mathrm{d}x}&=u \\
    \frac{\mathrm{d}v}{\mathrm{d}x}&=-v
\end{aligned}$$
容易解得$u=\mathrm{e}^x, v=\mathrm{e}^{-x}$，从而
$$\begin{aligned}
    y&=\frac{1}{2}(\mathrm{e}^x+\mathrm{e}^{-x})=\cosh x \\
    z&=\frac{1}{2}(\mathrm{e}^x-\mathrm{e}^{-x})=\sinh x
\end{aligned}$$

> 11. 求下列常系数非齐次线性方程的通解(建议用算子解法求特解)：
>
> $(1)y'-y=x^3+1$；
>
> $(2)y''+y=x^2-x+2$；
>
> $(3)y''-4y'+4y=\mathrm{e}^{2x}$；
>
> $(4)y''+y=x^2-x+2+9\sin x-3\sin 2x$

**思路：** 就是硬算，只有计算量大小的区别了。

**解答：** $(1)$ 解对应的齐次线性方程
$$y'-y=0$$
可得其通解$\tilde{y}=C\mathrm{e}^x$。现在求其特解，将方程写成算子形式可得
$$(D-1)y=x^3+1$$
因此其特解是
$$\begin{aligned}
    y^*&=\frac{1}{D-1}(x^3+1) \\
    &=(-1-D-D^2-D^3)(x^3+1) \\
    &=-(x^3+1)-3x^2-6x-6 \\
    &=-(x^3+3x^2+6x+7)
\end{aligned}$$
最后得到原方程的通解为
$$y=C\mathrm{e}^x-(x^3+3x^2+6x+7)$$

$(2)$ 解对应的齐次线性方程
$$y''+y=0$$
方程对应的特征方程为
$$r^2+1=0$$
其特征根为$r_{1, 2}=\pm\mathrm{i}$，因此通解$\tilde{y}=C_1\cos x+C_2\sin x$。现在求其特解，将方程写成算子形式可得
$$(D^2+1)y=x^2-x+2$$
因此其特解是
$$\begin{aligned}
    y^*&=\frac{1}{1+D^2}(x^2-x+2) \\
    &=(1-D^2)(x^2-x+2) \\
    &=(x^2-x+2)-2 \\
    &=x^2-x
\end{aligned}$$
最后得到原方程的通解为
$$y=C_1\cos x+C_2\sin x+x^2-x$$

$(3)$ 解对应的齐次线性方程
$$y''-4y'+4y=0$$
方程对应的特征方程为
$$r^2-4r+4=0$$
其特征根为$r=2$，这是一个二重根，因此通解$\tilde{y}=(C_1+C_2x)\mathrm{e}^{2x}$。现在求其特解，将方程写成算子形式可得
$$(D^2-4D+4)y=\mathrm{e}^{2x}$$
因此其特解是
$$\begin{aligned}
    y^*&=\frac{1}{(D-2)^2}(\mathrm{e}^{2x}) \\
    &=\mathrm{e}^{2x}\frac{1}{(D-2+2)^2}(1) \\
    &=\mathrm{e}^{2x}\frac{1}{D^2}(1) \\
    &=\frac{x^2}{2}\mathrm{e}^{2x}
\end{aligned}$$
最后得到原方程的通解为
$$y=(C_1+C_2x)\mathrm{e}^{2x}+\frac{x^2}{2}\mathrm{e}^{2x}$$

$(4)$ 解对应的齐次线性方程
$$y''+y=0$$
方程对应的特征方程为
$$r^2-4r+4=0$$
与$(2)$同理得到通解$\tilde{y}=C_1\cos x+C_2\sin x$。现在求其特解，将方程写成算子形式可得
$$(D^2+1)y=x^2-x+2+9\sin x-3\sin 2x$$
因此其特解是
$$\begin{aligned}
    y^*&=\frac{1}{D^2+1}(x^2-x+2+9\sin x-3\sin 2x) \\
    &=\frac{1}{1+D^2}(x^2-x+2)+\frac{1}{D^2+1}(9\sin x-3\sin 2x) \\
    &=x^2-x+\frac{1}{D^2+1}\left[\text{Im}(9\mathrm{e}^{\mathrm{i}x}-3\mathrm{e}^{2\mathrm{i}x})\right] \\
    &=x^2-x+\mathrm{Im}\left[\frac{1}{D^2+1}(9\mathrm{e}^{\mathrm{i}x})\right]-\text{Im}\left[\frac{1}{D^2+1}(3\mathrm{e}^{2\mathrm{i}x})\right] \\
    &=x^2-x+\text{Im}\left[\mathrm{e}^{\mathrm{i}x}\frac{1}{(D+\mathrm{i})^2+1}(9)\right]-\text{Im}\left[\mathrm{e}^{2\mathrm{i}x}\frac{1}{(D+2\mathrm{i})+1}(3)\right] \\
    &=x^2-x+\text{Im}\left[\mathrm{e}^{\mathrm{i}x}\frac{1}{D(D+2\mathrm{i})}(9)\right]-\text{Im}\left[\mathrm{e}^{2\mathrm{i}x}\frac{1}{D^2+4\mathrm{i}D-3}(3)\right] \\
    &=x^2-x+\text{Im}\left[\mathrm{e}^{\mathrm{i}x}\frac{1}{D}(\frac{9}{2}\mathrm{i})\right]-\text{Im}\left[\mathrm{e}^{2\mathrm{i}x}(-1)\right] \\
    &=x^2-x+\text{Im}(\frac{9}{2}\mathrm{i}x\cdot\mathrm{e}^{\mathrm{i}x})-\text{Im}(-\mathrm{e}^{2\mathrm{i}x}) \\
    &=x^2-x+\frac{9}{2}x\cos x+\sin 2x
\end{aligned}$$
最后得到原方程的通解为
$$y=C_1\cos x+C_2\sin x+x^2-x+\frac{9}{2}x\cos x+\sin 2x$$

> 12. 求常系数线性微分方程组$\begin{cases}
    x'+y'+x+y=2t \\
    x'+2y'-y=3t
\end{cases}$的特解，其中$x, y$都是$t$的函数。

**思路：** 先将方程组写成算子形式，然后进行消元求解。

**解答：** 将方程组写成算子形式
$$\begin{cases}
    Dx+Dy+x+y=2t \\
    Dx+2Dy-y=3t
\end{cases}$$
解得$Dy=t+x+2y, Dx=t-2x-3y$。从前一式中可知
$$x=Dy-2y-t$$
与上式比较可得
$$\begin{aligned}
    t-2(Dy-2y-t)-3y&=D^2y-2Dy-1 \\
    t-2Dy+4y+2t-3y&=D^2y-2Dy-1 \\
    D^2y-y&=-3t-1 \\
    y&=\frac{1}{D^2-1}(-3t-1) \\
    &=-3t-1
\end{aligned}$$
代入上式可得
$$\begin{aligned}
    x&=-3-2(-3t-1)-t \\
    &=5t-1
\end{aligned}$$

> 13. 求欧拉方程$x^3y'''+x^2y''-4xy'=3x^2$的通解。

**思路：** 欧拉方程的形式是$x^ny^{(n)}+p_1x^{n-1}y^{(n-1)}+\cdots+p_{n-1}xy'+p_ny=q(x)$，解法是作换元$x=\mathrm{e}^t$。

**解答：** 令$x=\mathrm{e}^t$，记$D=\frac{\mathrm{d}}{\mathrm{d}t}$，则原方程变为
$$\begin{aligned}
    D(D-1)(D-2)y+D(D-1)y-4Dy&=3\mathrm{e}^{2t} \\
    (D^3-2D^2-3D)y&=3\mathrm{e}^{2t}
\end{aligned}$$
先求对应齐次线性方程的通解，对应的特征方程为
$$\begin{aligned}
    r^3-2r^2-3r&=0 \\
    r(r+1)(r-3)&=0
\end{aligned}$$
因此特征根为$r_1=0, r_2=-1, r_3=3$，通解$\tilde{y}=C_1+C_2\mathrm{e}^{-t}+C_3\mathrm{e}^{3t}$。现在求其特解，容易算得
$$\begin{aligned}
    y^*&=\frac{1}{D^3-2D^2-3D}(3\mathrm{e}^{2t}) \\
    &=\mathrm{e}^{2t}\frac{1}{(D+2)^3-2(D+2)^2-3(D+2)}(3) \\
    &=\mathrm{e}^{2t}\frac{1}{D^3+4D^2+D-6}(3) \\
    &=-\frac{1}{2}\mathrm{e}^{2t}
\end{aligned}$$
因此通解为
$$\begin{aligned}
    y&=C_1+C_2\mathrm{e}^{-t}+C_3\mathrm{e}^{3t}-\frac{1}{2}\mathrm{e}^{2t} \\
    &=C_1+C_2x^{-1}+C_3x^3-\frac{1}{2}x^2
\end{aligned}$$

> 14. 设二阶非齐次线性方程$y''+p(x)y'+q(x)y=f(x)$有三个特解$\mathrm{e}^x, \mathrm{e}^x+\mathrm{e}^{\frac{x}{2}}, \mathrm{e}^x+\mathrm{e}^{-x}$。求该方程的表达式和通解。

**思路：** 三个特解配三个未知函数吗？有点意思。直接全部代进去求解。

**解答：** 将三个特解都代入方程得到
$$\begin{aligned}
    \mathrm{e}^x+\mathrm{e}^xp(x)+\mathrm{e}^xq(x)&=f(x) \\
    \mathrm{e}^x+\frac{1}{4}\mathrm{e}^{\frac{x}{2}}+(\mathrm{e}^x+\frac{1}{2}\mathrm{e}^{\frac{x}{2}})p(x)+(\mathrm{e}^x+\mathrm{e}^{\frac{x}{2}})q(x)&=f(x) \\
    \mathrm{e}^x+\mathrm{e}^{-x}+(\mathrm{e}^x-\mathrm{e}^{-x})p(x)+(\mathrm{e}^x+\mathrm{e}^{-x})q(x)&=f(x)
\end{aligned}$$
将第一式代入第二、三式可知
$$\begin{aligned}
    \frac{1}{4}\mathrm{e}^{\frac{x}{2}}+\frac{1}{2}\mathrm{e}^{\frac{x}{2}}p(x)+\mathrm{e}^{\frac{x}{2}}q(x)&=0 \\
    \mathrm{e}^{-x}-\mathrm{e}^{-x}p(x)+\mathrm{e}^{-x}q(x)&=0 \\
    p(x)&=q(x)+1
\end{aligned}$$
然后可以得到
$$\begin{aligned}
    \frac{1}{4}+\frac{1}{2}[q(x)+1]+q(x)&=0 \\
    \frac{3}{2}q(x)&=-\frac{3}{4} \\
    q(x)&=-\frac{1}{2}
\end{aligned}$$
从而$p(x)=\frac{1}{2}$。再代入第一式就得到
$$\begin{aligned}
    \mathrm{e}^x+\frac{1}{2}\mathrm{e}^x-\frac{1}{2}\mathrm{e}^x&=f(x) \\
    f(x)&=\mathrm{e}^x
\end{aligned}$$
因此方程的表达式为
$$y''+\frac{1}{2}y'-\frac{1}{2}y=\mathrm{e}^x$$
其对应的齐次线性方程的特征根为$r_1=-1, r_2=\frac{1}{2}$，因此通解$\tilde{y}=C_1\mathrm{e}^{-x}+C_2\mathrm{e}^{\frac{x}{2}}$。对比题设条件可知原方程的通解$y=C_1\mathrm{e}^{-x}+C_2\mathrm{e}^{\frac{x}{2}}+\mathrm{e}^x$。

> 15. 设二阶线性微分方程为$y''+p(x)y'+q(x)y=f(x)$，它有特解$y_1=x^4+\frac{x^4}{5}\ln x, y_2=\frac{1}{x}+\frac{x^4}{5}\ln x, y_3=\frac{x^4}{5}\ln x$。求该方程的表达式和通解。

**思路：** 要是直接代入计算，计算量会很大。但可以投机取巧。

**解答：** 直接猜$y_3$是特解，将$y_1-y_3$和$y_2-y_3$代入对应的齐次线性方程
$$\begin{aligned}
    12x^2+4x^3p(x)+x^4q(x)&=0 \\
    \frac{2}{x^3}-\frac{1}{x^2}p(x)+\frac{1}{x}q(x)&=0
\end{aligned}$$
也就是
$$\begin{aligned}
    x^2q(x)+4xp(x)+12&=0 \\
    x^2q(x)-xp(x)+2&=0
\end{aligned}$$
解之，可得
$$\begin{aligned}
    p(x)&=-\frac{2}{x} \\
    q(x)&=-\frac{4}{x^2}
\end{aligned}$$
再将$p(x), q(x)$和$y_3$代入原方程可得
$$\begin{aligned}
    \frac{12}{5}x^2\ln x+\frac{7}{5}x^2-\frac{2}{x}(\frac{4}{5}x^3\ln x+\frac{1}{5}x^3)-\frac{4}{x^2}\cdot\frac{x^4}{5}\ln x&=f(x) \\
    \frac{12}{5}x^2\ln x+\frac{7}{5}x^2-\frac{8}{5}x^2\ln x-\frac{2}{5}x^2-\frac{4}{5}x^2\ln x&=f(x) \\

\end{aligned}$$
因此方程的表达式是

> 16. 已知二阶常系数线性方程$y''+ay'+by=c\mathrm{e}^x$有特解$y=\mathrm{e}^{-x}(1+x\mathrm{e}^{2x})$，求该方程的通解。

**思路：** 只给了一个特解，没办法，只好代进去算了。

**解答：** 将$y=\mathrm{e}^{-x}+x\mathrm{e}^x$代入方程可得
$$\begin{aligned}
    \mathrm{e}^{-x}+(x+2)\mathrm{e}^x+a[-\mathrm{e}^{-x}+(x+1)\mathrm{e}^x]+b(\mathrm{e}^{-x}+x\mathrm{e}^x)&=c\mathrm{e}^x \\
    (1-a+b)\mathrm{e}^{-x}+[(1+a+b)x+(2+a)]\mathrm{e}^x&=c\mathrm{e}^x
\end{aligned}$$
容易知道
$$\begin{cases}
    1-a+b=0 \\
    1+a+b=0 \\
    2+a=c
\end{cases}$$
求得$a=0, b=-1, c=2$，因此原方程为
$$y''-y=2\mathrm{e}^x$$
对应的齐次线性方程的通解$\tilde{y}=C_1\mathrm{e}^x+C_2\mathrm{e}^{-x}$。现在求其特解，将方程写成算子形式并求解可得
$$\begin{aligned}
    (D^2-1)y&=2\mathrm{e}^x \\
    y&=2\cdot\frac{1}{D^2-1}(\mathrm{e}^x) \\
    &=2\mathrm{e}^x\cdot\frac{1}{D(D+2)}(1) \\
    &=2\mathrm{e}^x\cdot\frac{1}{D}(\frac{1}{2}) \\
    &=x\mathrm{e}^x
\end{aligned}$$
因此原方程的通解为
$$y=C_1\mathrm{e}^x+C_2\mathrm{e}^{-x}+x\mathrm{e}^x$$

> 17. 设$y=y_1(x), y=y_2(x)$为方程$y'=P(x)y+Q(x)$的两个互异解。证明：对该方程的任一解$y(x)$，都有恒等式$\frac{y(x)-y_1(x)}{y_2(x)-y_1(x)}=C$成立，其中$C$为某一常数。

**思路：** 直接把$y_1, y_2$先代进去。

**解答：** 将$y_1(x), y_2(x)$代入方程可得
$$\begin{aligned}
    y'_1&=P(x)y_1+Q(x) \\
    y'_2&=P(x)y_2+Q(x)
\end{aligned}$$
因此有
$$\begin{aligned}
    (y_2-y_1)'&=P(x)(y_2-y_1) \\
    y_2-y_1&=C_1\mathrm{e}^{\int P(x)\mathrm{d}x}
\end{aligned}$$
因为$y_1, y_2$互异，故$C_1\neq 0$。同理可得
$$y-y_1=C_2\mathrm{e}^{\int P(x)\mathrm{d}x}$$
最后可得
$$\frac{y(x)-y_1(x)}{y_2(x)-y_1(x)}=\frac{C_2}{C_1}=C$$

> 18. 设$y_1(x), y_2(x), y_3(x)$均为非齐次线性方程$y''+P_1(x)y'+P_2(x)y=Q(x)$的特解，其中$P_1(x), P_2(x), Q(x)$连续且$\frac{y_2(x)-y_1(x)}{y_3(x)-y_1(x)}$不是常数。证明：$y(x)=(1-C_1-C_2)y_1(x)+C_1y_2(x)+C_2y_3(x)$为给定方程的通解($C_1, C_2$为任意常数)。

**思路：** 与上题类似。

**解答：** 计算可知$y_2-y_1, y_3-y_1$均为对应齐次线性方程的通解。由$\frac{y_2(x)-y_1(x)}{y_3(x)-y_1(x)}$不是常数可知它们线性无关。又由$y_1$为原方程的特解可知，原方程的通解可写为

    &=(1-C_1-C_2)y_1(x)+C_1y_2(x)+C_2y_3(x)

其中$C_1, C_2$为任意常数。

> 19. $(1)$ 设二阶常系数非齐次线性方程为$y''+py'+qy=f(x)$，其中$p, q$为常数。其对应的齐次方程的特征根分别为$r_1, r_2$。证明：该方程的通解为
> $$y=\mathrm{e}^{r_1x}\left[\int \mathrm{e}^{-r_1x}\mathrm{e}^{r_2x}\left(\int \mathrm{e}^{-r_2x}f(x)\mathrm{d}x+C_1\right)\mathrm{d}x+C_2\right]$$
> $(2)$求微分方程$y''+y'-2y=\frac{1}{\mathrm{e}^x+\mathrm{e}^{2x}}$的通解。

**提示：** 特征根满足
$$\begin{cases}
    r_1+r_2=-p \\
    r_1r_2=q
\end{cases}$$

**解答：** $(1)$ 令$u=y'-r_1y$，则易知$u'=y''-r_1y'$，因此原方程化为
$$u'-r_2u=f(x)$$
乘以积分因子$\mathrm{e}^{-r_2x}$可得
$$\begin{aligned}
    \mathrm{e}^{-r_2x}u'-r_2\mathrm{e}^{-r_2x}u&=\mathrm{e}^{-r_2x}f(x) \\
    (\mathrm{e}^{-r_2x}u)'&=\mathrm{e}^{-r_2x}f(x) \\
    u&=\mathrm{e}^{r_2x}\left(\int \mathrm{e}^{-r_2x}f(x)\mathrm{d}x+C_1\right)
\end{aligned}$$
也就是
$$\begin{aligned}
    y'-r_1y&=\mathrm{e}^{r_2x}\left(\int \mathrm{e}^{-r_2x}f(x)\mathrm{d}x+C_1\right)
\end{aligned}$$
乘以积分因子$\mathrm{e}^{-r_1x}$可得
$$\begin{aligned}
    (\mathrm{e}^{-r_1x}y)'&=\mathrm{e}^{-r_1x}\mathrm{e}^{r_2x}\left(\int \mathrm{e}^{-r_2x}f(x)\mathrm{d}x+C_1\right) \\
    y&=\mathrm{e}^{r_1x}\left[\int \mathrm{e}^{-r_1x}\mathrm{e}^{r_2x}\left(\int \mathrm{e}^{-r_2x}f(x)\mathrm{d}x+C_1\right)\mathrm{d}x+C_2\right]
\end{aligned}$$

$(2)$ 首先算出对应的齐次方程的特征根为$r_1=1, r_2=-2$。直接套用上面的公式可得
$$\begin{aligned}
    y&=\mathrm{e}^x\left[\int\mathrm{e}^{-x}\mathrm{e}^{-2x}\left(\int\frac{\mathrm{e}^{2x}}{\mathrm{e}^x+\mathrm{e}^{2x}}\mathrm{d}x+C_1\right)\mathrm{d}x+C_2\right] \\
    &=\mathrm{e}^x\left[\int\mathrm{e}^{-3x}\left(\int\frac{1}{1+t}\mathrm{d}t+C_1\right)\mathrm{d}x+C_2\right] \\
    &=\mathrm{e}^x\left[\int\mathrm{e}^{-3x}\left(\ln(1+\mathrm{e}^x)+C_1\right)\mathrm{d}x+C_2\right] \\
    &=\mathrm{e}^x\left[\int C_1\mathrm{e}^{-3x}\mathrm{d}x+\int \mathrm{e}^{-3x}\ln(1+\mathrm{e}^x)\mathrm{d}x+C_2\right] \\
    &=\mathrm{e}^x\left[C_1\mathrm{e}^{-3x}+C_2+\int \frac{\ln(1+\mathrm{e}^x)}{\mathrm{e}^{4x}}\mathrm{d}(\mathrm{e}^x)\right] \\
    &=\mathrm{e}^x\left(-\frac{\ln(1+\mathrm{e}^x)}{3\mathrm{e}^{3x}}+\frac{x}{3}-\frac{1}{3}\ln(1+\mathrm{e}^x)+\frac{1}{3\mathrm{e}^x}-\frac{1}{6\mathrm{e}^{2x}}+\frac{C_1}{\mathrm{e}^{3x}}+C_2\right) \\
    &=C_1\mathrm{e}^{-2x}+C_2\mathrm{e}^x+\frac{1}{3}\left(1+x\mathrm{e}^x-\frac{1}{2\mathrm{e}^x}\right)+\frac{1}{3}(\mathrm{e}^{-2x}+\mathrm{e}^x)\ln(1+\mathrm{e}^x)
\end{aligned}$$

> 20. 求不定积分$\int x^3\mathrm{e}^{2x}\mathrm{d}x$。

**思路：** 硬积当然是可以的，但还有另一种做法。

**解答1:** 直接积分可得
$$\begin{aligned}
    \int x^3\mathrm{e}^{2x}\mathrm{d}x&=\frac{1}{2}x^3\mathrm{e}^{2x}-\frac{3}{2}\int x^2\mathrm{e}^{2x}\mathrm{d}x \\
    &=\frac{1}{2}x^3\mathrm{e}^{2x}-\frac{3}{4}x^2\mathrm{e}^{2x}+\frac{3}{2}\int x\mathrm{e}^{2x}\mathrm{d}x \\
    &=\frac{1}{2}x^3\mathrm{e}^{2x}-\frac{3}{4}x^2\mathrm{e}^{2x}+\frac{3}{4}x\mathrm{e}^{2x}-\frac{3}{4}\int\mathrm{e}^{2x}\mathrm{d}x \\
    &=\left(\frac{1}{2}x^3-\frac{3}{4}x^2+\frac{3}{4}x-\frac{3}{8}\right)\mathrm{e}^{2x}+C
\end{aligned}$$

**解答2:** 用逆算子的形式可知，要求的是其中一个原函数是
$$\frac{1}{D}(x^3\mathrm{e}^{2x})$$
计算可得
$$\begin{aligned}
    \frac{1}{D}(x^3\mathrm{e}^{2x})&=\mathrm{e}^{2x}\cdot\frac{1}{D+2}(x^3) \\
    &=\mathrm{e}^{2x}\cdot\left(\frac{1}{2}-\frac{1}{4}D+\frac{1}{8}D^2-\frac{1}{16}D^3\right)(x^3) \\
    &=\left(\frac{1}{2}x^3-\frac{3}{4}x^2+\frac{3}{4}x-\frac{3}{8}\right)\mathrm{e}^{2x}
\end{aligned}$$
因此不定积分的结果就是
$$\left(\frac{1}{2}x^3-\frac{3}{4}x^2+\frac{3}{4}x-\frac{3}{8}\right)\mathrm{e}^{2x}+C$$

> 21. 解二阶偏微分方程$\frac{\partial^2 z}{\partial x\partial y}=\frac{1}{2x}\cdot\frac{\partial z}{\partial y}$，其中$z=z(x, y)$有连续的二阶偏导数。

> 22. 设$f$是二次可微函数，对于任何实数$x, y$都满足函数方程$f^2(x)-f^2(y)=f(x+y)f(x-y)$，且存在$v_0$，使得$f(v_0)\neq 0$。试求$f$的表达式。

**思路：** 可微就大概意味着要求偏导了。

**解答：** 令$x=y=0$可知$f^2(0)=0$，从而$f(0)=0$。方程两边对$x$求偏导得到
$$2f(x)f'(x)=f'(x+y)f(x-y)+f(x+y)f'(x-y)$$
再对$y$求偏导可得
$$0=f''(x+y)f(x-y)-f(x+y)f''(x-y)$$
令$u=x-y, v=x+y$，就得到
$$f(u)f''(v)=f(v)f''(u)$$
这对任意$u, v$都成立。令$k=\frac{f''(v_0)}{f(v_0)}$，则上式变为
$$f''(u)=kf(u)$$
现就$k$的不同取值求解方程。

1. 若$k>0$，容易知道
$$f(u)=C_1\mathrm{e}^{\sqrt{k}u}+C_2\mathrm{e}^{-\sqrt{k}u}$$
代入$u=0$可知$C_2=-C_1$，于是
$$\begin{aligned}
    f(u)&=2C_1\cdot\frac{\mathrm{e}^{\sqrt{k}u}-\mathrm{e}^{-\sqrt{k}u}}{2} \\
    &=A\sinh(\sqrt{k}u)
\end{aligned}$$

2. 若$k<0$，则可求得
$$f(u)=C_1\cos(\sqrt{-k}u)+C_2\sin(\sqrt{-k}u)$$
代入$u=0$可知$C_1=0$，于是
$$f(u)=A\sin(\sqrt{-k}u)$$

3. 若$k=0$，则显然
$$f(u)=Au+B$$
代入$u=0$可知$B=0$，于是
$$f(u)=Au$$

> 23. 设$F(x)$是$f(x)$的一个原函数，且$F(1)=\frac{\sqrt{2}}{4}\pi$。当$x>0$时，有$f(x)F(x)=\frac{\arctan\sqrt{x}}{\sqrt{x}(1+x)}$，求$f(x)$。

**思路：** 注意到$F'(x)=f(x)$...

**解答：** 令$y=F(x)$，则$y'=f(x)$，原方程可写为
$$y'y=\frac{\arctan\sqrt{x}}{\sqrt{x}(1+x)}$$
计算可知
$$\begin{aligned}
    y\frac{\mathrm{d}y}{\mathrm{d}x}&=\frac{\arctan\sqrt{x}}{\sqrt{x}(1+x)} \\
    \int y\mathrm{d}y&=\int\frac{\arctan\sqrt{x}}{\sqrt{x}(1+x)}\mathrm{d}x \\
    \frac{1}{2}y^2&=(\arctan\sqrt{x})^2+C
\end{aligned}$$
代入$x=1$可得

    C&=0
\end{aligned}$$
因此可得
$$\begin{aligned}
    f(x)&=\frac{\arctan\sqrt{x}}{\sqrt{x}(1+x)}\cdot\frac{1}{\sqrt{2}\arctan\sqrt{x}} \\
    &=\frac{1}{\sqrt{2x}(1+x)}

> 25. 设$p_1(x), p_2(x)$是连续函数，$y_1(x), y_2(x)$是方程$y''+p_1(x)y'+p_2(x)y=0$的两个线性无关的解。证明：如果$\alpha, \beta$是$y_1(x)$的两个零点，则在$\alpha, \beta$之间必存在$y_2(x)$的一个零点。

> 26. 试证：如果微分方程$\frac{\mathrm{d}y}{\mathrm{d}x}+p(x)y=q(x)$的积分曲线族被直线$x=k$所截，则在各交点处的所有切线交于同一点，其中$p(x)q(x)\neq 0$。

**思路：** 先写出切线的方程，然后随便拉两条积分曲线出来代两个点进去算一下，如果与交点无关就行了。

**解答：** 积分曲线族在各交点处的切线的形式为
$$y=[q(k)-f(k)p(k)](x-k)+f(k)$$
设$y_1(x), y_2(x)$是微分方程的两条积分曲线，与$x=k$的交点分别为$(k, a), (k, b)$，其中$a\neq b$，那么它们的切线为
$$\begin{aligned}
    y&=[q(k)-ap(k)](x-k)+a \\
    y&=[q(k)-bp(k)](x-k)+b
\end{aligned}$$
联立计算可得交点为$(k+\frac{1}{p(k)}, \frac{q(k)}{p(k)})$，这与$a, b$无关。因此，任何此类切线都经过该点。
