### **一般数列的极限**

> 1. 设$x_n=\frac{2^3-1}{2^3+1}\cdot\frac{3^3-1}{3^3+1}\cdots\frac{n^3-1}{n^3+1}$，求$\lim_{n\to\infty}x_n$。

**思路：** 注意到分子/分母分别是立方差/立方和的形式，于是试着写几项看看能不能抵消。
$$k^3-1=(k-1)(k^2+k+1), k^3+1=(k+1)(k^2-k+1)$$
其中
$$(k+2)-1=k+1, (k+1)^2-(k+1)+1=(k^2+2k+1)-(k+1)+1=k^2+k+1$$
所以大部分项都是可以抵消的，可以先算出结果再写一句“注意到”，反正别人也不想看你的计算过程。

**解答：** 注意到
$$x_n=\frac{2(n^2+n+1)}{3n(n+1)}$$
因此
$$\lim_{n\to\infty}x_n=\frac{2}{3}$$

> 2. 设$x_n=(1+\frac{1}{2})(1+\frac{1}{4})(1+\frac{1}{16})\cdots(1+\frac{1}{2^{2^n}})$，求$\lim_{n\to\infty}x_n$。

**思路：** 看到题目右边每个括号里的第二项以$(\frac{1}{2})^n$的速度递减，立马应该要想到平方差公式。

**解答：** 因为
$$(1-\frac{1}{2})x_n=(1-\frac{1}{2})(1+\frac{1}{2})(1+\frac{1}{4})(1+\frac{1}{16})\cdots(1+\frac{1}{2^{2^n}})=1-\frac{1}{2^{2^{n+1}}}$$
所以
$$\lim_{n\to\infty}x_n=\lim_{n\to\infty}\frac{1-\frac{1}{2^{2^{n+1}}}}{1-\frac{1}{2}}=2$$

> 3. 设$\alpha>0$，求$\lim_{n\to\infty}(\cos\frac{1}{n^{\alpha}})^n$。

**思路：** 看到这种奇怪的指数就应该要取对数。

**解答：**
$$\lim_{n\to\infty}(\cos\frac{1}{n^{\alpha}})^n=\exp(\lim_{n\to\infty}n\ln\cos\frac{1}{n^{\alpha}})$$
把极限单独抽出来，并作倒代换有
$$\lim_{n\to\infty}n\ln\cos\frac{1}{n^{\alpha}}=\lim_{t\to 0^+}\frac{\ln\cos t^{\alpha}}{t}=\lim_{t\to 0^+}\frac{-\sin t^{\alpha}\alpha t^{\alpha-1}}{\cos t^{\alpha}}=-\alpha\lim_{t\to 0^+}\frac{\sin t^{\alpha}}{t^{\alpha}}\cdot t^{2\alpha-1}=\begin{cases}
-\infty, &\alpha<\frac{1}{2} \\
-\alpha, &\alpha=\frac{1}{2} \\
0, &\alpha>\frac{1}{2}
\end{cases}$$
所以
$$\lim_{n\to\infty}(\cos\frac{1}{n^{\alpha}})^n=\begin{cases}
0, &\alpha<\frac{1}{2} \\
e^{-\frac{1}{2}}, &\alpha=\frac{1}{2} \\
1, &\alpha>\frac{1}{2}
\end{cases}$$

> 4. 设$x_n=\frac{1}{2}\cdot\frac{3}{4}\cdots\frac{2n-1}{2n}$，求$\lim_{n\to\infty}x_n$

**思路：** 可以看出极限是0，但是怎么看出上界呢？一个自然的考虑就是，上界是$O(\frac{1}{\sqrt{n}})$的。

**解答：** 用数学归纳法可以证明
$$0<x_n<\frac{1}{\sqrt{2n+1}}$$
因此极限就是0。

> 5. 设数列$\{a_n\}$有界，对于任何$n$总有$a_n\leq a_{n+2},a_n\leq a_{n+3}$，证明$\lim_{n\to\infty}a_n$存在。

**思路：** 这很明显就是子数列的证明题了...常用的结论就是把数列拆成数个子数列，然后证明它们有相同的极限。

**解答：** 取$\{a_n\}$的子列$\{a_{2n}\},\{a_{2n-1}\},\{a_{3n}\}$，由题目中的不等式可知它们单调有界，因此极限存在，分别记为$a,b,c$，接下来需要证明$a=b=c$。首先注意到$\{a_{2n}\}$与$\{a_{3n}\}$有公共子列$\{a_{6n}\}$且也收敛，所以$a=c$。然后考虑$\{a_{2n-1}\}$和$\{a_{3n}\}$的子列$\{a_{6n+3}\}$，同理有$b=c$。

> 6. 如果存在正整数$p$，使得$\lim_{n\to\infty}(a_{n+p}-a_n)=\lambda$，则$\lim_{n\to\infty}\frac{a_n}{n}=\frac{\lambda}{p}$。

**思路：** 别怕，不就多分几个子列吗？

**解答：** 构造$\{\frac{a_n}{n}\}$的子列$\{\frac{a_{np+k}}{np+k}\}$，其中$k=0,1,\cdots,p-1$。对于任一子列，我们有
$$\lim_{n\to\infty}\frac{a_{np+k}}{np+k}=\lim_{n\to\infty}\frac{a_{np+p+k}-a_{np+k}}{p}=\frac{\lambda}{p}$$

### **递推、迭代数列的极限**

> 7. 设数列$x_0=a,x_{n+1}=1+bx_n$，求$a,b$使得该数列收敛。

**思路：** 怪怪的...要不先写几项看看？多写了几项后发现一部分形式很像等比数列的求和。

**解答：** 由数学归纳法可以得到
$$x_n=1+b+\cdots+b^na=\frac{1-b^n}{1-b}+ab^n=\frac{1}{1-b}+(a-\frac{1}{1-b})b^n$$
前面那项无关紧要，重点看后面。若$|b|<1$，则无论$a$取什么值都收敛。但若$|b|\geq 1$，则$a=\frac{1}{1-b}$时数列才收敛。

> 8. 设$x_0=a,x_1=b,x_{n+1}=\frac{x_{n-1}+(2n-1)x_n}{2n}$，求$\lim_{n\to\infty}x_n$。

**思路：** 首先，你需要相信，递推形式肯定能变得更好看的。最自然的想法当然是两边同乘$2n$了。

**解答：** 注意到
$$x_{n+1}-x_n=-\frac{1}{2n}(x_n-x_{n-1})$$
因此
$$x_n-x_{n-1}=-\frac{1}{2(n-1)}(x_{n-1}-x_{n-2})=(-\frac{1}{2})^2\frac{1}{(n-1)(n-2)}(x_{n-2}-x_{n-3})=\cdots=(-\frac{1}{2})^{n-1}\frac{x_1-x_0}{(n-1)!}$$
由此，我们得到
$$\lim_{n\to\infty}x_n=\lim_{n\to\infty}x_0+\sum_{k=0}^{n-1}(x_{k+1}-x_k)=\lim_{n\to\infty}a+\sum_{k=0}^{n-1}(-\frac{1}{2})^k\cdot\frac{b-a}{k!}=a+(b-a)e^{-\frac{1}{2}}$$


> 9. 设数列递推公式$u_1=b,u_{n+1}=u_n^2+(1-2a)u_n+a^2$所确定。当$a,b$为何值时，$\{u_n\}$收敛？收敛到什么值？

**思路：** 这不是我们的完全平方公式吗？当然，得到好看的递推公式是不够的，还需要探寻它的更多性质。

**解答：** 注意到
$$u_{n+1}=u_n+(u_n-a)^2$$
即$\{u_n\}$单调增加，假设$\{u_n\}$的极限存在，记为$A$，则有$A=A+(A-a)^2$，那么$A=a$。若存在$i$使得$u_i>a$，则对所有$j>i$都有$u_j>a$，这显然是不合理的。因此只能是$u_i\leq a$，现在我们强迫$u_{i+1}\leq a$，即需要解$u+(u-a)^2\leq a$，结合所有条件可知$a-1\leq u\leq a$，即当$a-1\leq b\leq a$时，$\{u_n\}$收敛到$a$。

> 10. 设$\{x_n\}$定义如下：$x_1=\sqrt{5},x_{n+1}=x_n^2-2$，求极限$\frac{x_1x_2\cdots x_n}{x_{n+1}}$。

**思路：** 令$y_n=x_n^2$。这个只能说是见过的就赢了。

**解答：** 我们有$y_{n+1}=(y_n-2)^2$，则
$$y_{n+1}-4=y_n(y_n-4)=y_ny_{n-1}(y_{n-1}-4)=\cdots=y_ny_{n-1}\cdots y_2y_1(y_1-4)=y_1y_2\cdots y_n$$
用数学归纳法容易证明$y_n>4$，所以$y_n\to \infty$。因此
$$\lim_{n\to\infty}\frac{y_1y_2\cdots y_n}{y_{n+1}}=\lim_{n\to\infty}\frac{y_{n+1}-4}{y_{n+1}}=1$$
由此，原极限也等于1。

> 11. 设$x_1=2,x_{n+1}=2+\frac{1}{x_n}$，求$\lim_{n\to\infty}x_n$。

**思路：** 这不是单调数列...只好先算后证了。

```python
%matplotlib inline
import matplotlib.pyplot as plt

list = [2]
for i in range(50):
    x = list[i]
    x = 2 + (1/x)
    list.append(x)

plt.plot(list)
plt.xlabel('n', fontsize=15)
plt.ylabel('x_n', fontsize=15)
plt.ylim(2.35, 2.45)
plt.show()
```

**解答：** 假设极限存在，记为$A$。那么我们有
$$\begin{aligned}
A &= 2+\frac{1}{A} \\
A^2 - 2A - 1 &= 0 \\
A &= 1\pm\sqrt{2}
\end{aligned}$$
其中我们取$A=1+\sqrt{2}$。接下来我们证明$\lim_{n\to\infty}|x_n-A|=0$。
$$|x_n-(1+\sqrt{2})|=|2+\frac{1}{x_{n-1}}-1-\sqrt{2}|=|\frac{1}{x_{n-1}}-(1-\sqrt{2})|=|\frac{1}{x_{n-1}}-\frac{1}{A}|\leq \frac{1}{2}|x_{n-1}-A|$$
其中上面不等式的依据是$x_n>2$，这可以由数学归纳法证明。因此，当$n\to\infty$时
$$|x_n-A|\leq \frac{1}{2}|x_{n-1}-A|\leq (\frac{1}{2})^2|x_{n-2}-A|\leq \cdots \leq (\frac{1}{2})^{n-1}|x_1-A|\to 0$$

> 12. 证明数列$\sqrt{7},\sqrt{7-\sqrt{7}},\sqrt{7-\sqrt{7+\sqrt{7}}},\sqrt{7-\sqrt{7+\sqrt{7-\sqrt{7}}}},\cdots$收敛，并求其值。

**思路：** 寻找通项之间的关系吧！

**解答：** 我们注意到$x_{n+2}=\sqrt{7-\sqrt{7+x_n}}$，假设极限存在，记为$A$，那么计算可得$A=2$。接下来需要证明
$$\lim_{n\to\infty}|x_n-A|=0$$
记$f(x)=\sqrt{7-\sqrt{7+x}}$，则
$$|f(x)-2|=|f(x)-f(2)|=|f'(\xi)||x-2|$$
其中$\xi\in(0,7)$
$$|f'(\xi)|=\frac{1}{4\sqrt{7-\sqrt{7+\xi}}\sqrt{7+\xi}}=\alpha<1$$
因此
$$\lim_{n\to\infty}|x_{2n}-2|=\lim_{n\to\infty}\alpha^n|x_0-2|=0$$
$$\lim_{n\to\infty}|x_{2n+1}-2|=\lim_{n\to\infty}\alpha^n|x_1-2|=0$$
即$\lim_{n\to\infty}x_n=2$。

> 13. 设$a_0=0,a_1=1+\sin(-1),a_n=1+\sin(a_{n-1}-1)$，求$\lim_{n\to\infty}\frac{1}{n}\sum_{k=1}^na_k$

**思路：** 觉不觉得$\sin$里面的那一项很难看，想换元？当然，移项也可以看出。

### 数列的收敛速度

> 14. 设$x_1=\sin x_0>0,x_{n+1}=\sin x_n$，证明$\lim_{n\to\infty}x_n=0$且$\lim_{n\to\infty}x_n\sqrt{\frac{n}{3}}=1$

**思路：** 这种题型必须多练！！

```python
%matplotlib inline

list = [1]
for i in range(500):
    x = list[i]
    x = np.sin(x)
    list.append(x)

y = np.linspace(0, 500, 500)
z = np.sqrt(3/y)
plt.figure(figsize=(8, 6))
plt.plot(list, label=r'$x_n$', color='blue')
plt.plot(z, label=r'$\sqrt{3/n}$', color='red')
plt.xlabel('n', fontsize=15)
plt.ylabel(r'$x_n$', fontsize=15)
plt.grid(True, linestyle=':')
plt.legend()
plt.show()
```

**解答：** 容易证明当$x>0$时，$\sin x \leq x$且$\sin x \geq -1$，因此数列$\{x_n\}$单调递减有下界，极限存在，记为$A$。解$A=\sin A$得到$A=0$。下面证明第二个结论。
只需证明：
$$\lim_{n\to\infty}\frac{\frac{1}{x_n^2}}{n}=\frac{1}{3}$$
由`Heine定理`和`Stolz公式`可知：
$$\lim_{n\to\infty}\frac{\frac{1}{x_n^2}}{n}=\lim_{n\to\infty}\frac{1}{x_n^2}-\frac{1}{x_{n-1}^2}=\lim_{n\to\infty}\frac{x_{n-1}^2-x_n^2}{x_n^2x_{n-1}^2}=\lim_{t\to 0^+}\frac{t^2-\sin^2 t}{t^4}=\frac{1}{3}$$

> 15. 设$x_1=\sqrt{2},x_{n+1}=\sqrt{2+x_n}$，证明$x_n-2=o(\frac{1}{3^n})$

**思路：** 证明收敛速度一般都是用比值法的。

```python
%matplotlib inline

list = [np.sqrt(2)]
for i in range(10):
    x = list[i]
    x = np.sqrt(2+x)
    list.append(x)

for i in range(11):
    list[i] -= 2

y = np.linspace(0, 10, 11)
z = np.power(3, -y)
plt.figure(figsize=(8, 6))
plt.plot(list, label=r'$x_n-2$', color='blue')
plt.plot(z, label=r'$\frac{1}{3^n}$', color='red')
plt.ylim(-0.1, 0.1)
plt.grid(True, linestyle=':')
plt.xlabel('n', fontsize=15)
plt.ylabel(r'$x_n$', fontsize=15)
plt.legend()
plt.show()
```

**解答：** $\lim_{n\to\infty}x_n=2$是容易证明的，现在只证明其收敛速度快于$\frac{1}{3^n}$。注意到
$$|x_n-2|=|\sqrt{2+x_{n-1}}-2|=\frac{|x_{n-1}-2|}{\sqrt{2+x_{n-1}}+2}=\frac{|x_{n-2}-2|}{(\sqrt{2+x_{n-1}}+2)(\sqrt{2+x_{n-2}}+2)}=\cdots=\frac{|x_1-2|}{(\sqrt{2+x_1}+2)(\sqrt{2+x_2}+2)\cdots(\sqrt{2+x_{n-1}}+2)}<\frac{|x_1-2|}{(2+\sqrt{2})^{n-1}}$$
因为分母每一项都是大于$2+\sqrt{2}$的，因此下面只需通过比值法验证结论即可。
$$\lim_{n\to\infty}\frac{\frac{|x_1-2|}{(2+\sqrt{2})^{n-1}}}{\frac{1}{3^n}}=\lim_{n\to\infty}3|x_1-2|(\frac{3}{2+\sqrt{2}})^{n-1}=0$$

> 16. 设$f(x)=\frac{x+2}{x+1},x_0=1,x_{n+1}=f(x_n)$，求$\lim_{n\to\infty}x_n$，并估计数列趋于极限的速度。

**思路：** 感觉又不是单调数列。。。估计要先算后证了

```python
%matplotlib inline

list = [1]
for i in range(10):
    x = list[i]
    x = 1 + (1/(x+1))
    list.append(x)

for i in range(11):
    list[i] -= np.sqrt(2)

y = np.linspace(0, 10, 11)
z = np.power(4, -y)
plt.plot(list, label=r'$x_n-\sqrt{2}$', color='blue')
plt.plot(z, label=r'$\frac{1}{4^n}$', color='red')
plt.xlabel('n', fontsize=15)
plt.ylabel(r'$x_n$', fontsize=15)
plt.ylim(-0.1, 0.1)
plt.grid(True, linestyle=':')
plt.legend()
plt.show()
```

**解答：** 设极限存在，记为$A$，则容易得到$A=\sqrt{2}$。现在证明$\lim_{n\to\infty}|x_n-A|=0$。
$$|x_n-\sqrt{2}|=|\frac{1}{x_{n-1}+1}-(\sqrt{2}-1)|=|\frac{1}{x_{n-1}+1}-\frac{1}{\sqrt{2}+1}|=\frac{|x_{n-1}-\sqrt{2}|}{(x_{n-1}+1)(\sqrt{2}+1)}<\frac{1}{4}|x_{n-1}-\sqrt{2}|$$
因此
$$\lim_{n\to\infty}|x_n-\sqrt{2}|=\lim_{n\to\infty}(\frac{1}{4})^n|x_0-\sqrt{2}|=0$$
由上述证明过程可以看出，数列收敛到$\sqrt{2}$的速度要比$\frac{1}{4^n}$收敛到0的速度快。

### 数列极限的综合题

> 17. 设$x_1,x_2,\cdots,x_n,\cdots$是将方程$\tan x=x$的全部正根按从小到大的次序排列的数列，求极限$\lim_{n\to\infty}(x_n-x_{n-1})$。

**思路：** 几何直观很重要！可以考虑两边取$\arctan$
$$\arctan(\tan x)=\arctan x$$

```python
%matplotlib inline

x = np.linspace(0, 30, 30000)
x_tan = np.delete(x, np.where(np.abs(np.cos(x)) < 0.01))

y_tan = np.tan(x_tan) 
y_x = x

plt.figure(figsize=(8, 6))

plt.scatter(x_tan, y_tan, s=2, label=r'$y = \tan(x)$', color='blue',alpha=0.7)

plt.plot(x, y_x, label=r'$y = x$', color='red', linewidth=2.5)

asymptotes = [np.pi/2 + k*np.pi for k in range(0, 17)]
for pos in asymptotes:
    plt.axvline(x=pos, color='gray', linestyle=':', alpha=0.4)

plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.grid(True, linestyle=':', alpha=0.3)

plt.xlim(0, 30)
plt.ylim(-30, 30)

plt.legend(loc='upper left')

plt.tight_layout()
plt.show()
```

**解答：** $x_n$是$y=\tan x$与$y=x$的交点的横坐标。设$y_n=\arctan x_n$，则由$\tan x$的周期性可知$x_n=n\pi + y_n$。容易知道$\{y_n\}$和$\{x_n\}$一样都是单调递增的，但是$\{y_n\}$有上界$\frac{\pi}{2}$，因此收敛，即$\lim_{n\to\infty}(y_n-y_{n-1})=0$。所以
$$\lim_{n\to\infty}(x_n-x_{n-1})=\lim_{n\to\infty}(\pi+y_n-y_{n-1})=\pi$$

> 18. 设$a>0,f_n(x)=x^n+nx-a$，证明$f_n(x)$在$(0,+\infty)$上有唯一的零点$x_n$，并求$\lim_{n\to\infty}(1+x_n)^n$。

**思路：** 唯一零点是挺简单的，一般都是单调递增/递减函数，然后取一个函数值大于0和小于0的点就行。然而，后面的极限用夹逼准则我真是没想出来。

**解答：** 首先注意到$f_n(x)$连续，且$f_n(0)=-a<0,f_n(+\infty)=+\infty>0$。此外$f'_n(x)=nx^{n-1}+n>0$，因此$f_n(x)$在$(0,\infty)$上单调递增。根据介值定理，$f_n(x)$在$(0,+\infty)$上有唯一的零点$x_n$。然后，当$n>a$时，$\frac{a}{n}-\frac{a}{n^2}\leq x_n \leq \frac{a}{n}$。由夹逼准则可知极限为$e^a$。

> 19. 证明$f_n(x)=e^x+x^{2n+1}$有唯一的实根$x_n$，证明$\lim_{n\to\infty}x_n=A$存在并求出$A$，最后证明$n\to\infty$时，$x_n-A$和$\frac{1}{n}$是同阶无穷小。

**思路：** 唯一实根比较容易，要求出$x_n$的极限，就要考虑能不能利用等式来对其进行变形。比阶数无非就那几招：`比值法`，`Stolz定理`，`拉格朗日中值定理`。。。

**解答：** 注意到
$$f_n(-1)=\frac{1}{e}-1<0, f_n(1)=e+1>0, f'_n(x)=e^x+(2n+1)x^{2n}>0$$
因此$f_n(x)$在$(-1,1)$有唯一的实根$x_n$。然后我们有
$$x_n^{2n+1}=-e^{x_n}, x_n=-e^{\frac{x_n}{2n+1}}$$
因此
$$\lim_{n\to\infty}x_n=-\exp[\lim_{n\to\infty}\frac{x_n}{2n+1}]=-1$$
最后，由`拉格朗日中值定理`：
$$x_n-A=1-e^{\frac{x_n}{2n+1}}=e^{\xi}(0-\frac{x_n}{2n+1}),(\xi\to 0)$$
所以
$$\lim_{n\to\infty}\frac{x_n-A}{1/n}=\lim_{n\to\infty}-e^{\xi}x_n\frac{n}{2n+1}=\frac{1}{2}$$

> 20. 给定3个实数$a_1, b_1, c_1$。将每个数换成另外两个的算术平均值，即$a_2=\frac{b_1+c_1}{2},b_2=\frac{a_1+c_1}{2},c_2=\frac{a_1+b_1}{2}$，如此下去构成三个数列$\{a_n\},\{b_n\},\{c_n\}$。证明：这三个数列的极限都等于$\frac{S}{3},S=a_1+b_1+c_1.$

**思路：** 对数列两两作差。这就是纯纯的套路，初见谁能想出来啊。

**解答：** 令$A_n=a_n-b_n, B_n=b_n-c_n, C_n=c_n-a_n$。计算得到
$$A_n=-\frac{1}{2}A_{n-1}, B_n=-\frac{1}{2}B_{n-1}, C_n=-\frac{1}{2}C_{n-1}$$
因此这三个数列的极限都是0.然后注意到
$$a_n+b_n+c_n=a_{n-1}+b_{n-1}+c_{n-1}=\cdots=S$$
因此
$$\lim_{n\to\infty}a_n=\lim_{n\to\infty}(S-b_n-c_n)=\lim_{n\to\infty}\frac{1}{3}(S+A_n-C_n)=\frac{S}{3}$$
同理可证其他。

> 21. 若
> $$\lim_{n\to\infty}\frac{n^{2009}}{n^b-(n-1)^b}=a\neq 0$$
> 求$a, b$的值

**思路1:** 考虑分母，当$n\to\infty$，
$$n^b-(n-1)^b=n^b(1-(1-\frac{1}{n})^b)=n^b(1-e^{b\ln(1-\frac{1}{n})})\sim n^b(-b\ln(1-\frac{1}{n}))\sim n^b(-b\cdot (-\frac{1}{n}))=bn^{b-1}$$

**解答1:** 因为
$$\lim_{n\to\infty}\frac{n^{2009}}{n^b-(n-1)^b}=\lim_{n\to\infty}\frac{n^{2009}}{bn^{b-1}}=\lim_{n\to\infty}b^{-1}n^{2010-b}=a\neq 0$$
所以$b=2010, a=\frac{1}{2010}$

**思路2:** 考虑分母
$$n^b-(n-1)^b=n^b-\sum_{k=0}^b\binom{b}{k}n^k(-1)^{n-k}=\sum_{k=0}^{b-1}\binom{b}{k}n^k(-1)^{n-k}$$

**解答2:** 因为
$$\lim_{n\to\infty}\frac{n^{2009}}{n^b-(n-1)^b}=\lim_{n\to\infty}\frac{n^{2009}}{\sum_{k=0}^{b-1}\binom{b}{k}n^k(-1)^{n-k}}=a\neq 0$$
根据`“抓大头”`，其实也就是同除以$n^{2009}$，$b=2010$。此时$a=(\binom{2010}{2009})^{-1}=\frac{1}{2010}$

> 23. 求下列曲线的渐近线
> $$(1)y=xe^{x^{-2}};(2)y=\frac{x^{1+x}}{(1+x)^x}$$

**思路：** 垂直渐近线是最容易被忽略的，斜渐近线和水平渐近线还好说。斜渐近线的求法：以正无穷方向为例，设斜渐近线为$y=ax+b$，则
$$a=\lim_{x\to +\infty}\frac{f(x)}{x}$$
$$b=\lim_{x\to +\infty}(f(x)-ax)$$

**解答：** $(1)$ 因为
$$\lim_{x\to 0}xe^{\frac{1}{x^2}}=+\infty$$
因此曲线有垂直渐近线$x=0$，下面看斜渐近线，设斜渐近线为$y=ax+b$
$$a=\lim_{x\to +\infty}\frac{xe^{\frac{1}{x^2}}}{x}=\lim_{x\to +\infty}e^{\frac{1}{x^2}}=1$$
$$b=\lim_{x\to +\infty}(xe^{\frac{1}{x^2}}-x)=\lim_{x\to +\infty}x(e^{\frac{1}{x^2}}-1)=\lim_{x\to +\infty}\frac{x}{x^2}=0$$
即一条斜渐近线为
$$y=x$$
对$x\to -\infty$进行相似的计算可以得到同一条斜渐近线。

$(2)$ 对$x=0$处进行试探发现并没有垂直渐近线，于是直接开始求斜渐近线。设斜渐近线为$y=ax+b$
$$a=\lim_{x\to +\infty}\frac{x^{1+x}}{(1+x)^x}\cdot\frac{1}{x}=\lim_{x\to +\infty}(1-\frac{1}{1+x})^x=e^{-1}$$
$$b=\lim_{x\to +\infty}x(\frac{1}{(1+\frac{1}{x})^x}-\frac{1}{e})=\lim_{t\to 0^+}\frac{e(1+t)^{-\frac{1}{t}}-1}{et}=\frac{1}{e}\lim_{t\to 0^+}\frac{e^{1-\frac{1}{t}\ln(1+t)}-1}{t}=\frac{1}{e}\lim_{t\to 0^+}\frac{1-\frac{1}{t}\ln(1+t)}{t}=\frac{1}{e}\lim_{t\to 0^+}\frac{1-\frac{1}{t}(t-\frac{t^2}{2}+o(t^2))}{t}=\frac{1}{2e}$$
因此斜渐近线为
$$y=\frac{x}{e}+\frac{1}{2e}$$

> 24. 设$a>0$，参数方程：$x=\frac{3at}{1+t^3},y=\frac{3at^2}{1+t^3}$表示的曲线称为笛卡尔叶形线。
> 
> $(1)$求出该曲线在直角坐标下的方程。
>
> $(2)$仅就直角坐标下的方程求出该曲线的渐近线。
>
> $(3)$仅就参数方程求出该曲线的渐近线。

**思路：** 这个题属于是典型的百闻不如一见，做过一次就会了，没做过的怎么都想不出来，特别是后两小问。

```python
%matplotlib inline
a = 1
x = np.linspace(-4, 4, 1000)
y = np.linspace(-4, 4, 1000)
X, Y = np.meshgrid(x, y)

F = X**3 + Y**3 - 3*a*X*Y

plt.figure(figsize=(8, 8))
contour = plt.contour(X, Y, F, levels=[0], colors='crimson', linewidths=2.5)

plt.plot(x, -x - a, ':', color='teal', alpha=0.7)

plt.grid(alpha=0.3)
plt.axis('equal')
plt.xlabel('x', fontsize=15)
plt.ylabel('y', fontsize=15)
plt.show()
```

**解答：** $(1)$注意到
$$\frac{y}{x}=t$$
因此
$$x=\frac{3a\frac{y}{x}}{1+\frac{y^3}{x^3}}$$
$$x^3+y^3=3axy$$

$(2)$方程可变形为
$$\frac{x^3}{y}+y^2=3ax$$
先看看有没有垂直渐近线，如果有垂直渐近线$x=c$，对方程两边同时取$x\to c$的极限，左边趋向正无穷，右边趋向定值$3ac$，矛盾！故曲线没有垂直渐近线。对于斜渐近线，设为$y=kx+b$，方程可变形为
$$\frac{1}{\frac{y}{x}}+(\frac{y}{x})^2=\frac{3a}{x}$$
从这个变形我们就能看出$\lim_{x\to\pm\infty}\frac{y}{x}=k$存在。在等式两边令$x\to\pm\infty$有
$$\frac{1}{k}+k^2=0$$
得到$k=-1$，然后我们有
$$b=\lim_{x\to\pm\infty}(y+x)=\lim_{x\to\pm\infty}\frac{x^3+y^3}{x^2-xy+y^2}=\lim_{x\to\pm\infty}\frac{3axy}{x^2-xy+y^2}=\lim_{x\to\pm\infty}\frac{3a\frac{y}{x}}{1-\frac{y}{x}+(\frac{y}{x})^2}=-a$$
因此渐近线就是
$$y=-x-a$$

$(3)$可以看出$x\to\pm\infty$的等价条件就是$t\to -1$，因此
$$k=\lim_{x\to\pm\infty}\frac{y}{x}=\lim_{t\to -1}t=-1$$
$$b=\lim_{x\to\pm\infty}(y+x)=\lim_{t\to -1}\frac{3a(t^2+t)}{1+t^3}=\lim_{t\to -1}\frac{3a(2t+1)}{3t^2}=-a$$
同样可以得到渐近线为
$$y=-x-a$$

> 25. 求极坐标方程的曲线$\Gamma:\rho=\frac{1}{3\theta-\pi}$在直角坐标下的渐近线方程。

**思路：** 根据两个坐标之间的转化：
$$\begin{cases}
x=\rho\cos\theta \\
y=\rho\sin\theta
\end{cases}$$
仿照上题的$(3)$即可轻松求解。

```python
%matplotlib inline
theta = np.linspace(0, 2*np.pi, 1000)
x = np.cos(theta) / (3*theta-np.pi)
y = np.sin(theta) / (3*theta-np.pi)
X, Y = np.meshgrid(x, y)

F = np.tan(theta)*X - Y

plt.figure(figsize=(8, 8))
contour = plt.contour(X, Y, F, levels=[0], colors='crimson', linewidths=2.5)

plt.plot(x, np.sqrt(3)*x+(2/3), ':', color='teal', alpha=0.7)

plt.grid(alpha=0.3)
plt.axis('equal')
plt.xlabel('x', fontsize=15)
plt.ylabel('y', fontsize=15)
plt.show()
```

**解答：** 因为
$$x=\rho\cos\theta=\frac{\cos\theta}{3\theta-\pi},y=\rho\sin\theta=\frac{\sin\theta}{3\theta-\pi}$$
可以看出$x\to\pm\infty$的等价条件就是$\theta\to\frac{\pi}{3}$。设渐近线为$y=kx+b$，计算可得
$$k=\lim_{x\to\pm\infty}\frac{y}{x}=\lim_{\theta\to\frac{\pi}{3}}\tan\theta=\sqrt{3}$$
$$b=\lim_{x\to\pm\infty}(y-\sqrt{3}x)=\lim_{\theta\to\frac{\pi}{3}}\frac{\sin\theta-\sqrt{3}\cos\theta}{3\theta-\pi}=\lim_{\theta\to\frac{\pi}{3}}\frac{\cos\theta+\sqrt{3}\sin\theta}{3}=\frac{2}{3}$$
因此曲线在直角坐标下的渐近线方程就是
$$y=\sqrt{3}x+\frac{2}{3}$$

> 27. 设$f_1(x)=x,f_2(x)=x^x,f_3(x)=x^{x^x},\cdots$。证明
> $$\lim_{x\to 0^+} f_n(x)=\begin{cases}
0,&\text{当n是奇数} \\
1,&\text{当n是偶数}
\end{cases}$$

**思路：** 这种无限递归的形式用数学归纳法比较方便。

**解答：** 用数学归纳法，当$n=1, 2$时，显然有
$$\lim_{x\to 0^+}f_1(x)=\lim_{x\to 0^+}x=0$$
$$\lim_{x\to 0^+}f_2(x)=\lim_{x\to 0^+}x^x=1$$
假设当$n=2k-1, 2k$时结论成立，则当$n=2k+1, 2k+2$时，由归纳假定
$$\lim_{x\to 0^+}f_{2k+1}(x)=\lim_{x\to 0^+}x^{f_{2k}(x)}=0$$
设$0<x<\delta<1$，使得
$$\frac{1}{2}<f_{2k}(x)<\frac{3}{2}$$
则有
$$x^{\frac{3}{2}}<x^{f_{2k}(x)}<x^{\frac{1}{2}}$$
从而
$$x^{x^{\frac{1}{2}}}<x^{x^{f_{2k}(x)}}<x^{x^{\frac{3}{2}}}$$
但是我们知道
$$\lim_{x\to 0^+}x^{x^{\frac{1}{2}}}=\lim_{x\to 0^+}x^{x^{\frac{3}{2}}}=1$$
因此由夹逼定理，我们得到
$$\lim_{x\to 0^+}f_{2k+2}(x)=\lim_{x\to 0^+}x^{x^{f_{2k}(x)}}=1$$
*由数学归纳法原理可知结论成立*

> 28. 设在$(-\infty, +\infty)$上$f(x)\ge 0$，且对于任意$x_1, x_2$，总有$\frac{1}{2}f(x_1)+\frac{1}{2}f(x_2)\le f(\frac{x_1+x_2}{2})$成立。证明：$f(x)$恒为常数。

**思路：** 证明函数在每一点的取值都相同是很难的，不如用反证法假设有一点大于(小于)另一点，引出矛盾。根据题目给出的形式来看，其中一点最好取0.

**解答：** 假设有一点$x_1$使得$f(x_1)<f(0)$，则
$$f(x_1)=f(\frac{0+2x_1}{2})\ge \frac{1}{2}f(0)+\frac{1}{2}f(2x_1)\ge \frac{1}{2}f(0)+\frac{1}{4}f(0)+\frac{1}{4}f(4x_1)\ge\cdots$$
$$f(x_1)\ge \frac{1}{2}f(0)+\frac{1}{4}f(0)+\frac{1}{8}f(0)+\cdots+\frac{1}{2^n}f(0)+\frac{1}{2^n}f(2^nx_1)$$
注意题设条件中的$f(x)\ge 0$，则
$$f(0)>f(x_1)\ge \frac{1}{2}f(0)+\frac{1}{4}f(0)+\frac{1}{8}f(0)+\cdots+\frac{1}{2^n}f(0)=f(0)\cdot\frac{\frac{1}{2}-(\frac{1}{2})^n}{1-\frac{1}{2}}$$
取$n\to\infty$的极限，我们得到
$$f(0)>f(x_1)\ge f(0)$$
这显然矛盾，故没有这样的点。又假设有点$x_2$使得$f(x_2)>f(0)$，则
$$f(0)=f(\frac{-x_2+x_2}{2})\ge \frac{1}{2}f(-x_2)+\frac{1}{2}f(x_2)>\frac{1}{2}f(-x_2)+\frac{1}{2}f(0)$$
$$f(-x_2)<f(0)$$
由上面的证明可推出矛盾，故$f(x)\equiv f(0)$。
