## 4.2的练习

### 多元函数微分学的几何应用

> 1. 求曲线$l:x=t-\sin t, y=1-\cos t, z=4\sin\frac{t}{2}$在点$(\frac{\pi}{2}-1, 1, 2\sqrt{2})$处的切线和法平面。

**思路：** 考察的是曲线在参数方程下的切线和法平面的求法。

**解答：** 容易知道该点所对应的参数$t_0=\frac{\pi}{2}$，然后可以求出
$$\begin{aligned}
    x'(t)&=1-\cos t \\
    y'(t)&=\sin t \\
    z'(t)&=2\cos\frac{t}{2}
\end{aligned}$$
因此切线为
$$x+1-\frac{\pi}{2}=y-1=\frac{z}{\sqrt{2}}-2$$
法平面方程为
$$\begin{aligned}
    x+1-\frac{\pi}{2}+y-1+\sqrt{2}(z-2\sqrt{2})&=0 \\
    x+y+\sqrt{2}z&=4+\frac{\pi}{2}
\end{aligned}$$

> 2. 求空间曲线$L:x(t)=t(\mathrm{e}^t-1), y(t)=t\sin t, z(t)=t^3+t^2+1$，在点$P_0(0, 0, 1)$处的切线。

**思路：** 和上题完全类似...吗？

**解答：** 容易看出对应的参数$t_0=0$，然后计算可得切向量为$(x'(0), y'(0), z'(0))=(0, 0, 0)$，不符合切向量不为零向量的要求，只能回归定义。设$Q(x, y, z)$为$L$上任意一点，对应的参数为$\Delta t$，则割线$PQ$的方向向量为$\overrightarrow{PQ}=(x, y, z-1)$，注意到$\Delta t$是个常数，因此也可取割线的方向向量为$\mathbf{F}=(\frac{x}{\Delta t^2}, \frac{y}{\Delta t^2}, \frac{z-1}{\Delta t^2})$。当$\Delta t\to 0$，$Q$就趋近于$P_0$，此时$\mathbf{F}$也可以作为点$P_0$处的切向量，计算可得
$$\begin{aligned}
    \lim_{\Delta t\to 0}\frac{x(\Delta t)}{\Delta t^2}&=\lim_{\Delta t\to 0}\frac{\mathrm{e}^{\Delta t}-1}{\Delta t}=1 \\
    \lim_{\Delta t\to 0}\frac{y(\Delta t)}{\Delta t^2}&=\lim_{\Delta t\to 0}\frac{\sin\Delta t}{\Delta t}=1 \\
    \lim_{\Delta t\to 0}\frac{z(\Delta t)-1}{\Delta t^2}&=\lim_{\Delta t\to 0}(\Delta t+1)=1
\end{aligned}$$
所以切线方程为
$$\frac{x}{1}=\frac{y}{1}=\frac{z-1}{1}$$

**总结：** 当公式失效时，就需要通过定义来求切线，$\Delta t$取多少次这个问题，感觉就像是在做泰勒展开的极限题。

> 3. 设曲线$L$是由两个曲面$F(x, y, z)=x+y+z+z^2=0, G(x, y, z)=x+y^2+z+z^3=0$相交而成。点$P(1, -1, -1)$在$L$上。求点$P$处的切线在$xOy$坐标面上的投影。

**思路：** 考察的是隐函数形式的切线方程。

**解答1：** 方程组确定了两个隐函数$y=y(x), z=z(x)$，在方程组两端同时对$x$求导可得
$$\begin{cases}
    1+y'+z'+2zz'=0 \\
    1+2yy'+z'+3z^2z'=0
\end{cases}$$
代入点$P$坐标可得
$$\begin{cases}
    1+y'(1)-z'(1)=0 \\
    1-2y'(1)+4z'(1)=0
\end{cases}$$
解得$y'(1)=-\frac{5}{2}, z'(1)=-\frac{3}{2}$，因此点$P$处的切线为
$$\frac{x-1}{1}=\frac{y+1}{-5/2}=\frac{z+1}{-3/2}$$
它在$xOy$坐标上的投影是
$$\frac{x-1}{1}=\frac{y+1}{-5/2}$$
也就是$5x+2y-3=0$。

**解答2:** 两个曲面的法向量依次为$\mathbf{n}_F=(1, 1, 1+2z), \mathbf{n}_G=(1, 2y, 1+3z^2)$。在点$P$处就有$\mathbf{n}_F=(1, 1, -1), \mathbf{n}_G=(1, -2, 4)$。这两个曲面切平面的交线$L$就是点$P$处的切线
$$\begin{cases}
    (x-1)+(y+1)-(z+1)=0 \\
    (x-1)-2(y+1)+4(z+1)=0
\end{cases}$$
从这两个方程中消去$z$，就可以得到投影为$5x+2y-3=0$。

**总结：** 隐函数法非常通用，但由切平面之间的交线求切线的方法似乎能减少一点计算量，也值得学习。

> 4. 求下列曲面的切平面和法线：
> 
> $(1)$ 求曲面$\mathrm{e}^z-z+xy=3$，在点$(2, 1, 0)$处的切平面和法线。
>
> $(2)$ 求椭球面$x^2+2y^2+z^2=1$上平行于平面$x-y+2z=0$的切平面。
>
> $(3)$ 求曲面$z=xy$上一点，使得该点处的法线垂直于平面$x+3y+z+9=0$，并求法线方程。

**思路：** 一直求法向量即可。

**解答：** $(1)$ 容易求得在曲面上一点$M(x, y, z)$处的法向量为$(y, x, \mathrm{e}^z-1)$，代入点$(2, 1, 0)$可得所需切向量$\mathbf{v}=(1, 2, 0)$，因此切平面方程为
$$\begin{aligned}
    (x-2)+2(y-1)&=0 \\
    x+2y&=4
\end{aligned}$$
法线方程为
$$\frac{x-2}{1}=\frac{y-1}{2}=\frac{z}{0}$$

$(2)$ 容易求得在椭球面上一点$M(x, y, z)$处的法向量为$(2x, 4y, 2z)$，从而$k(x, 2y, z)(k\neq 0)$也是其法向量。平面的法向量为$(1, -1, 2)$，比较可知
$$\begin{cases}
    kx=1 \\
    2ky=-1 \\
    kz=2
\end{cases}$$
代入椭球面方程可得
$$\begin{aligned}
    \left(\frac{1}{k}\right)^2+2\left(-\frac{1}{2k}\right)^2+\left(\frac{2}{k}\right)^2&=1 \\
    \frac{11}{2k^2}&=1 \\
    k&=\pm\sqrt{\frac{11}{2}}
\end{aligned}$$
由此得到椭球面上两个点
$$P_1\left(\frac{\sqrt{2}}{\sqrt{11}}, -\frac{\sqrt{2}}{2\sqrt{11}},\frac{2\sqrt{2}}{\sqrt{11}}\right), P_2\left(-\frac{\sqrt{2}}{\sqrt{11}}, \frac{\sqrt{2}}{2\sqrt{11}}, -\frac{2\sqrt{2}}{\sqrt{11}}\right)$$
从而可得两个切平面方程
$$\begin{aligned}
    (x-\frac{\sqrt{2}}{\sqrt{11}})-(y+\frac{\sqrt{2}}{2\sqrt{11}})+2(z-\frac{2\sqrt{2}}{\sqrt{11}})&=0 \\
    x-y+2z&=\sqrt{\frac{11}{2}}
\end{aligned}$$
以及
$$x-y+2z=-\sqrt{\frac{11}{2}}$$

$(3)$ 令$F(x, y, z)=xy-z=0$，容易计算得到曲面上一点$M(x, y, z)$处的法向量为$(y, x, -1)$，从而$k(y, x, -1)(k\neq 0)$也是其法向量。平面的法向量为$(1, 3, 1)$，因此
$$\begin{cases}
    ky=1 \\
    kx=3 \\
    -k=1
\end{cases}$$
从而可求出点为$(-3, -1, 3)$，法线为
$$\frac{x+3}{1}=\frac{y+1}{3}=\frac{z-3}{1}$$

> 5. 求过直线$L:\begin{cases}
    3x-2y-z=5 \\
    x+y+z=0
\end{cases}$与曲面$2x^2-2y^2+2z=\frac{5}{8}$相切的切平面方程。

**思路：** 重点在于写出过直线$L$的平面方程，剩下的就是计算问题了。

**解答：** 过直线$L$的平面方程为
$$(3x-2y-z-5)+\lambda(x+y+z)=0, (\lambda\in\R)$$
其法向量为$(3+\lambda, -2+\lambda, -1+\lambda)$。曲面上一点$M(x, y, z)$的法向量为$(4x, -4y, 2)$，也可以是$k(2x, -2y, 1)$。比较可得$\lambda=1+k$，因此有
$$\begin{cases}
    4+k=2kx \\
    -1+k=-2ky
\end{cases}$$
也就是
$$\begin{cases}
    x=\frac{4+k}{2k} \\
    y=\frac{1-k}{2k}
\end{cases}$$
代入平面方程有
$$\begin{aligned}
    \frac{(k+4)^2}{2k}-\frac{(k-1)^2}{2k}+kz&=5 \\
    10k+15+2k^2z&=10k \\
    z&=-\frac{15}{2k^2}
\end{aligned}$$
代入曲面方程可得
$$\begin{aligned}
    2\cdot\left(\frac{k+4}{2k}\right)^2-2\cdot\left(\frac{1-k}{2k}\right)^2-\frac{30}{2k^2}&=\frac{5}{8} \\
    \frac{(k+4)^2-(k-1)^2-30}{2k^2}&=\frac{5}{8} \\
    8(10k+15-30)&=10k^2 \\
    k^2-4(2k-3)&=0 \\
    (k-2)(k-6)&=0
\end{aligned}$$
当$k=2$时，切平面方程为
$$\begin{aligned}
    6(x-\frac{3}{2})+(y+\frac{1}{4})+2(z+\frac{15}{8})&=0 \\
    6x+y+2z&=5
\end{aligned}$$
当$k=6$时，切平面方程为
$$\begin{aligned}
    10(x-\frac{5}{6})+5(y+\frac{5}{12})+6(z+\frac{15}{72})&=0 \\
    10x+5y+6z&=5
\end{aligned}$$

> 6. 长度为$a$的线段两端分别在$x$轴和$y$轴上滑动，求这样的线段簇的包络线。

**思路：** 可参考正文例2，但首先要写出线段簇的方程。

**解答：** 由对称性，我们只需考虑第一象限的情况。设此时一端在点$A(c, 0)$上，其中$0<c<a$，则另一端在点$B(0, \sqrt{a^2-c^2})$上。那么，线段的方程为
$$\frac{x}{c}+\frac{y}{\sqrt{a^2-c^2}}=1$$
也就是
$$\frac{x}{\frac{c}{a}}+\frac{y}{\sqrt{1-(\frac{c}{a})^2}}=a$$
令$\frac{c}{a}=\sin\theta$，则方程可写为
$$\frac{x}{\sin\theta}+\frac{y}{\cos\theta}=a$$
其中$\theta\in(0, \frac{\pi}{2})$。由此，我们只需解方程组
$$\begin{cases}
    F=\frac{x}{\sin\theta}+\frac{y}{\cos\theta}-a=0 \\
    F_{\theta}=-\frac{x}{\sin^2\theta}\cos\theta+\frac{y}{\cos^2\theta}\sin\theta=0
\end{cases}$$
由第二式可得
$$\frac{x}{\sin^3\theta}=\frac{y}{\cos^3\theta}$$
设$x=k\sin^3\theta, y=k\cos^3\theta$，代入第一式可得$k=a$，从而$x=a\sin^3\theta, b=a\cos^3\theta$。包络线的方程为
$$x^{\frac{2}{3}}+y^{\frac{2}{3}}=a^{\frac{2}{3}}$$

> 7. 设椭圆簇为$\frac{x^2}{a^2}+\frac{y^2}{b^2}=1$，其中的每个椭圆的面积都为常数$S$。求该椭圆簇的包络线。

**思路：** 只需通过$S$是定值，对方程进行变形，然后套用上一题的做法即可。

**解答：** 由椭圆面积公式$S=\pi ab$可知$b=\frac{S}{\pi a}$，从而椭圆簇方程可写为
$$\frac{x^2}{a^2}+\frac{(\pi ay)^2}{S^2}=1$$
只需解方程组
$$\begin{cases}
    F=\frac{x^2}{a^2}+\frac{(\pi ay)^2}{S^2}-1=0 \\
    F_a=-\frac{2x^2}{a^3}+\frac{2a\pi^2y^2}{S^2}=0
\end{cases}$$
计算可知
$$\begin{aligned}
    \frac{2}{a}\cdot\left[\frac{(\pi ay)^2}{S^2}-1\right]+\frac{2a\pi^2y^2}{S^2}&=0 \\
    \frac{a\pi^2y^2}{S^2}+\frac{a\pi^2y^2}{S^2}&=\frac{1}{a} \\
    \frac{(\pi ay)^2}{S^2}&=\frac{1}{2}
\end{aligned}$$
代入第一式可得$a^2=2x^2$，再代入上述结果就得到
$$\begin{aligned}
    \frac{2\pi^2x^2y^2}{S^2}&=\frac{1}{2} \\
    |xy|&=\frac{S}{2\pi}
\end{aligned}$$

> 8. 以曲线$L:y=x^2$上的每一点为圆心，$R$为半径作圆。求这样的圆簇的包络线。验证所求的包络线与$L$的距离为$R$的等距线是一样的。

**思路：** 除了典中典的解方程组求包络线外，本题还需要求等距线，回顾正文例3就知道该怎么做了。

**解答：** 圆簇的方程可写为
$$(x-c)^2+(y-c^2)^2=R^2$$
其中$c\in\R$。要求包络线就只需解方程组
$$\begin{cases}
    F=(x-c)^2+(y-c^2)^2-R^2=0 \\
    F_c=-2(x-c)-4c(y-c^2)=0
\end{cases}$$
计算可知
$$\begin{aligned}
    x-c&=-2c(y-c^2) \\
    (x-c)^2&=4c^2(y-c^2)^2
\end{aligned}$$
代入第一式可得
$$\begin{aligned}
    (4c^2+1)(y-c^2)^2&=R^2 \\
    y-c^2&=\pm\frac{R}{\sqrt{4c^2+1}} \\
    y&=c^2\pm\frac{R}{\sqrt{4c^2+1}}
\end{aligned}$$
由此可得
$$\begin{aligned}
    x-c&=-2c\cdot\left(\pm\frac{R}{\sqrt{4c^2+1}}\right) \\
    x&=c\mp\frac{2cR}{\sqrt{4c^2+1}}
\end{aligned}$$
因此包络线有两条。现在求$L$的距离为$R$的等距线。任取$L$上一点$M(c, c^2)$，则在该点的切向量是$(1, 2c)$，因此单位法向量是$(-\frac{2c}{\sqrt{4c^2+1}},\frac{1}{\sqrt{4c^2+1}})$。等距线上对应的点为
$$M_1\left(c-\frac{2cR}{\sqrt{4c^2+1}},c^2+\frac{R}{\sqrt{4c^2+1}}\right), M_2\left(c+\frac{2cR}{\sqrt{4c^2+1}},c^2-\frac{R}{\sqrt{4c^2+1}}\right)$$
这就证明了结论。

> 9. 设椭圆$C:\frac{x^2}{a^2}+\frac{y^2}{b^2}=1$。求与$C$的距离为$h$的外等距线$C'$的方程。

**思路：** 和上题思路一致，尝试使用换元法来简化计算。

**解答：** 设$x=a\cos\theta, y=b\sin\theta$，则在$C$上任一点$M(x, y)$处的切向量为$(-a\sin\theta, b\cos\theta)$，因此单位法向量为
$$\mathbf{v}=\pm\left(\frac{b\cos\theta}{\sqrt{a^2\sin^2\theta+b^2\cos^2\theta}}, \frac{a\sin\theta}{\sqrt{a^2\sin^2\theta+b^2\cos^2\theta}}\right)$$
由此可得，与$C$的距离为$h$的外等距线$C'$上对应的点为
$$N\left(a\cos\theta+\frac{bh\cos\theta}{\sqrt{a^2\sin^2\theta+b^2\cos^2\theta}}, b\sin\theta+\frac{ah\sin\theta}{\sqrt{a^2\sin^2\theta+b^2\cos^2\theta}}\right)$$
因此，与$C$的距离为$h$的外等距线$C'$的方程就是
$$\begin{aligned}
    x&=a\cos\theta+\frac{bh\cos\theta}{\sqrt{a^2\sin^2\theta+b^2\cos^2\theta}} \\
    y&=b\sin\theta+\frac{ah\sin\theta}{\sqrt{a^2\sin^2\theta+b^2\cos^2\theta}}
\end{aligned}$$
其中$\theta\in[0, 2\pi]$。

> 10. 如果直线与椭球面$\frac{x^2}{a^2}+\frac{y^2}{b^2}+\frac{z^2}{c^2}=1$只有一个交点(称为切点)，则称该直线是椭球面的一条切线。
>
> $(1)$ 若$\alpha=(l, m, n)$是非零向量，证明：椭球面上所有平行于$\alpha$的切线的切点都位于过原点的某个平面上。
>
> $(2)$ 若$P(u, v, w)$是椭球面之外的一个点，证明：椭球面上所有过点$P$的切线的切点都在某个平面上。

**思路：** 曲面的切线？没学过啊！但是我们可以将其转化为切平面来求解。

**解答：** 椭球面上任一点$M(x_0, y_0, z_0)$的法向量为$(\frac{x_0}{a^2}, \frac{y_0}{b^2}, \frac{z_0}{c^2})$，因此过该点的切平面为
$$\frac{x_0}{a^2}(x-x_0)+\frac{y_0}{b^2}(y-y_0)+\frac{z_0}{c^2}(z-z_0)=0$$
也就是
$$\frac{x_0x}{a^2}+\frac{y_0y}{b^2}+\frac{z_0z}{c^2}=1$$
$(1)$ 若在点$M$处的切线$l$平行于$\alpha$，则切平面的法向量$\mathbf{n}\perp\alpha$，也就是
$$\frac{lx_0}{a^2}+\frac{my_0}{b^2}+\frac{nz_0}{c^2}=0$$
这说明该点位于过原点的这个平面上
$$\frac{l}{a^2}x+\frac{m}{b^2}y+\frac{n}{c^2}z=0$$

$(2)$ 记号同上，此时切平面也经过点$P(u, v, w)$，因此
$$\frac{u}{a^2}x_0+\frac{v}{b^2}y_0+\frac{w}{c^2}z_0=1$$
因此切点都位于这个平面上
$$\frac{u}{a^2}x+\frac{v}{b^2}y+\frac{w}{c^2}z=1$$

> 11. 设$F(u, v)$可微，$l, m, n$不同时为零。证明：曲面$F(lx-mz, ly-nz)=0$是柱面。

**提示：** 需要证明在曲面上任意点处都存在完全位于该曲面的直线，且它们都平行于某非零向量。

**解答：** 由柱面的定义，只需证明存在一个非零向量$\alpha=(a, b, c)$，使得在曲面上任一点处都有完全位于曲面上的直线，且它们都平行于$\alpha$。由于直线完全位于曲面上，而直线的切线就是自身，那么直线就位于曲面的切平面上。设$G(x, y, z)=F(lx-mz, ly-nz)$，则在曲面上任一点处的法向量为$\mathbf{N}=(G_x, G_y, G_z)=(lF_1, lF_2, -mF_1-nF_2)$。此时应有$\alpha\cdot\mathbf{N}=0$，也就是
$$alF_1+blF_2-c(mF_1+nF_2)=0$$
我们直接取$\alpha=(m, n, l)$就能满足上式，从而曲面上任一点处的法向量都垂直于$\alpha$。设曲面上任一点$M(x_0, y_0, z_0)$，则$F(lx_0-mz_0, ly_0-nz_0)$，以$\alpha$为方向向量且过点$M$的直线为$L:x=x_0+mt, y=y_0+nt, z=z_0+lt$，代入曲面方程可得
$$\begin{aligned}
    F(lx-mz, ly-nz)&=F(lx_0+lmt-mz_0-lmt, ly_0+lnt-nz_0-lnt) \\
    &=F(lx_0-mz_0, ly_0-nz_0) \\
    &=0
\end{aligned}$$
由此可知$L$在曲面上。

### 多元函数的极值与极值

> 12. 讨论函数$f(x, y)=(y-x^2)(y-2x^2)$的极值。

**思路：** 没有定义域限制，直接开始求偏导数。

**解答：** 计算可知$f(x, y)=y^2-3x^2y+2x^4$，求偏导数可知
$$\begin{aligned}
    \frac{\partial f}{\partial x}&=8x^3-3xy=x(8x^2-3y) \\
    \frac{\partial f}{\partial y}&=2y-3x^2
\end{aligned}$$
由极值的必要条件，令两个偏导数为0，则只可能是$x=y=0$，即$(0, 0)$是函数的驻点(不一定是极值点！)。求二阶偏导数可得
$$\left.\frac{\partial^2 z}{\partial x^2}\right|_{(0, 0)}=24(0)^2-3(0)=0$$
这说明$(0, 0)$也不是极值点，因此$f(x, y)$无极值。

> 13. 求函数$z=x^4+y^4-x^2-2xy-y^2$的极值。

**思路：** 和上题完全相同。

**解答：** 由极值的必要条件，求两个偏导数可得
$$\begin{aligned}
    \frac{\partial z}{\partial x}&=4x^3-2x-2y=0 \\
    \frac{\partial z}{\partial y}&=4y^3-2x-2y=0
\end{aligned}$$
两式相减可得$4(x^3-y^3)=0$，这意味着$x=y$。代入上式可得
$$\begin{aligned}
    4(x^3-x)&=0 \\
    x(x+1)(x-1)&=0
\end{aligned}$$
即函数的驻点为$(0, 0), (1, 1), (-1, 1)$。求二阶偏导数可得
$$\begin{aligned}
    A&=\frac{\partial^2 z}{\partial x^2}=12x^2-2 \\
    B&=\frac{\partial^2 z}{\partial x\partial y}=-2 \\
    C&=\frac{\partial^2 z}{\partial y^2}=12y^2-2
\end{aligned}$$
在点$(0, 0)$处，$\Delta=AC-B^2=0$，因此其不是极值点。在点$(1, 1)$处，$\Delta>0$且$A>0$，因此在该点函数有极小值$z_{\min}=-2$。同理在点$(-1, -1)$处函数有极小值$z_{\min}=-2$。

> 14. 求由方程$x^2+2xy+y^2+2x+y=0$所确定的隐函数$y=y(x)$的极值。

**思路：** 一元函数吗？那就更简单了。

**解答：** 方程对$x$求导可得
$$\begin{aligned}
    2x+2y+2xy'+2yy'+2+y'&=0 \\
    (2x+2y+1)y'&=-(2x+2y+2) \\
    y'&=-\frac{2(x+y+1)}{2x+2y+1}
\end{aligned}$$
由极值的必要条件，我们有$x+y+1=0$，即$x=-(y+1)$，代入原方程可得
$$\begin{aligned}
    [-(y+1)]^2-2y(y+1)+y^2-2(y+1)+y&=0 \\
    y^2+2y+1-2y^2-2y+y^2-2y-2+y&=0 \\
    -y-1&=0 \\
    y&=-1
\end{aligned}$$
此时对应$x=0$。要知道这是极大值还是极小值，只需求二阶导数
$$\begin{aligned}
    y''&=-\frac{(2+2y')(2x+2y+1)-2(x+y+1)(2+2y')}{(2x+2y+1)^2} \\
    &=\frac{2+2y'}{(2x+2y+1)^2}
\end{aligned}$$
因此，由$y'(0)=0$，可知$y''(0)=2>0$，因此隐函数$y=y(x)$有极小值$y_{\min}=y(0)=-1$。

> 15. 设$f(x, y)$有二阶连续偏导数，$g(x, y)=f(\mathrm{e}^{xy}, x^2+y^2)$，且
> $$f(x, y)=1-x-y+o(\sqrt{(x-1)^2+y^2}), x\to 1, y\to 0$$
> 证明：$g(x, y)$在$(0, 0)$取得极值，判断此极值是极大值还是极小值，并求出此极值。

**思路：** 题目给出了二阶连续偏导数的条件以及一个看起来像是可微定义的式子，或许可以从此入手。

**解答：** 可将题中式子改写为
$$f(x, y)=-(x-1)-y+o(\sqrt{(x-1)^2+y^2}), x\to 1, y\to 0$$
由
$$\lim_{(x, y)\to(1, 0)}\left(f(x, y)-(1-x-y)\right)=0$$
可推出$f(1, 0)=0$，因此
$$f(x, y)-f(1, 0)=-(x-1)-y+o(\sqrt{(x-1)^2+y^2})$$
$f(x, y)$在点$(1, 0)$处可微，且
$$f_1(1, 0)=-1, f_2(1, 0)=-1$$
对$g(x, y)$求两个偏导数可得
$$\begin{aligned}
    g_x&=f_1\cdot y\mathrm{e}^{xy}+f_2\cdot2x \\
    g_y&=f_1\cdot x\mathrm{e}^{xy}+f_2\cdot2y
\end{aligned}$$
由此可知$g_x(0, 0)=0, g_y(0,0)=0$，因此$(0, 0)$为函数的驻点。计算二阶偏导数可得
$$\begin{aligned}
    g_{xx}&=(f_{11}\cdot y\mathrm{e}^{xy}+f_{12}\cdot 2x)y\mathrm{e}^{xy}+f_1\cdot y^2\mathrm{e}^{xy}+(f_{21}\cdot y\mathrm{e}^{xy}+f_{22}\cdot 2x)2x+2f_2 \\
    g_{xy}&=(f_{11}\cdot x\mathrm{e}^{xy}+f_{12}\cdot 2y)y\mathrm{e}^{xy}+f_1\cdot(\mathrm{e}^{xy}+xy\mathrm{e}^{xy})+(f_{21}\cdot x\mathrm{e}^{xy}+f_{22}\cdot 2y)2x \\
    g_{yy}&=(f_{11}\cdot x\mathrm{e}^{xy}+f_{12}\cdot 2y)x\mathrm{e}^{xy}+f_1\cdot x^2\mathrm{e}^{xy}+(f_{21}\cdot x\mathrm{e}^{xy}+f_{22}\cdot 2y)2y+2f_2
\end{aligned}$$
简单计算可知$A=g_{xx}(0, 0)=2f_2(1, 0)=-2, B=g_{xy}(0, 0)=f_1(0, 0)=-1, C=f_{yy}(0, 0)=2f_2(1, 0)=-2$，因此$\Delta=AC-B^2=3>0$且$A<0$，此为极大值。此时$g(0, 0)=f(1, 0)=0$。

> 16. 求函数$u=x-2y+2z$，在约束条件$x^2+y^2+z^2=1$下的极值。

**思路：** 看到有约束条件，直接拉格朗日乘数法。

**解答：** 构造拉格朗日函数$F=x-2y+2z+\lambda(x^2+y^2+z^2-1)$，只需解方程组
$$\begin{cases}
    F_x=1+2\lambda x=0 \\
    F_y=-2+2\lambda y=0 \\
    F_z=2+2\lambda z=0 \\
    x^2+y^2+z^2=1
\end{cases}$$
第二、三式相加可知$2\lambda(y+z)=0$，但由第一式可知$\lambda\neq 0$，因此有$y=-z$。第三式减第二式可得
$$\begin{aligned}
    2+\lambda(z-y)&=0 \\
    2\lambda z&=-2 \\
    z&=-\frac{1}{\lambda}
\end{aligned}$$
从而$y=\frac{1}{\lambda}$，又从第一式可得$x=-\frac{1}{2\lambda}$。将其代入第四式可得
$$\begin{aligned}
    \frac{1}{4\lambda^2}+\frac{1}{\lambda^2}+\frac{1}{\lambda^2}&=1 \\
    4\lambda^2&=9 \\
    \lambda&=\pm\frac{3}{2}
\end{aligned}$$
若$\lambda=\frac{3}{2}$，则得到点$(-\frac{1}{3}, \frac{2}{3}, -\frac{2}{3})$，此时$u(-\frac{1}{3}, \frac{2}{3}, -\frac{2}{3})=-3$。若$\lambda=-\frac{3}{2}$，则得到点$(\frac{1}{3}, -\frac{2}{3}, \frac{2}{3})$，此时$u(\frac{1}{3}, -\frac{2}{3}, \frac{2}{3})=3$。因此$u$在约束条件下的极小值为$-3$，极大值为$3$。

> 17. 求函数$u=xyz$在约束条件$x^2+y^2+z^2=1$及$x+y+z=0$下的极值。

**思路：** 同样是利用拉格朗日乘数法，但计算较为复杂。

**解答：** 构造拉格朗日函数$F=xyz+\lambda(x^2+y^2+z^2-1)+\mu(x+y+z)$，只需解方程组
$$\begin{cases}
    F_x=yz+2\lambda x+\mu=0 \\
    F_y=xz+2\lambda y+\mu=0 \\
    F_z=xy+2\lambda z+\mu=0 \\
    x^2+y^2+z^2=1 \\
    x+y+z=0
\end{cases}$$
第一式减第二式可得
$$\begin{aligned}
    z(y-x)+2\lambda(x-y)&=0 \\
    (x-y)(2\lambda-z)&=0
\end{aligned}$$
同理可得$(y-z)(2\lambda-x)=0, (x-z)(2\lambda-y)=0$。

1. 若$x\neq y$，则$z=2\lambda$，从而
$$\begin{aligned}
    (y-2\lambda)(2\lambda-x)&=0 \\
    2\lambda(x+y)-4\lambda^2-xy&=0 \\
    xy&=-8\lambda^2 \\
    y&=-\frac{8\lambda^2}{x}
\end{aligned}$$
其中显然$x\neq 0$，代入第五式可知
$$\begin{aligned}
    x-\frac{8\lambda^2}{x}+2\lambda&=0 \\
    x^2+2\lambda x-8\lambda^2&=0 \\
    (x+2\lambda)(x-4\lambda)&=0
\end{aligned}$$
若$x=-2\lambda$，则$y=4\lambda$。代入第四式可知
$$\begin{aligned}
    (-2\lambda)^2+(4\lambda)^2+(2\lambda)^2&=1 \\
    24\lambda^2&=1 \\
    \lambda&=\pm\frac{1}{2\sqrt{6}}
\end{aligned}$$
此时得到两个驻点$(\frac{1}{\sqrt{6}}, -\frac{2}{\sqrt{6}}, -\frac{1}{\sqrt{6}}), (-\frac{1}{\sqrt{6}}, \frac{2}{\sqrt{6}}, \frac{1}{\sqrt{6}})$。若$x=4\lambda$，则$y=-2\lambda$，经过相同的计算可得两个驻点$(\frac{2}{\sqrt{6}}, -\frac{1}{\sqrt{6}}, \frac{1}{\sqrt{6}}), (-\frac{2}{\sqrt{6}}, \frac{1}{\sqrt{6}}, -\frac{1}{\sqrt{6}})$。

2. 若$x=y$，则由第五式易知$z=-2x$，代入第四式可知
$$\begin{aligned}
    x^2+x^2+(-2x)^2&=1 \\
    x&=\pm\frac{1}{\sqrt{6}}
\end{aligned}$$
从而可得两个驻点$(\frac{1}{\sqrt{6}}, \frac{1}{\sqrt{6}}, -\frac{2}{\sqrt{6}}), (-\frac{1}{\sqrt{6}}, -\frac{1}{\sqrt{6}}, \frac{2}{\sqrt{6}})$。

分别计算函数在这六个驻点处的值，可得函数的极大值为$u(\frac{1}{\sqrt{6}}, -\frac{2}{\sqrt{6}}, -\frac{1}{\sqrt{6}})=u(-\frac{2}{\sqrt{6}}, \frac{1}{\sqrt{6}}, -\frac{1}{\sqrt{6}})=u(-\frac{1}{\sqrt{6}}, -\frac{1}{\sqrt{6}}, \frac{2}{\sqrt{6}})=\frac{1}{3\sqrt{6}}$，极小值为$u(-\frac{1}{\sqrt{6}}, \frac{2}{\sqrt{6}}, \frac{1}{\sqrt{6}})=u(\frac{2}{\sqrt{6}}, -\frac{1}{\sqrt{6}}, \frac{1}{\sqrt{6}})=u(\frac{1}{\sqrt{6}}, \frac{1}{\sqrt{6}}, -\frac{2}{\sqrt{6}})=-\frac{1}{3\sqrt{6}}$。

> 18. 求函数$z=x^2+y^2$在约束条件$\frac{x}{a}+\frac{y}{b}=1$下的极值。

**思路：** 约束条件不是有界点集，因此无法由最值方法得知这是极大值还是极小值，不如转化为一元函数来求极值。

**解答：** 从条件可得
$$\begin{aligned}
    \frac{y}{b}&=1-\frac{x}{a} \\
    y&=b-\frac{b}{a}x
\end{aligned}$$
只需求一元函数$z(x)=x^2+(b-\frac{b}{a}x)^2$的极值。求导可得
$$\begin{aligned}
    z'&=2x-2\cdot\frac{b}{a}(b-\frac{b}{a}x) \\
    &=2x+\frac{2b^2}{a^2}x-\frac{2b^2}{a} \\
    &=2\left[\left(1+\frac{b^2}{a^2}\right)x-\frac{b^2}{a}\right]
\end{aligned}$$
因此$z'(\frac{ab^2}{a^2+b^2})=0$。此时有驻点$(\frac{ab^2}{a^2+b^2}, \frac{a^2b}{a^2+b^2})$，求二阶导数可得
$$z''=2+\frac{2b^2}{a^2}>0$$
因此这是极小值，且
$$\begin{aligned}
    z_{\min}&=\left(\frac{ab^2}{a^2+b^2}\right)^2+\left(\frac{a^2b}{a^2+b^2}\right)^2 \\
    &=\frac{a^2b^2(a^2+b^2)}{(a^2+b^2)^2} \\
    &=\frac{a^2b^2}{a^2+b^2}
\end{aligned}$$

> 19. 求函数$z=x^2+xy+y^2+x-y+1$在由$y=x+2, x=0, y=0$所围的有界闭区域$\bar{D}$上的最值。

**思路：** 闭区域上的最值可分为区域内部和边界来讨论。

**解答：** 在闭区域内部$D$，只需求两个偏导数并令其等于零
$$\begin{cases}
    \frac{\partial z}{\partial x}=2x+y+1=0 \\
    \frac{\partial z}{\partial y}=x+2y-1=0
\end{cases}$$
容易解得$x=-1, y=1$，但此点在区域边界，且$z(-1, 1)=0$，因此在$D$内函数取不到最值。接下来讨论在边界$\partial D$上函数的取值。首先，若$x=0$，则$y\in[0, 2], z(y)=y^2-y+1$，在这情况下的最小值为$z(0, \frac{1}{2})=\frac{3}{4}$，最大值为$z(0, 2)=3$。然后，若$y=0$，则$x\in[-2, 0], z(x)=x^2+x+1$，在这情况下的最小值为$z(-\frac{1}{2}, 0)=\frac{3}{4}$，最大值为$z(-2, 0)=3$。最后，$y=x+2$这段可由参数方程表示
$$\begin{cases}
    x=t \\
    y=t+2
\end{cases}, t\in[-2, 0]$$
此时
$$\begin{aligned}
    z(t)&=t^2+t(t+2)+(t+2)^2+t-(t+2)+1 \\
    &=t^2+t^2+2t+t^2+4t+4+t-t-2+1 \\
    &=3t^2+6t+3 \\
    &=3(t+1)^2
\end{aligned}$$
在这情况下的最小值就是$z(-1, 1)=0$，最大值为$z(-2, 0)=z(0, 2)=3$。最后得到在有界闭区域$\bar{D}$上的最小值为$z(-1, 1)=0$，最大值为$z(-2, 0)=z(0, 2)=3$。

> 20. 求$u=x+2y+3z$在$x^2+y^2+z^2\le 1$上的最值。

**思路：** 与上题相同，都是分区域内部和边界上两种情况来讨论。

**解答：** 在区域$x^2+y^2+z^2<1$内，计算偏导数可知
$$\begin{cases}
    u_x=1 \\
    u_y=2 \\
    u_z=3
\end{cases}$$
因此函数没有驻点。现在讨论边界$x^2+y^2+z^2=1$上函数的取值，构造拉格朗日函数$F=x+2y+3z+\lambda(x^2+y^2+z^2-1)$，只需解方程组
$$\begin{cases}
    F_x=1+2\lambda x=0 \\
    F_y=2+2\lambda y=0 \\
    F_z=3+2\lambda z=0 \\
    x^2+y^2+z^2=1
\end{cases}$$
在前三式中将$x, y, z$用$\lambda$表示后代入第四式可得
$$\begin{aligned}
    \left(-\frac{1}{2\lambda}\right)^2+\left(-\frac{2}{2\lambda}\right)^2+\left(-\frac{3}{2\lambda}\right)^2&=1 \\
    4\lambda^2&=14 \\
    \lambda&=\pm\frac{\sqrt{14}}{2}
\end{aligned}$$
由此可得最小值$u(-\frac{1}{\sqrt{14}}, -\frac{2}{\sqrt{14}}, -\frac{3}{\sqrt{14}})=-\sqrt{14}$，最大值$u(\frac{1}{\sqrt{14}}, \frac{2}{\sqrt{14}}, \frac{3}{\sqrt{14}})=\sqrt{14}$。

> 21. 设$z=f(x, y)=(x^2-y^2)\mathrm{e}^{-x^2-y^2}$，讨论$f(x, y)$在定义域$\R^2$上的最值。

**思路：** 直接开始求偏导数，但重点在于需要论证求得的极值就是最值

**解答：** 求两个偏导数可得
$$\begin{aligned}
    z_x&=2x\mathrm{e}^{-x^2-y^2}+(x^2-y^2)\cdot(-2x)\mathrm{e}^{-x^2-y^2} \\
    &=2\mathrm{e}^{-x^2-y^2}(x-x^3+xy^2) \\
    z_y&=-2y\mathrm{e}^{-x^2-y^2}+(x^2-y^2)\cdot(-2y)\mathrm{e}^{-x^2-y^2} \\
    &=-2\mathrm{e}^{-x^2-y^2}(y+yx^2-y^3)
\end{aligned}$$
两式相加可得
$$\begin{aligned}
    x^3+y^3-xy^2-x^2y-x-y&=(x+y)(x^2-xy+y^2)-xy(x+y)-(x+y) \\
    &=(x+y)(x^2-2xy+y^2-1) \\
    &=(x+y)[(x-y)^2-1]
\end{aligned}$$
令其等于0，那么就只有三种情况。

1. 当$x=-y$，则只能是$(x, y)=(0, 0)$，则$f(0, 0)=0$。

2. 当$x-y=1$，则
$$\begin{aligned}
    z&=[x^2-(x-1)^2]\mathrm{e}^{-x^2-(x-1)^2} \\
    &=(2x-1)\mathrm{e}^{-2x^2+2x-1}
\end{aligned}$$
此时求导可得
$$\begin{aligned}
    z'&=2\mathrm{e}^{-2x^2+2x-1}+(2x-1)(-4x+2)\mathrm{e}^{-2x^2+2x-1} \\
    &=\mathrm{e}^{-2x^2+2x-1}(2-8x^2+8x-2) \\
    &=-8x\mathrm{e}^{-2x^2+2x-1}(x-1)
\end{aligned}$$
那么有$f(0, -1)=-\mathrm{e}^{-1}, f(1, 0)=\mathrm{e}^{-1}$。

3. 当$x-y=-1$，则
$$\begin{aligned}
    z&=[x^2-(x+1)^2]\mathrm{e}^{-x^2-(x+1)^2} \\
    &=(-2x-1)\mathrm{e}^{-2x^2-2x-1}
\end{aligned}$$
此时求导可得
$$\begin{aligned}
    z'&=-2\mathrm{e}^{-2x^2-2x-1}+(-2x-1)\cdot(-4x-2)\mathrm{e}^{-2x^2-2x-1} \\
    &=\mathrm{e}^{-2x^2-2x-1}(-2+8x^2+8x+2) \\
    &=8x\mathrm{e}^{-2x^2-2x-1}(x+1)
\end{aligned}$$
那么有$f(0, 1)=-\mathrm{e}^{-1}, f(-1, 0)=\mathrm{e}^{-1}$。

现在讨论以上的极值必是最值。令$x=r\cos\theta, y=r\sin\theta$，则$f(x, y)=(r^2\cos^2\theta-r^2\sin^2\theta)\mathrm{e}^{-r^2}=r^2\cos 2\theta\mathrm{e}^{-r^2}$，因此$0\le|f(x, y)|\le r^2\mathrm{e}^{-r^2}\to 0(r\to +\infty)$。根据极限的保号性，取足够大的$R$，使得在区域$\bar{D}:x^2+y^2\le R$内包含了函数的所有极值点，而在区域外有$-\mathrm{e}^{-1}<f(x, y)<\mathrm{e}^{-1}$。那么，最小值就是$f(0, \pm 1)=-\mathrm{e}^{-1}$，最大值就是$f(\pm 1, 0)=\mathrm{e}^{-1}$。

> 22. 求函数$z=x^3+y^3-3xy$的极值，并讨论函数的最值。

**思路：** 还是直接开始求偏导数，找到驻点。但明显可以看出函数是没有极值的。

**解答：** 求极值只需令两个偏导数为零，也就是解方程组
$$\begin{cases}
    z_x=3x^2-3y=0 \\
    z_y=3y^2-3x=0
\end{cases}$$
这样可得到两个驻点$(0, 0)$和$(1, 1)$。求二阶偏导数可知
$$\begin{aligned}
    z_{xx}&=6x \\
    z_{xy}&=-3 \\
    z_{yy}&=6y
\end{aligned}$$
在点$(0, 0)$，计算可得$A=C=0, B=-3, \Delta=AC-B^2=-9<0$，因此这不是极值点。在点$(1, 1)$，计算可得$A=C=6, B=-3, \Delta=AC-B^2=27>0$，因此这是极小值点，且$z(1, 1)=-1$。但只要令$y=0, x\to+\infty$，就可知$z\to+\infty$，同理令$x\to-\infty$，就可知$z\to-\infty$，因此函数无最值。

**总结：** 本题实际说明，在一元函数成立的“唯一极值即最值”，在多元函数中不再成立。

> 23. 求函数$z=x^2-xy+y^2$在有界闭区域$D:|x|+|y|\le 1$上的最大值与最小值。

**思路：** 还是分成区域内部和边界讨论。

**解答：** 在区域内部$|x|+|y|<1$，只需求两个偏导数并令其等于零
$$\begin{cases}
    z_x=2x-y=0 \\
    z_y=2y-x=0
\end{cases}$$
这样可得到驻点$(0, 0)$。计算二阶偏导数可知$A=z_{xx}=2>0, B=z_{xy}=-1, C=z_{yy}=2, \Delta=AC-B^2=3>0$，因此在区域内部$z_{\min}=0$。现在考虑在边界$|x|+|y|=1$上函数的取值，这是由$x+y=\pm 1, x-y=\pm 1$围成的正方形，在$x+y=1(x\in[0, 1])$这段线段上，$z_{\min}=z(\frac{1}{2}, \frac{1}{2})=\frac{1}{4}, z_{\max}=z(1, 0)=z(0, 1)=1$。其余线段可作类似计算。最终得到$z=x^2-xy+y^2$在$D$上的最大值为$z(\pm 1, 0)=z(0, \pm 1)=1$，最小值为$z(0, 0)=0$。

### 多元函数微分学的综合题

> 24. 证明空间曲面的最远(近)点的垂线原理：
>
> $(1)$ 在空间中，如果$\Sigma$是光滑闭曲面，点$Q$不在$\Sigma$上，点$\bar{P}$是$\Sigma$上与$Q$的最远(近)点，则直线$\bar{P}Q$在点$\bar{P}$与$\Sigma$垂直，即$\bar{P}Q$是$\Sigma$在点$\bar{P}$处的法线。
>
> $(2)$ 设在空间中光滑闭曲面$\Sigma_1, \Sigma_2$不相交，点$P_1, P_2$分别是$\Sigma_1, \Sigma_2$之间的最远(近)点，则直线$P_1P_2$是$\Sigma_1, \Sigma_2$的公垂线，则$P_1P_2$既垂直于$\Sigma_1$又垂直于$\Sigma_2$。

**提示：** 可建立$Q$到$\Sigma$上某点$P$的距离函数，用拉格朗日乘数法求最值。

**解答：** $(1)$ 设$Q=(x_0, y_0, z_0)\notin\Sigma, P=(x, y, z)\in\Sigma$是$\Sigma$上与$Q$的最远(近)点，光滑闭曲面$\Sigma$的参数方程为$\Sigma:F(x, y, z)=0$，则$P, Q$之间的距离平方为$\rho^2=(x-x_0)^2+(y-y_0)^2+(z-z_0)^2$。此时，只需求$\rho^2$的最大(小)值。为此，构造拉格朗日函数$L=(x-x_0)^2+(y-y_0)^2+(z-z_0)^2+\lambda F(x, y, z)$，则只需解方程组
$$\begin{cases}
    L_x=2x-2x_0+\lambda F_x=0 \\
    L_y=2y-2y_0+\lambda F_y=0 \\
    L_z=2z-2z_0+\lambda F_z=0 \\
    F(x, y, z)=0
\end{cases}$$
若$\lambda=0$，则必有$(x_0, y_0, z_0)=(x, y, z)\in\Sigma$，这就与$Q\notin\Sigma$矛盾。故$\lambda\neq 0$，那么就有
$$\begin{aligned}
    \frac{x-x_0}{y-y_0}&=\frac{F_x}{F_y} \\
    \frac{x-x_0}{z-z_0}&=\frac{F_x}{F_z} \\
    \frac{y-y_0}{z-z_0}&=\frac{F_y}{F_z}
\end{aligned}$$
从而
$$\frac{x-x_0}{F_x}=\frac{y-y_0}{F_y}=\frac{z-z_0}{F_z}$$
这是一条过$P, Q$的直线，其方向向量为$(F_x, F_y, F_z)$，恰为$\Sigma$在点$P$处的法向量，因此$\bar{P}Q$是$\Sigma$在点$\bar{P}$处的法线。

$(2)$ 因为$\Sigma_1, \Sigma_2$不相交，$P_1$是$P_2$到$\Sigma_1$的最远(近)点，因此由上，$P_1P_2$在点$P_1$与$\Sigma_1$垂直。同理，$P_1P_2$在点$P_2$与$\Sigma_2$垂直。于是$P_1P_2$是$\Sigma_1, \Sigma_2$的公垂线。
