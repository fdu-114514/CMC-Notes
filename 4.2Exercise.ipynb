## 4.2的练习

### 多元函数的极限与连续

> 1. 求下列极限
> 
> $$(1)\lim_{(x, y)\to(2, -\frac{1}{2})}(2+xy)^{\frac{1}{y+xy^2}}$$
> 
> $$(2)\lim_{(x, y)\to(+\infty, +\infty)}(x^2+y^2)\mathrm{e}^{-(x+y)}$$

**思路：** 第一小题明显与$\mathrm{e}$有关，第二小题直觉可知是0，但如何严格证明呢？

**解答：** $(1)$ 直接计算可得
$$\begin{aligned}
    \lim_{(x, y)\to(2, -\frac{1}{2})}(2+xy)^{\frac{1}{y+xy^2}}&=\lim_{(x, y)\to(2, -\frac{1}{2})}\left\{[1+(xy+1)]^{\frac{1}{xy+1}}\right\}^{\frac{1}{y}} \\
    &=\mathrm{e}^{-2}
\end{aligned}$$

$(2)$ 对式子进行变形可得
$$\lim_{(x, y)\to(+\infty, +\infty)}(x^2+y^2)\mathrm{e}^{-(x+y)}=\lim_{(x, y)\to(+\infty, +\infty)}\left(\frac{x^2}{\mathrm{e}^x}\mathrm{e}^{-y}+\frac{y^2}{\mathrm{e}^y}\mathrm{e}^{-x}\right)$$
容易知道
$$\lim_{(x, y)\to(+\infty, +\infty)}\frac{x^2}{\mathrm{e}^x}\mathrm{e}^y=\lim_{x\to+\infty}\frac{x^2}{\mathrm{e}^x}\cdot\lim_{y\to+\infty}\mathrm{e}^{-y}=0\times 0=0$$
同理
$$\lim_{(x,y)\to(+\infty, +\infty)}\frac{y^2}{\text{e}^y}\text{e}^{-x}=0$$
因此
$$\lim_{(x, y)\to(+\infty, +\infty)}(x^2+y^2)\mathrm{e}^{-(x+y)}=0$$

> 2. 设
> 
> $$(1)f(x, y)=\begin{cases}
    \frac{x^2y^2}{x^2y^2+(x-y)^2}, &(x, y)\neq (0,0) \\
    0, &(x, y)=(0, 0)
\end{cases}$$
> $$(2)g(x, y)=\begin{cases}
    \frac{x^4y^4}{(x^2+y^4)^3}, &(x, y)\neq (0,0) \\
    0, &(x, y)=(0, 0)
\end{cases}$$
> 证明：这两个函数在原点不连续。

**思路：** 只要证明函数趋于原点时的极限不存在就可以了。

**解答：** $(1)$ 取$x=ky$可得
$$\begin{aligned}
    \lim_{(x, y)\to (0, 0)}f(x, y)&=\lim_{y\to 0}\frac{k^2y^4}{k^2y^4+(k-1)^2y^2} \\
    &=\lim_{y\to 0}\frac{k^2y^2}{(k-1)^2+k^2y^2}
\end{aligned}$$
显然在$k=1$时此极限为1，否则为0，因此$\lim_{(x, y)\to (0, 0)}f(x, y)$不存在，函数在原点不连续。

$(2)$ 取$x=y^2$可得
$$\begin{aligned}
    \lim_{(x, y)\to(0, 0)}g(x, y)&=\lim_{y\to 0}\frac{y^{12}}{8y^{12}} \\
    &=\frac{1}{8}
\end{aligned}$$
但若取$x=y$就有
$$\begin{aligned}
    \lim_{(x, y)\to(0, 0)}g(x, y)&=\lim_{y\to 0}\frac{y^8}{(y^4+y^2)^3} \\
    &=\lim_{y\to 0}\frac{y^8}{y^{12}+3y^{10}+3y^8+y^6} \\
    &=\lim_{y\to 0}\frac{y^2}{1+3y^2+3y^4+y^6} \\
    &=0
\end{aligned}$$
因此$\lim_{(x, y)\to (0, 0)}g(x, y)$不存在，函数在原点不连续。

### 偏导数

> 3. 求下列函数的偏导数：
> $(1)u=(\frac{x}{y})^z$；
> 
> $(2)u=x^{y^z}$；
> 
> $(3)z=(3x+2y-1)^{2x+3y}$；
> 
> $(4)z=\arctan\frac{x^2}{y}$

**思路：** 评价为基础题，敢算就行。

**解答：** $(1)$ 直接计算可得
$$\begin{aligned}
    \frac{\partial u}{\partial x}&=\frac{z}{y}\left(\frac{x}{y}\right)^{z-1} \\
    \frac{\partial u}{\partial y}&=-\frac{xz}{y^2}\left(\frac{x}{y}\right)^{z-1} \\
    \frac{\partial u}{\partial z}&=\left(\frac{x}{y}\right)^z\ln\frac{x}{y}
\end{aligned}$$

$(2)$ 由$\ln u=y^z\ln x$可知
$$\begin{aligned}
    \frac{\partial u}{\partial x}&=y^z\cdot x^{y^z-1} \\
    \frac{\partial u}{\partial y}&=x^{y^z}\cdot z\ln x\cdot y^{z-1} \\
    \frac{\partial u}{\partial z}&=x^{y^z}\cdot \ln x\cdot y^z\ln y
\end{aligned}$$

$(3)$ 由$\ln z=(2x+3y)\ln(3x+2y-1)$可知
$$\begin{aligned}
    \frac{\partial z}{\partial x}&=(3x+2y-1)^{2x+3y}\cdot\left[2\ln(3x+2y-1)+\frac{3(2x+3y)}{3x+2y-1}\right] \\
    \frac{\partial z}{\partial y}&=(3x+2y-1)^{2x+3y}\cdot\left[3\ln(3x+2y-1)+\frac{2(2x+3y)}{3x+2y-1}\right]
\end{aligned}$$

$(4)$ 直接计算可得
$$\begin{aligned}
    \frac{\partial z}{\partial x}&=\frac{1}{1+(\frac{x^2}{y})^2}\cdot\frac{2x}{y} \\
    &=\frac{y^2}{x^4+y^2}\cdot\frac{2x}{y} \\
    &=\frac{2xy}{x^4+y^2} \\
    \frac{\partial z}{\partial y}&=\frac{1}{1+(\frac{x^2}{y})^2}\cdot\left(-\frac{x^2}{y^2}\right) \\
    &=-\frac{y^2}{x^4+y^2}\cdot\frac{x^2}{y^2} \\
    &=-\frac{x^2}{x^4+y^2}
\end{aligned}$$

> 4. 设$z=f(x, y)=\frac{x\cos(y-1)-(y-1)\cos x}{1+\sin x+\sin(y-1)}$，求$\left.\frac{\partial z}{\partial x}\right|_{(0, 1)}, \left.\frac{\partial z}{\partial y}\right|_{(0, 1)}$。

**思路：** 函数形式如此复杂，求偏导函数显然不是一个好的做法。应该使用偏导数的定义。

**解答：** 由偏导数的定义
$$\begin{aligned}
    \left.\frac{\partial z}{\partial x}\right|_{(0, 1)}&=\left.\frac{\mathrm{d}}{\mathrm{d}x}f(x, 1)\right|_{x=0} \\
    &=\left.\frac{\mathrm{d}}{\mathrm{d}x}\left(\frac{x}{1+\sin x}\right)\right|_{x=0} \\
    &=\left.\frac{1+\sin x-x\cos x}{(1+\sin x)^2}\right|_{x=0} \\
    &=1 \\
    \left.\frac{\partial z}{\partial y}\right|_{(0, 1)}&=\left.\frac{\mathrm{d}}{\mathrm{d}y}f(0, y)\right|_{y=1} \\
    &=\left.\frac{\mathrm{d}}{\mathrm{d}y}\left[\frac{1-y}{1+\sin(y-1)}\right]\right|_{y=1} \\
    &=\left.\frac{-[1+\sin(y-1)]-(1-y)\cos(y-1)}{[1+\sin(y-1)]^2}\right|_{y=1} \\
    &=-1
\end{aligned}$$

> 5. 设$w=f(u, v)$可微，求复合函数$w=f(x^2-y^2, \mathrm{e}^{xy})$的偏导数。

**思路：** 考验你的链式法则功底！

**解答：** 由多元复合函数求导的链式法则
$$\begin{aligned}
    \frac{\partial w}{\partial x}&=f'_1\cdot\frac{\partial u}{\partial x}+f'_2\cdot\frac{\partial v}{\partial x} \\
    &=2xf'_1+y\mathrm{e}^{xy}f'_2 \\
    \frac{\partial w}{\partial y}&=f'_1\cdot\frac{\partial u}{\partial y}+f'_2\cdot\frac{\partial v}{\partial y} \\
    &=-2yf'_1+x\mathrm{e}^{xy}f'_2
\end{aligned}$$

> 6. 设$u=f(x, y)$为可微函数。
> 
> $(1)$若$u=f(x, y)$满足方程$x\frac{\partial f}{\partial x}+y\frac{\partial f}{\partial y}=0$，试证$f(x, y)$在极坐标系中只是$\theta$的函数，而与$r$无关。
> 
> $(2)$若$u=f(x, y)$满足方程$\frac{1}{x}\cdot\frac{\partial f}{\partial x}-\frac{1}{y}\cdot\frac{\partial f}{\partial y}=0$，试证$f(x, y)$在极坐标系中只是$r$的函数，而与$\theta$无关。

**思路：** 本质上还是对链式法则的考察，只是需要多做点东西。

**解答：** 设$u=f(x, y)=F(r, \theta)$

$(1)$只需证明$\frac{\partial F}{\partial r}=0$，计算可知
$$\begin{aligned}
    \frac{\partial F}{\partial r}&=f'_1\cos\theta+f'_2\sin\theta \\
    &=\frac{r\cos\theta f'_1+r\sin\theta f'_2}{r} \\
    &=0
\end{aligned}$$

$(2)$ 只需证明$\frac{\partial F}{\partial \theta}=0$，计算可知
$$\begin{aligned}
    \frac{\partial F}{\partial \theta}&=f'_1\cdot(-r\sin\theta)+f'_2\cdot r\cos\theta \\
    &=x\frac{\partial f}{\partial y}-y\frac{\partial f}{\partial x} \\
    &=-xy\left(\frac{1}{x}\cdot\frac{\partial f}{\partial x}-\frac{1}{y}\cdot\frac{\partial f}{\partial y}\right) \\
    &=0
\end{aligned}$$

> 7. 设$f(x, y)$有连续的偏导数，且$f(x, x^2)=1, f_x(x, x^2)=x$，求$f_y(x, x^2)$。

**思路：** 直接对$x$求导即可，注意链式法则的运用。

**解答：** 由链式法则可知
$$\begin{aligned}
    \frac{\mathrm{d}}{\mathrm{d}x}f(x, x^2)&=0 \\
    f_x(x, x^2)\cdot1+f_y(x, x^2)\cdot(2x)&=0 \\
    2xf_y(x, x^2)&=-x \\
    f_y(x, x^2)&=-\frac{1}{2}
\end{aligned}$$

> 8. 设$F(x, x+y, x+y+z)=0$，其中$F$可微，求$\frac{\partial z}{\partial x}$和$\frac{\partial z}{\partial y}$。

**思路：** 直接开始求偏导。

**解答：** 方程对$x$求偏导可得
$$\begin{aligned}
    F'_1+F'_2+F'_3(1+\frac{\partial z}{\partial x})&=0 \\
    \frac{\partial z}{\partial x}&=-\frac{F'_1+F'_2+F'_3}{F'_3}
\end{aligned}$$
而对$y$求偏导可得
$$\begin{aligned}
    F'_2+F'_3(1+\frac{\partial z}{\partial y})&=0 \\
    \frac{\partial z}{\partial y}&=-\frac{F'_2+F'_3}{F'_3}
\end{aligned}$$

> 9. 设$F(x, y, z)$可微，$x=x(y, z), y=y(x, z), z=z(x, y)$都是方程$F(x, y, z)=0$所确定的隐函数，证明：$\frac{\partial x}{\partial y}\cdot\frac{\partial y}{\partial z}\cdot\frac{\partial z}{\partial x}=-1$。

**思路：** 考虑使用隐函数求导法。

**解答：** 由隐函数求导法
$$\begin{aligned}
    \frac{\partial x}{\partial y}\cdot\frac{\partial y}{\partial z}\cdot\frac{\partial z}{\partial x}&=-\frac{F_y}{F_x}\cdot\left(-\frac{F_z}{F_y}\right)\cdot\left(-\frac{F_x}{F_z}\right) \\
    &=-1
\end{aligned}$$

**注：** 本题说明偏导数符号$\frac{\partial z}{\partial x}$不能像一元函数的导数$\frac{\mathrm{d}z}{\mathrm{d}x}$一样进行分式运算。

> 10. 由方程$\Phi(cx-az, cy-bz)=0$确定了函数$z=z(x, y)$，求$a\frac{\partial z}{\partial x}+b\frac{\partial z}{\partial y}$。

**思路：** 直接开始求偏导。

**解答：** 原方程对$x$求偏导可得
$$\begin{aligned}
    \Phi'_1(c-a\frac{\partial z}{\partial x})+\Phi'_2(-b\frac{\partial z}{\partial x})&=0 \\
    \frac{\partial z}{\partial x}&=\frac{c\Phi'_1}{a\Phi'_1+b\Phi'_2}
\end{aligned}$$
再对$y$求偏导可得
$$\begin{aligned}
    \Phi'_1(-a\frac{\partial z}{\partial y})+\Phi'_2(c-b\frac{\partial z}{\partial y})&=0 \\
    \frac{\partial z}{\partial y}&=\frac{c\Phi'_2}{a\Phi'_1+b\Phi'_2}
\end{aligned}$$
因此
$$\begin{aligned}
    a\frac{\partial z}{\partial x}+b\frac{\partial z}{\partial y}&=\frac{ac\Phi'_1+bc\Phi'_2}{a\Phi'_1+b\Phi'_2} \\
    &=c
\end{aligned}$$

> 11. 由方程组$\begin{cases}
    x^2+y^2+z^2=6 \\
    x+y+z=0
\end{cases}$确定了隐函数$y=y(x), z=z(x)$。求导数$\frac{\mathrm{d}y}{\mathrm{d}x}, \frac{\mathrm{d}z}{\mathrm{d}x}$。

**思路：** 注意$y$和$z$都仅与$x$有关，不要导错了。

**解答：** 方程组对$x$求导可得
$$\begin{cases}
    2x+2y\frac{\mathrm{d}y}{\mathrm{d}x}+2z\frac{\mathrm{d}z}{\mathrm{d}x}=0 \\
    1+\frac{\mathrm{d}y}{\mathrm{d}x}+\frac{\mathrm{d}z}{\mathrm{d}x}=0
\end{cases}$$
计算可得
$$\begin{aligned}
    \frac{\mathrm{d}y}{\mathrm{d}x}&=\frac{z-x}{y-z} \\
    \frac{\mathrm{d}z}{\mathrm{d}x}&=\frac{x-y}{y-z}
\end{aligned}$$

> 12. 设由方程组$\begin{cases}
    x^2u-yv^2=2 \\
    y^2u^2+x^2v=1
\end{cases}$确定了隐函数$u=u(x, y), v=v(x, y)$。求它们的偏导数。

**思路：** 当题目出现方程组时，求全微分是一种比较方便的方法。

**解答：** 对方程组求全微分，得到
$$\begin{cases}
    (2ux\mathrm{d}x+x^2\mathrm{d}u)-(v^2\mathrm{d}y+2vy\mathrm{d}v)=0 \\
    (2u^2y\mathrm{d}y+2uy^2\mathrm{d}u)+(2vx\mathrm{d}x+x^2\mathrm{d}v)=0
\end{cases}$$
$$\begin{cases}
    x^2\mathrm{d}u-2vy\mathrm{d}v=-2ux\mathrm{d}x+v^2\mathrm{d}y \\
    2uy^2\mathrm{d}u+x^2\mathrm{d}v=-2vx\mathrm{d}x-2u^2y\mathrm{d}y
\end{cases}$$
$$\begin{cases}
    (x^4+4uvy^3)\mathrm{d}u=(-2ux^3-4v^2xy)\mathrm{d}x+(x^2v^2-4u^2vy^2)\mathrm{d}y \\
    (x^4+4uvy^3)\mathrm{d}v=(-2vx^3+4u^2xy^2)\mathrm{d}x+(-2u^2x^2y-2uv^2y^2)\mathrm{d}y
\end{cases}$$
因此可求得
$$\begin{aligned}
    \frac{\partial u}{\partial x}&=-\frac{2ux^3+4v^2xy}{x^4+4uvy^3} \\
    \frac{\partial u}{\partial v}&=\frac{x^2v^2-4u^2vy^2}{x^4+4uvy^3} \\
    \frac{\partial v}{\partial x}&=\frac{4u^2xy^2-2vx^3}{x^4+4uvy^3} \\
    \frac{\partial v}{\partial y}&=-\frac{2u^2x^2y+2uv^2y^2}{x^4+4uvy^3}
\end{aligned}$$

> 13. 设$\mathbf{l}_j(j=1, 2, \cdots, n)$是平面上点$P_0$处点$n\ge 2$个方向向量，相邻两个向量之间的夹角为$2\pi/n$。若函数$f(x, y)$在点$P_0$有连续偏导数，证明$\sum_{j=1}^n\frac{\partial f(P_0)}{\partial\mathbf{l}_j}=0$。

**思路：** 既然有连续偏导数，那么容易表示$\frac{\partial f(P_0)}{\partial\mathbf{l}_j}$，唯一的难点是如何化简了。

**解答：** 设$P_0$的坐标为$(x_0, y_0)$，方向向量$\mathbf{l}_j=(\cos\alpha_j, \cos\beta_j)$。
$$\begin{aligned}
    \sum_{j=1}^n\frac{\partial f(P_0)}{\partial\mathbf{l}_j}&=\sum_{j=1}^n\left[f_x(x_0, y_0)\cos\alpha_j+f_y(x_0, y_0)\cos\beta_j\right] \\
    &=f_x(x_0, y_0)\sum_{j=1}^n\cos\alpha_j+f_y(x_0, y_0)\sum_{j=1}^n\cos\beta_j \\
\end{aligned}$$
其中
$$\begin{aligned}
    \sum_{j=1}^n\cos\alpha_j&=\sum_{j=0}^{n-1}\text{Re}(\mathrm{e}^{\mathrm{i}(\alpha_1+\frac{2\pi j}{n})}) \\
    &=\text{Re}\left(\sum_{j=0}^{n-1}\mathrm{e}^{\mathrm{i}\alpha_1}\cdot\mathrm{e}^{\frac{2\pi j}{n}\mathrm{i}}\right) \\
    &=\text{Re}\left(\mathrm{e}^{\mathrm{i}\alpha_1}\cdot\frac{1-\mathrm{e}^{2\pi\mathrm{i}}}{1-\mathrm{e}^{\frac{2\pi}{n}\mathrm{i}}}\right) \\
    &=\text{Re}(\mathrm{e}^{\mathrm{i}\alpha_1}\cdot 0) \\
    &=0
\end{aligned}$$
同理
$$\sum_{j=1}^n\cos\beta_j=0$$
因此
$$\sum_{j=1}^n\frac{\partial f(P_0)}{\partial\mathbf{l}_j}=0$$

> 14. 问在空间的哪些点上，函数$u=x^3+y^3+z^3-3xyz$的梯度平行于$y$轴。

**思路：** 只需回想梯度的定义即可。

**解答：** 直接计算可知$u$的梯度为
$$\begin{aligned}
    \text{grad}u&=\left(\frac{\partial u}{\partial x}, \frac{\partial u}{\partial y}, \frac{\partial u}{\partial z}\right) \\
    &=(3x^2-3yz, 3y^2-3xz, 3z^2-3xy)
\end{aligned}$$
$y$轴的方向向量为$(0, 1, 0)$，因此只要令$3x^2-3yz=3z^2-3x=0$，且$3y^2-3xz\neq 0$即可。因此，只要是满足下列方程组的点，函数在这些点的梯度都平行于$y$轴：
$$\begin{cases}
    x^2=yz \\
    z^2=xy
\end{cases}, \text{且}y^2\neq xz$$

### 全微分

> 15. 求下列函数的全微分：
>
> $(1)z=x^y$；
> 
> $(2)u=xyz+\sin(xy)$

**思路：** 就硬算。

**解答：** $(1)$计算可知
$$\begin{aligned}
    \mathrm{d}z&=\frac{\partial z}{\partial x}\mathrm{d}x+\frac{\partial z}{\partial y}\mathrm{d}y \\
    &=yx^{y-1}\mathrm{d}x+xy\ln x\mathrm{d}y
\end{aligned}$$

$(2)$计算可知
$$\begin{aligned}
    \mathrm{d}u&=\frac{\partial u}{\partial x}\mathrm{d}x+\frac{\partial u}{\partial y}\mathrm{d}y+\frac{\partial u}{\partial z}\mathrm{d}z \\
    &=[yz+y\cos(xy)]\mathrm{d}x+[xz+x\cos(xy)]\mathrm{d}y+xy\mathrm{d}z
\end{aligned}$$

> 16. 利用微分的演算公式求全微分：
> 
> $(1)z=\arctan\frac{x}{x^2+y^2}$，求$\mathrm{d}z$；
> 
> $(2)u^3-3(x+y)u^2+z^3=0$，求$\mathrm{d}u$。

**思路：** 微分注意计算即可。

**解答：** $(1)$ 令$u=z-\arctan\frac{x}{x^2+y^2}=0$，计算可知
$$\begin{aligned}
    \mathrm{d}u&=\mathrm{d}(z-\arctan\frac{x}{x^2+y^2})=0 \\
    \mathrm{d}z&=\frac{1}{1+(\frac{x}{x^2+y^2})^2}\cdot\frac{(x^2+y^2)\mathrm{d}x-x(2x\mathrm{d}x+2y\mathrm{d}y)}{(x^2+y^2)^2} \\
    &=\frac{(x^2+y^2)^2}{(x^2+y^2)^2+x^2}\cdot\frac{(y^2-x^2)\mathrm{d}x-2xy\mathrm{d}y}{(x^2+y^2)^2} \\
    &=\frac{y^2-x^2}{(x^2+y^2)^2+x^2}\mathrm{d}x-\frac{2xy}{(x^2+y^2)^2+x^2}\mathrm{d}y
\end{aligned}$$

$(2)$ 直接计算可得
$$\begin{aligned}
    \mathrm{d}[u^3-3(x+y)u^2+z^3]&=0 \\
    3u^2\mathrm{d}u-3u^2\mathrm{d}x-6ux\mathrm{d}u-3u^2\mathrm{d}y-6uy\mathrm{d}u+3z^2\mathrm{d}z&=0 \\
    (u^2-2ux-2uy)\mathrm{d}u&=u^2\mathrm{d}x+u^2\mathrm{d}y-3z^2\mathrm{d}z \\
    \mathrm{d}u&=\frac{u^2\mathrm{d}x+u^2\mathrm{d}y-3z^2\mathrm{d}z}{u^2-2(x+y)u}
\end{aligned}$$

> 17. 设二元函数$f(x, y)=|x-y|\varphi(x, y)$，其中$\varphi(x, y)$在点$(0, 0)$处连续。证明$f(x, y)$在点$(0, 0)$处可微的充分必要条件是$\varphi(0, 0)=0$。

**思路：** 重要的一点是，从可微可以推出两个偏导数存在。

**解答：** 必要性：设$f(x, y)$在点$(0, 0)$处可微，则存在偏导数$f_x(0, 0)$和$f_y(0, 0)$。其中
$$\begin{aligned}
    f_x(0, 0)=\lim_{x\to 0}\frac{f(x, 0)-f(0, 0)}{x-0}=\lim_{x\to 0}\frac{|x|\varphi(x, 0)}{x}
\end{aligned}$$
且有
$$\begin{aligned}
    \lim_{x\to 0^+}\frac{|x|\varphi(x, 0)}{x}&=\varphi(0, 0) \\
    \lim_{x\to 0^-}\frac{|x|\varphi(x, 0)}{x}&=-\varphi(0, 0)
\end{aligned}$$
这两者相等，故必有$\varphi(0, 0)=0$。

充分性：设$\varphi(0, 0)=0$，由上面的步骤倒推就知道在点$(0, 0)$的两个偏导数都存在，且$f_x(0, 0)=f_y(0, 0)=0$。现在证明当$(x, y)\to(0, 0)$，有$|x-y|\varphi(x, y)=o(\sqrt{x^2+y^2})$。由上面的计算及微分的定义可得
$$\frac{f(x, y)-f(0, 0)-f_x(0, 0)x-f_y(0, 0)y}{\sqrt{x^2+y^2}}=\frac{|x-y|\varphi(x, y)}{\sqrt{x^2+y^2}}$$
有不等式
$$\frac{|x-y|}{\sqrt{x^2+y^2}}\le\frac{|x|+|y|}{\sqrt{x^2+y^2}}\le 2$$
因此得到
$$\lim_{(x, y)\to(0, 0)}\frac{|x-y|\varphi(x, y)}{\sqrt{x^2+y^2}}=0$$
最后，当$(x, y)\to(0, 0)$时，就有
$$\begin{aligned}
    \Delta f&=f(x, y)-f(0, 0) \\
    &=f_x(0, 0)x+f_y(0, 0)y+|x-y|\varphi(x, y) \\
    &=0x+0y+o(\sqrt{x^2+y^2})
\end{aligned}$$
这符合函数在点$(0, 0)$可微的定义。

> 18. 已知下列式子是某函数$u$的全微分$\mathrm{d}u$，求$u$的表达式(原函数)：$(1)\mathrm{d}u=(2x\cos y+y^2\cos x)\mathrm{d}x+(2y\sin x-x^2\sin y)\mathrm{d}y$；$(2)\mathrm{d}u=\frac{x\mathrm{d}y-y\mathrm{d}x}{x^2+y^2}$；$(3)\mathrm{d}u=(x^2-2yz)\mathrm{d}x+(y^2-2xz)\mathrm{d}y+(z^2-2xy)\mathrm{d}z$；$(4)\mathrm{d}u=(1-\frac{1}{y}+\frac{y}{z})\mathrm{d}x+(\frac{x}{z}+\frac{x}{y^2})\mathrm{d}y-\frac{xy}{z^2}\mathrm{d}z$

**思路：** 凑微分练习题，发挥你全部的注意力吧！

**解答：** $(1)$ 注意到

    &=\mathrm{d}(x^2\cos y+y^2\sin x)
\end{aligned}$$
因此$u=x^2\cos y+y^2\sin x+C$，其中$C$是任意常数。

$(2)$ 注意到
$$\begin{aligned}
    \mathrm{d}\left(\frac{y}{x}\right)&=\frac{x\mathrm{d}y-y\mathrm{d}x}{x^2} \\

\end{aligned}$$
因此$u=\arctan(\frac{y}{x})+C$，其中$C$是任意常数。

$(3)$ 注意到

\end{aligned}$$
因此$u=\frac{x^3}{3}+\frac{y^3}{3}+\frac{z^3}{3}-2xyz+C$，其中$C$是任意常数。

$(4)$ 注意到
$$\begin{aligned}
    \mathrm{d}u&=\mathrm{d}x-\frac{y\mathrm{d}x-x\mathrm{d}y}{y^2}+\frac{yz\mathrm{d}x+xz\mathrm{d}y-xy\mathrm{d}z}{z^2} \\

\end{aligned}$$
因此$u=x-\frac{x}{y}+\frac{xy}{z}+C$，其中$C$是任意常数。

> 20. 求下列函数的高阶偏导数：
>
> $(1)u=\frac{x+y}{x-y}$，求$\frac{\partial^{m+n}u}{\partial x^m\partial y^n}$
> 
> $(2)u=(x-a)^p(y-b)^q$，求$\frac{\partial^{p+q}u}{\partial x^p\partial y^q}$

**思路：** 回忆正文的例子是如何求解的，一般都是进行一些变形后看出某个变量求导方便就先导。

**解答：** $(1)$ 作变形
$$u=\frac{x+y}{x-y}=1+\frac{2y}{x-y}$$
此时先对$x$求$m$阶偏导，得到
$$\begin{aligned}
    \frac{\partial^mu}{\partial x^m}&=(-1)^mm!\cdot\frac{2y}{(x-y)^{m+1}} \\
    &=-2\cdot m!\cdot\left[\frac{y-x+x}{(y-x)^{m+1}}\right] \\
    &=-2\cdot m!\cdot\left[\frac{1}{(y-x)^m}+\frac{x}{(y-x)^{m+1}}\right]
\end{aligned}$$
再对$y$求$n$阶偏导得到
$$\begin{aligned}
    \frac{\partial^{m+n}u}{\partial x^m\partial y^n}&=-2\cdot m!\cdot\left[\frac{(-m)(-m-1)\cdots(-m-n+1)}{(y-x)^{m+n}}+\frac{(-m-1)(-m-2)\cdots(-m-n)x}{(y-x)^{m+n+1}}\right] \\
    &=(-1)^{n+1}2(m+n-1)!\cdot\frac{m(y-x)+(m+n)x}{(y-x)^{m+n+1}} \\
    &=(-1)^{n+1}2(m+n-1)!\cdot\frac{my+nx}{(y-x)^{m+n+1}}
\end{aligned}$$

$(2)$ 先直接对$x$求$p$阶偏导得到
$$\frac{\partial^p u}{\partial x^p}=p!(y-b)^q$$
再对$y$求$q$阶偏导就有
$$\frac{\partial^{p+q}u}{\partial x^p\partial y^q}=\frac{\partial^q}{\partial y^q}\left(\frac{\partial^p u}{\partial x^p}\right)=p!q!$$

因此有$\Delta z=0$。

\end{aligned}$$
因此有$\Delta z=0$。

    2u\frac{\partial u}{\partial x}+2v\frac{\partial v}{\partial x}&\equiv 0 \\

\end{aligned}$$
同理，对$y$求偏导就有
$$\left(\frac{\partial u}{\partial y}\right)^2+u\frac{\partial^2 u}{\partial y^2}+\left(\frac{\partial v}{\partial y}\right)^2+v\frac{\partial^2 v}{\partial y^2}\equiv 0$$

**解答：** 对题目中的等式求积分可得
$$\begin{aligned}
    \int \frac{\partial z}{\partial x}\mathrm{d}x&=\int \left(-\sin y+\frac{1}{1-xy}\right)\mathrm{d}x \\
    z(x, y)&=-x\sin y-\frac{1}{y}\ln|1-xy|+\varphi(y)
\end{aligned}$$
其中$\varphi(y)$是只与$y$有关的函数。代入$x=1$可得
$$\begin{aligned}
    z(1, y)&=\sin y \\
    -\sin y-\frac{1}{y}\ln(1-y)+\varphi(y)&=\sin y \\
    \varphi(y)&=2\sin y+\frac{1}{y}\ln|1-y|
\end{aligned}$$
代回上式可得
$$\begin{aligned}
    z(x, y)&=2\sin y+\frac{1}{y}\ln|1-y|-x\sin y-\frac{1}{y}|1-xy| \\
    &=(2-x)\sin y+\frac{1}{y}\ln\left|\frac{1-y}{1-xy}\right|
\end{aligned}$$

> 31. 求解偏微分方程$y\frac{\partial z}{\partial x}-x\frac{\partial z}{\partial y}=0$。

**思路：** 试试$z=x^2+y^2$，发现满足方程。那么就可以想，是不是所有$z=\varphi(x^2+y^2)$都满足方程呢？

**解答：** 作换元$u=x, v=x^2+y^2$，那么计算可得
$$\begin{aligned}
    \frac{\partial z}{\partial x}&=\frac{\partial z}{\partial u}\cdot\frac{\partial u}{\partial x}+\frac{\partial z}{\partial v}\cdot\frac{\partial v}{\partial x}=\frac{\partial z}{\partial u}+2x\cdot\frac{\partial z}{\partial v} \\
    \frac{\partial z}{\partial y}&=\frac{\partial z}{\partial u}\cdot\frac{\partial u}{\partial y}+\frac{\partial z}{\partial v}\cdot\frac{\partial v}{\partial y}=2y\cdot\frac{\partial z}{\partial v}
\end{aligned}$$
代入方程可得
$$\begin{aligned}
    y\left(\frac{\partial z}{\partial u}+2x\cdot\frac{\partial z}{\partial v}\right)-2xy\cdot\frac{\partial z}{\partial v}&=0 \\
    y\frac{\partial z}{\partial u}&=0
\end{aligned}$$
由于$y$是任意的，因此$\frac{\partial z}{\partial u}=0$，即$z$只与$v$有关。于是可设$z=\varphi(x^2+y^2)$，其中$\varphi$是任意可导函数。

> 32. 设函数$u=f(\sqrt{x^2+y^2})$，其中$f$有连续的二阶导数，且满足$\frac{\partial^2 u}{\partial x^2}+\frac{\partial^2 u}{\partial y^2}=x^2+y^2$，求函数$u$的表达式。

**思路：** 和正文的例子一样，换元计算即可。

**解答：** 令$r=\sqrt{x^2+y^2}$，则
$$\begin{aligned}
    \frac{\partial u}{\partial x}&=\frac{x}{r}f'(r) \\
    \frac{\partial^2 u}{\partial x^2}&=\frac{r-x\cdot\frac{x}{r}}{r^2}f'(r)+\frac{x^2}{r^2}f''(r) \\
    &=\frac{r^2-x^2}{r^3}f'(r)+\frac{x^2}{r^2}f''(r) \\
    \frac{\partial u}{\partial y}&=\frac{y}{r}f'(r) \\
    \frac{\partial^2 u}{\partial y^2}&=\frac{r^2-y^2}{r^3}f'(r)+\frac{y^2}{r^2}f''(r)
\end{aligned}$$
相加后，可得
$$\begin{aligned}
    \frac{\partial^2 u}{\partial x^2}+\frac{\partial^2 u}{\partial y^2}&=x^2+y^2 \\
    \frac{x^2+y^2}{r^3}f'(r)+\frac{x^2+y^2}{r^2}f''(r)&=r^2 \\
    f''(r)+\frac{1}{r}f'(r)&=r^2 \\
    f'(r)&=\mathrm{e}^{-\int r^{-1}\mathrm{d}r}\left(\int r^2\mathrm{e}^{\int r^{-1}\mathrm{d}r}+C_1\right) \\
    &=\frac{1}{r}\left(\frac{r^4}{4}+C_1\right) \\
    &=\frac{r^3}{4}+C_1\cdot\frac{1}{r} \\
    f(r)&=\int \left(\frac{r^3}{4}+C_1\cdot\frac{1}{r}\right)\mathrm{d}r \\
    &=\frac{r^4}{16}+C_1\ln|r|+C_2
\end{aligned}$$
因此$u=\frac{1}{16}(x^2+y^2)^2+C_1\ln|x^2+y^2|+C_2$，其中$C_1, C_2$是任意常数。

> 33. 设$u=f(\ln\sqrt{x^2+y^2})$，其中$f$有连续的二阶导数，且满足$\frac{\partial^2 u}{\partial x^2}+\frac{\partial^2 u}{\partial y^2}=(x^2+y^2)^{\frac{3}{2}}$，求$f(v)$的表达式。

**思路：** 完全类似于上题。

**解答：** 令$v=\ln\sqrt{x^2+y^2}=\frac{1}{2}\ln(x^2+y^2)$，那么
$$\begin{aligned}
    \frac{\partial u}{\partial x}&=f'(v)\cdot\frac{x}{x^2+y^2} \\
    &=\frac{x}{\mathrm{e}^{2v}}f'(v) \\
    \frac{\partial^2 u}{\partial x^2}&=\frac{\mathrm{e}^{2v}-x\cdot 2\cdot\frac{\partial v}{\partial x}\cdot\mathrm{e}^{2v}}{\mathrm{e}^{4v}}f'(v)+\frac{x^2}{\mathrm{e}^{4v}}f''(v) \\
    &=\frac{1-2x\cdot\frac{x}{x^2+y^2}}{\mathrm{e}^{2v}}f'(v)+\frac{x^2}{(x^2+y^2)^2}f''(v) \\
    &=\frac{y^2-x^2}{(x^2+y^2)^2}f'(v)+\frac{x^2}{(x^2+y^2)^2}f''(v) \\
    \frac{\partial^2 u}{\partial y^2}&=\frac{x^2-y^2}{(x^2+y^2)^2}f'(v)+\frac{y^2}{(x^2+y^2)^2}f''(v)
\end{aligned}$$
相加后，可得
$$\begin{aligned}
    \frac{\partial^2 u}{\partial x^2}+\frac{\partial^2 u}{\partial y^2}&=(x^2+y^2)^{\frac{3}{2}} \\
    \frac{1}{\mathrm{e}^{2v}}f''(v)&=(\mathrm{e}^{2v})^{\frac{3}{2}} \\
    f''(v)&=\mathrm{e}^{5v} \\
    f'(v)&=\frac{1}{5}\mathrm{e}^{5v}+C_1 \\
    f(v)&=\frac{1}{25}\mathrm{e}^{5v}+C_1v+C_2
\end{aligned}$$
其中$C_1, C_2$是任意常数。

> 34. 证明：在某个开区域上有恒等式$f(x, y)=\arctan x+\arctan y-\arctan\frac{x+y}{1-xy}\equiv\pi$，并求出使之成立的开区域。

**思路：** 回想多元函数在某区域上恒为常数的条件是其偏导数都为0，导一下看看实力。

**解答：** 计算可得
$$\begin{aligned}
    \frac{\partial f}{\partial x}&=\frac{1}{1+x^2}-\frac{1}{1+(\frac{x+y}{1-xy})^2}\cdot\frac{1-xy-(-y)(x+y)}{(1-xy)^2} \\
    &=\frac{1}{1+x^2}-\frac{(1-xy)^2}{1-2xy+x^2y^2+x^2+2xy+y^2}\cdot\frac{1+y^2}{(1-xy)^2} \\
    &=\frac{1}{1+x^2}-\frac{1+y^2}{(1+x^2)(1+y^2)} \\
    &=0 \\
    \frac{\partial f}{\partial y}&=\frac{1}{1+y^2}-\frac{1}{1+(\frac{x+y}{1-xy})^2}\cdot\frac{1-xy-(-x)(x+y)}{(1-xy)^2} \\
    &=0
\end{aligned}$$
因此$f(x, y)$在其定义域上的偏导数恒为0。因为$f$的定义域为$xy\neq 1$，所以$xy=1$将$\mathbb{R}^2$分为三个区域：
$$D_1:xy<1, D_2:xy>1,x>0, D_3:xy>1, x<0$$
那么$f(x, y)$在这三个区域内都恒为常数。注意到$f(0, 0)=0, f(\sqrt{3}, \sqrt{3})=\pi, f(-\sqrt{3}, -\sqrt{3})=-\pi$，因此使等式成立的开区域为$xy>1, x>0$。

> 35. 设二元函数$f(x, y)$有一阶连续偏导数，且$f(0, 1)=f(1, 0)$。证明：在单位圆周$x^2+y^2=1$上至少存在两个不同的点满足$y\frac{\partial f}{\partial x}=x\frac{\partial f}{\partial y}$。

**思路：** 既然都在单位圆周上了，那实际上就是个一元函数。

**解答：** 设$F(\theta)=f(\cos\theta, \sin\theta)$，则$F$在区间$\theta\in[0, 2\pi]$上可导且连续，而且有$F(0)=F(\frac{\pi}{2})=F(2\pi)$。由罗尔定理，存在一点$\xi\in(0, \frac{\pi}{2})$和一点$\eta\in(\frac{\pi}{2}, 2\pi)$，使得$F'(\xi)=F'(\eta)=0$。此外，我们有
$$\begin{aligned}
    \frac{\mathrm{d}F}{\mathrm{d}\theta}&=\frac{\partial f}{\partial x}\cdot(-\sin\theta)+\frac{\partial f}{\partial y}\cdot\cos\theta \\
    &=x\frac{\partial f}{\partial y}-y\frac{\partial f}{\partial x}
\end{aligned}$$
因此结论成立。

> 36. 设$F(x, y)$有连续的二阶偏导数，且$F_y\neq 0$。则由方程$F(x, y)=C$确定的隐函数$y=y(x)$是线性函数的充要条件为$F_y^2\cdot F_{xx}-2F_x\cdot F_y\cdot F_{xy}+F_x^2\cdot F_{yy}=0$。

**思路：** 线性函数的二阶导数为0，但要注意求导时的计算。

**解答：** 必要性：若$y=y(x)$是线性函数，则$y''=0$。由隐函数求导法可知
$$\begin{aligned}
    \frac{\mathrm{d}y}{\mathrm{d}x}&=-\frac{F_x}{F_y} \\
    \frac{\mathrm{d}^2y}{\mathrm{d}x^2}&=-\frac{(F_x)'_xF_y-F_x(F_y)'_x}{F_y^2} \\
    &=-\frac{(F_{xx}+F_{xy}\cdot y')F_y-(F_{xy}+F_{yy}\cdot y')F_x}{F_y^2} \\
    &=-\frac{1}{F_y^2}\cdot\left[\left(F_{xx}-F_{xy}\cdot\frac{F_x}{F_y}\right)\cdot F_y-\left(F_{xy}-F_{yy}\cdot\frac{F_x}{F_y}\right)\cdot F_x\right] \\
    &=-\frac{F_y^2\cdot F_{xx}-2F_x\cdot F_y\cdot F_{xy}+F_x^2\cdot F_{yy}}{F_y^3}
\end{aligned}$$
由此可知$F_y^2\cdot F_{xx}-2F_x\cdot F_y\cdot F_{xy}+F_x^2\cdot F_{yy}=0$。

充分性：上式逆推可得$y''=0$，即$y=y(x)$是线性函数。

> 37. 设$f(x, y)$在区域$D:x^2+y^2\le R^2$上连续，满足$x\frac{\partial f}{\partial x}+ky\frac{\partial f}{\partial y}=0$，其中$k$是正整数。证明：在$D$上$f(x, y)$恒为常数。

**提示：** 任取$(x_0, y_0)\in D$后令$F(t)=f(t, \frac{y_0}{x_0^k}t^k)$。

**解答：** 任取$(x_0, y_0)\in D, x_0\neq 0$，作连接点$(x_0, y_0)$与原点的曲线$L:x=t, y=\frac{y_0}{x_0^k}x^k, t\in[0, x_0]$。构造辅助函数$F(t)=f(t, \frac{y_0}{x_0^k}t^k)$，则
$$\begin{aligned}
    F'(t)&=f'_1(t, \frac{y_0}{x_0^k}t^k)+k\frac{y_0}{x_0^k}\cdot t^{k-1}f'_2(t, \frac{y_0}{x_0^k}t^k) \\
    &=\frac{1}{t}\left[tf'_1(t, \frac{y_0}{x_0^k}t^k)+k\frac{y_0}{x_0^k}t^kf'_2(t, \frac{y_0}{x_0^k}t^k)\right] \\
    &=0
\end{aligned}$$
这就说明$F(t)$在$[0, x_0]$上恒为常数。分别取$t=0, x_0$可知$f(0, 0)=f(x_0, y_0)$。由$(x_0, y_0)$的任意性即得结论。

> 38. 对于函数$F(x, y)$，如果存在常数$k$，使得对于任何$x, y$及$t>0$恒有$F(tx, ty)=t^kF(x, y)$成立，则称$F(x, y)$是$k$次齐次函数。证明：可微函数$F(x, y)$是$k$次齐次函数的充要条件为对任何$x, y$恒有$xF'_1(x, y)+yF'_2(x, y)=kF(x, y)$成立。

**提示：** 必要性容易证明，充分性可令$f(t)=\frac{F(tx, ty)}{t^k}$，证明$f(t)$恒为常数。

**解答：** 必要性：若$F(x, y)$是$k$次齐次函数，则直接在定义式左右两端对$t$求导数得到
$$xF'_1(tx, ty)+yF'_2(tx, ty)=kt^{k-1}F(x, y)$$
令$t=1$就能得到所需结论。

充分性：若对任何$x, y$恒有$xF'_1(x, y)+yF'_2(x, y)=kF(x, y)$成立，令$f(t)=\frac{F(tx, ty)}{t^k}$，求导可得
$$\begin{aligned}
    f'(t)&=\frac{t^k[xF'_1(tx, ty)+yF'_2(tx, ty)]-kt^{k-1}F(tx, ty)}{t^{2k}} \\
    &=\frac{t^{k-1}[txF'_1(tx, ty)+tyF'_2(tx, ty)-kF(tx, ty)]}{t^{2k}} \\
    &=0
\end{aligned}$$
因此$f(t)$恒为常数$F(x, y)$，移项就有$F(tx, ty)=t^kF(x, y)$。
