## 4.2 多元函数的极限、连续与微分

### 二元函数的极限与连续

二元函数$z=f(x, y)$在点$(x_0, y_0)$的二重极限记为
$$\lim_{(x, y)\to(x_0, y_0)}f(x, y)=A$$
此二重极限存在的充要条件是动点$(x, y)$以任何方式趋向$(x_0, y_0)$是，$f(x, y)$的极限都存在且相等。若不满足此条件，则$f(x, y)$在$(x_0, y_0)$的极限不存在。此外，若有
$$\lim_{(x, y)\to(x_0, y_0)}f(x, y)=f(x_0, y_0)$$
则称$f(x, y)$在点$(x_0, y_0)$连续。

有界闭区域上的连续函数必有最大值和最小值。

初等函数在定义域上处处连续。

### 偏导数

函数$z=f(x, y)$在点$(x_0, y_0)$处固定一个自变量，对另一个自变量求导数，称为偏导数，也就是
$$f_x(x_0, y_0)=\left.\frac{\mathrm{d}}{\mathrm{d}x}f(x ,y_0)\right|_{x=x_0}$$
$$f_y(x_0, y_0)=\left.\frac{\mathrm{d}}{\mathrm{d}y}f(x_0, y)\right|_{y=y_0}$$
还有其他的表示方式，如$f_x(x_0, y_0)$也可写成$\left.\frac{\partial z}{\partial x}\right|_{(x_0, y_0)}, \left.\frac{\partial f}{\partial x}\right|_{(x_0, y_0)}, \left.z_x\right|_{(x_0, y_0)}, z_x(x_0, y_0), f_x(x_0, y_0), f_1'(x_0, y_0)$等。

若$f_x(x_0, y_0)$和$f_y(x_0, y_0)$都存在，则$f(x, y)$在点$(x_0, y_0)$可偏导，但函数在该点`未必连续`。

对上述偏导数再求偏导数称为二阶偏导数，有如下组合
$$\frac{\partial}{\partial x}\left(\frac{\partial z}{\partial x}\right)=\frac{\partial^2 z}{\partial x^2}, \text{或记为}f_{xx}(x, y),f_{11}''(x, y),f_{11}'',z_{11} \\
\frac{\partial}{\partial y}\left(\frac{\partial z}{\partial x}\right)=\frac{\partial z^2}{\partial x\partial y},\text{或记为}f_{xy}(x, y),f_{12}''(x, y),f_{12}'',z_{12} \\
\frac{\partial}{\partial x}\left(\frac{\partial z}{\partial y}\right)=\frac{\partial z^2}{\partial y\partial x},\text{或记为}f_{yx}(x, y),f_{21}''(x, y),f_{21}'',z_{21} \\
\frac{\partial}{\partial y}\left(\frac{\partial z}{\partial y}\right)=\frac{\partial z^2}{\partial y^2},\text{或记为}f_{yy}(x, y),f_{22}''(x, y),f_{22}'',z_{22}$$
同理可以定义更高阶的偏导数。若$f(x, y)$的两个二阶混合偏导数$f_{xy}(x_0, y_0)$和$f_{xy}(x_0, y_0)$在点$(x_0, y_0)$处都`连续`，则$f_{xy}(x_0, y_0)=f_{yx}(x_0, y_0)$。

### 多元复合函数求导的链式法则

1. 设$z=f(u, v)$可微，$u=\varphi(t), v=\psi(t)$可导，则复合函数$z=f[\varphi(t), \psi(t)]$可导，且
$$\frac{\mathrm{d}z}{\mathrm{d}t}=\frac{\partial z}{\partial u}\frac{\mathrm{d}u}{\mathrm{d}t}+\frac{\partial z}{\partial v}\frac{\mathrm{d}v}{\mathrm{d}t}$$

2. 设函数$z=f(u, v), u=\varphi(x, y)$和$v=\psi(x, y)$都可微，则复合函数$z=f[\varphi(x, y), \psi(x, y)]$存在偏导数，且
$$\frac{\partial z}{\partial x}=\frac{\partial z}{\partial u}\frac{\partial u}{\partial z}+\frac{\partial z}{\partial v}\frac{\partial v}{\partial x}, \frac{\partial z}{\partial y}=\frac{\partial z}{\partial u}\frac{\partial u}{\partial y}+\frac{\partial z}{\partial v}\frac{\partial v}{\partial y}$$
我一般称之为哪里有$x$导哪里。

### 全微分

若函数$z=f(x, y)$在点$(x, y)$的全增量$\Delta z=f(x+\Delta x, y+\Delta y)-f(x, y)$可表示为$\Delta z=A\Delta x+B\Delta y+o(\rho)(\rho\to 0)$，且$A, B$是与$\Delta x, \Delta y$无关而至于$(x, y)$有关的常数，$\rho=\sqrt{(\Delta x)^2+(\Delta y)^2}$。则称$z=f(x, y)$在点$(x, y)$可微分，记$\mathrm{d}z=A\Delta x+B\Delta y$，称其为$z=f(x, y)$在点$(x, y)$的全微分。若记$\mathrm{d}x=\Delta x, \mathrm{d}y=\Delta y$，则全微分也可以表示为$\mathrm{d}z=A\mathrm{d}x+B\mathrm{d}y$。

若$z=f(x, y)$可微，则有
$$\mathrm{d}z=\frac{\partial z}{\partial x}\mathrm{d}x+\frac{\partial z}{\partial y}\mathrm{d}y$$
无论$x, y$是自变量还是中间变量，这个运算总是正确的，这称为一阶微分的形式不变性。

若$\frac{\partial z}{\partial x}, \frac{\partial z}{\partial y}$在点$(x, y)$处连续，则函数在该点可微。

若$z=f(x, y)$在点$(x_0, y_0)$可微，则在该点也连续，反之不然。

总结起来，我们有
$$\text{偏导数连续}\xRightarrow{}\text{可微} \\
\text{可微}\xRightarrow{}\text{连续}$$

### 隐函数的求导法

1. 设$F(x, y)$在点$(x_0, y_0)$的某一邻域内具有连续偏导数，且
$$F(x_0, y_0)=0, F_y(x_0, y_0)\neq 0$$
则方程$F(x, y)=0$的某一邻域内可唯一确定具有连续导数的隐函数$y=f(x)$，它满足以下条件：
$$y_0=f(x_0), \frac{\mathrm{d}y}{\mathrm{d}x}=-\frac{F_x}{F_y}$$

2. 我们可以将上述结论推广到三元，甚至$n$元函数的情况，这里给出三元函数的情况。设$F(x, y, z)$在点$(x_0, y_0, z_0)$的某一邻域内具有连续偏导数，且
$$F(x_0, y_0, z_0)=0, F_z(x_0, y_0, z_0)\neq 0$$
则方程$F(x, y, z)=0$的某一邻域内可唯一确定具有连续导数的隐函数$z=f(x, y)$，它满足以下条件：
$$z_0=f(x_0, y_0), \frac{\partial z}{\partial x}=-\frac{F_x}{F_z}, \frac{\partial z}{\partial y}=-\frac{F_y}{F_z}$$

### 方向导数与梯度

1. 给定点$(x_0, y_0)$和方向$\mathbf{l}=(\cos\alpha, \cos\beta)$，对于$f(x, y)$若存在极限
$$\lim_{\rho\to 0^+}\frac{f(x_0+\rho\cos\alpha, y_0+\rho\cos\beta)-f(x_0, y_0)}{\rho}$$
则记为
$$\left.\frac{\partial f}{\partial\mathbf{l}}\right|_{(x_0, y_0)}$$
称为$f(x, y)$在点$(x_0, y_0)$处沿方向$\mathbf{l}$的方向导数。它表示$f(x, y)$在点$(x_0, y_0)$沿方向$\mathbf{l}$的变化率。

2. 实际计算中，若$f(x, y)$可微，则它在点$(x_0, y_0)$沿方向$\mathbf{l}$的方向导数为
$$\left.\frac{\partial f}{\partial\mathbf{l}}\right|_{(x_0, y_0)}=f_x(x_0, y_0)\cos\alpha+f_y(x_0, y_0)\cos\beta$$

3. 可微函数$f(x, y)$在点$(x_0, y_0)$确定的向量
$$\text{grad}f(x_0, y_0)=(f_x(x_0, y_0), f_y(x_0, y_0))$$
称为梯度，它是$f(x, y)$增长速度最快的方向，增长的最快速度是梯度的模$\left|\text{grad}f(x_0, y_0)\right|$。

### 二元函数的泰勒公式

设$z=f(x, y)$在点$(x_0, y_0)$的某邻域$U$内有$n+1$阶的连续偏导数，则对于$U$中的任何点$(x_0+h, y_0+k)$，都存在$\theta\in(0, 1)$，使得
$$\begin{aligned}
    f(x_0+h, y_0+k)&=f(x_0, y_0)+(h\frac{\partial}{\partial x}+k\frac{\partial}{\partial y})f(x_0, y_0) \\
    &+\frac{1}{2!}(h\frac{\partial}{\partial x}+k\frac{\partial}{\partial y})^2f(x_0, y_0) \\
    &+\cdots+\frac{1}{n!}(h\frac{\partial}{\partial x}+k\frac{\partial}{\partial y})^nf(x_0, y_0) \\
    &+\frac{1}{(n+1)!}(h\frac{\partial}{\partial x}+k\frac{\partial}{\partial y})^{n+1}f(x_0+\theta h, y_0+\theta k)
\end{aligned}$$
其中
$$(h\frac{\partial}{\partial x}+k\frac{\partial}{\partial y})^mf(x_0, y_0)=\left.\sum_{p=0}^m\binom{m}{p}h^pm^{m-p}\frac{\partial^mf}{\partial x^p\partial y^{m-p}}\right|_{(x_0, y_0)}$$
例如$n=1$时就有
$$\begin{aligned}
    f(x_0+h, y_0+k)&=f(x_0, y_0)+[hf_x(x_0, y_0)+kf_y(x_0, y_0)] \\
    &+\frac{1}{2!}[h^2f_{xx}(x_0+\theta h, y_0+\theta k)+2hkf_{xy}(x_0+\theta h, y_0+\theta k)+k^2f_{yy}(x_0+\theta h, y_0+\theta k)]
\end{aligned}$$

若函数$f(x, y)$在区域$D$上有偏导数，则它在$D$上恒为常数的充要条件为：它的两个偏导数在$D$上恒为零。

### 例题

> 1. 求下列极限：
> $$(1)\lim_{(x, y)\to(0, 0)}(1+x^2+y^2)^{\frac{xy}{x^2+y^2}}$$
> $$(2)\lim_{(x, y)\to(\infty, \infty)}\frac{x+y}{x^2-xy+y^2}$$

**解答：** $(1)$ 对极限做一点变形
$$\begin{aligned}
    \lim_{(x, y)\to(0, 0)}(1+x^2+y^2)^{\frac{xy}{x^2+y^2}}&=\lim_{(x, y)\to(0, 0)}[(1+x^2+y^2)^{\frac{1}{x^2+y^2}}]^{xy}
\end{aligned}$$
由于
$$\lim_{(x, y)\to(0, 0)}[(1+x^2+y^2)^{\frac{1}{x^2+y^2}}]=\lim_{u\to 0}(1+u)^\frac{1}{u}=\text{e} \\
\lim_{(x, y)\to(0, 0)}xy=0$$
故原极限为1。

$(2)$根据不等式$x^2+y^2\ge 2xy$，我们有
$$0\le\left|\frac{x+y}{x^2-xy+y^2}\right|\le\frac{|x|+|y|}{x^2+y^2-|xy|}\le\frac{|x|+|y|}{|xy|}=\frac{1}{|x|}+\frac{1}{|y|}$$
由于
$$\lim_{(x, y)\to(\infty, \infty)}\left(\frac{1}{|x|}+\frac{1}{|y|}\right)=0$$
故由夹逼准则，原极限为0。

**注：** $(1)$提醒我们，在多元函数求极限时也可以运用一元函数求极限的一些结论。$(2)$则说明了不等式在求二元函数极限时的重要性。

> 2. 设函数$f(x, y)=\begin{cases}
    \frac{x^2y}{x^4+y^2}, &(x, y)\neq (0, 0) \\
    0, &(x, y)=(0, 0)
\end{cases}$，证明在点$(0, 0)$处函数的偏导数存在，但在该点函数不连续。

**解答：** 由偏导数的定义，计算可得
$$f_x(0, 0)=\left.\frac{\mathrm{d}}{\mathrm{d}x}f(x, 0)\right|_{x=0}=0$$
同理可得$f_y(0, 0)=0$，因此在点$(0, 0)$处，函数的两个偏导数都存在。但是，若沿任何直线$y=kx$趋于$(0, 0)$时，有
$$\lim_{(x, y)\to(0, 0)}f(x, y)=\lim_{x\to 0}\frac{kx^3}{x^4+k^2x^2}=0$$
而沿抛物线$y=x^2$趋于$(0, 0)$时，又有
$$\lim_{(x, y)\to(0, 0)}f(x, y)=\lim_{x\to 0}\frac{x^4}{x^4+x^4}=\frac{1}{2}$$
两者极限不同，因此$\lim_{(x, y)\to(0, 0)}f(x, y)$不存在，在点$(0, 0)$处函数不连续。

**注：** 本例告诉我们，就算沿任何直线趋于某点时，函数极限均存在且相等，也不代表函数在该点就连续，在$x, y$次数不相同时要更加留意。

> 3. 设$f(x, y)=(1+xy)^y$，求$f_x(1, 1)$和$f_y(1, 1)$。

**解答1:** 先求偏导函数
$$f(x, y)=(1+xy)^y$$
$$f_x(x, y)=y^2(1+xy)^{y-1}$$
$$\begin{aligned}
    f_y(x, y)&=(1+xy)^y\left(\ln(1+xy)+\frac{xy}{1+xy}\right)
\end{aligned}$$
代入就有$f_x(1, 1)=1, f_y(1, 1)=2\ln 2+1$。

**解答2:** 由定义
$$\begin{aligned}
    f_x(1, 1)&=\left.\frac{\mathrm{d}}{\mathrm{d}x}(1+x)\right|_{x=1}=1 \\
    f_y(1, 1)&=\left.\frac{\mathrm{d}}{\mathrm{d}y}(1+y)^y\right|_{y=1} \\
    &=\left.(1+y)^y[\ln(1+y)+\frac{y}{1+y}]\right|_{y=1} \\
    &=2\ln 2+1
\end{aligned}$$

**注：** 解答1是较为通用的解法，但当函数形式复杂时，求偏导函数就很困难，此时使用定义是很好的方法。其实这和一元函数的情况是一样的。

> 4. 设函数$f(x, y)=\sqrt{x^2+y^2}$。证明$f(x, y)$在点$(0, 0)$处连续，沿任何方向的方向导数都存在，但在$(0, 0)$处的偏导数不存在。

**解答：** 因初等函数在定义域上处处连续，故$f(x, y)$在点$(0, 0)$处连续。设方向向量为$\mathbf{l}=(\cos\alpha, \cos\beta)$，则在点$(0, 0)$处的方向导数为
$$\begin{aligned}
    \left.\frac{\partial f}{\partial\mathbf{l}}\right|_{(0, 0)}&=\lim_{\rho\to 0^+}\frac{f(\rho\cos\alpha, \rho\cos\beta)-f(0, 0)}{\rho} \\
    &=\lim_{\rho\to 0^+}\frac{\sqrt{\rho^2\cos^2\alpha+\rho^2\cos^2\beta}}{\rho} \\
    &=1
\end{aligned}$$
因此任意方向的方向导数都存在。但是，由定义
$$\begin{aligned}
    f_x(0, 0)&=\left.\frac{\mathrm{d}}{\mathrm{d}x}f(x, 0)\right|_{x=0} \\
    &=\left.\frac{\mathrm{d}}{\mathrm{d}x}|x|\right|_{x=0}
\end{aligned}$$
因$|x|$在$x=0$处的导数不存在，故$f_x(0, 0)$不存在。同理$f_y(0, 0)$不存在。

> 5. 证明：函数
> $$f(x, y)=\begin{cases}
    \frac{x^2y}{x^2+y^2}, &(x, y)\neq (0, 0) \\
    0, &(x, y)=(0, 0)
\end{cases}$$
> 在原点连续，偏导数存在，但在原点不可微。

**解答：** 先证连续性
$$0\le\left|\frac{x^2y}{x^2+y^2}\right|\le\frac{x^2}{x^2+y^2}|y|\le |y|\to 0((x, y)\to (0,0))$$
因此$f(x, y)$在原点连续。由偏导数的定义
$$\begin{aligned}
    f_x(0, 0)&=\left.\frac{\mathrm{d}}{\mathrm{d}x}f(x, 0)\right|_{x=0} \\
    &=0
\end{aligned}$$
同理可得$f_y(0, 0)=0$，因此$f(x, y)$在原点存在偏导数。假定$f(x, y)$在原点可微，则有
$$f(\Delta x, \Delta y)-f(0, 0)=\frac{(\Delta x)^2\Delta y}{(\Delta x)^2+(\Delta y)^2}=o(\rho)(\rho\to 0)$$
其中$\rho=\sqrt{(\Delta x)^2+(\Delta y)^2}$。因此，应有极限
$$\lim_{(\Delta x, \Delta y)\to (0, 0)}\frac{\frac{(\Delta x)^2\Delta y}{(\Delta x)^2+(\Delta y)^2}}{\sqrt{(\Delta x)^2+(\Delta y)^2}}=\lim_{(\Delta x, \Delta y)\to(0, 0)}\frac{(\Delta x)^2\Delta y}{[(\Delta x)^2+(\Delta y)^2]^{\frac{3}{2}}}=0$$
取$\Delta x=\Delta y$，则有
$$\begin{aligned}
    \lim_{(\Delta x, \Delta y)\to(0, 0)}\frac{(\Delta x)^2\Delta y}{[(\Delta x)^2+(\Delta y)^2]^{\frac{3}{2}}}&=\lim_{\Delta x\to 0}\frac{(\Delta x)^3}{(2(\Delta x)^2)^{\frac{3}{2}}} \\
    &=\frac{1}{2\sqrt{2}}
\end{aligned}$$
但取$\Delta y=0$，就有
$$\lim_{(\Delta x, \Delta y)\to(0, 0)}\frac{(\Delta x)^2\Delta y}{[(\Delta x)^2+(\Delta y)^2]^{\frac{3}{2}}}=0$$
因此这个极限不存在，与假设矛盾！最后，函数在原点不可微。

**注：** 这说明偏导数存在只是可微的必要条件，不是充分条件。

> 6. 用两种方法求下列函数的全微分$\mathrm{d}z$：
> 
> $(1)z=\mathrm{e}^{\frac{y^2}{x}}$；
> 
> $(2)$由方程$\mathrm{e}^z=xyz$确定的隐函数$z=z(x, y)$

**方法1:** 先求偏导数，在写出全微分表达式

$(1)$因为
$$\begin{aligned}
    \frac{\partial z}{\partial x}&=-\frac{y^2}{x^2}\mathrm{e}^{\frac{y^2}{x}} \\
    \frac{\partial z}{\partial y}&=\frac{2y}{x}\mathrm{e}^{\frac{y^2}{x}}
\end{aligned}$$
因此
$$\mathrm{d}z=\frac{\partial z}{\partial x}\mathrm{d}x+\frac{\partial z}{\partial y}\mathrm{d}y=\mathrm{e}^{\frac{y^2}{x}}\left(-\frac{y^2}{x^2}\mathrm{d}x+\frac{2y}{x}\mathrm{d}y\right)$$

$(2)$用隐函数求导法，设$F(x, y, z)=\mathrm{e}^z-xyz$，则$F(x, y, z)\equiv 0, F_z(x, y, z)=\mathrm{e}^z-xy=\mathrm{e}^z(1-z^{-1})$不恒为0
$$\begin{aligned}
    \frac{\partial z}{\partial x}&=-\frac{F_x}{F_z}=\frac{yz}{\mathrm{e}^z-xy}=\frac{z}{x(z-1)} \\
    \frac{\partial z}{\partial y}&=-\frac{F_y}{F_z}=\frac{xz}{\mathrm{e}^z-xy}=\frac{z}{y(z-1)}
\end{aligned}$$
因此
$$\mathrm{d}z=\frac{z}{x(z-1)}\mathrm{d}x+\frac{z}{y(z-1)}\mathrm{d}y$$

**方法2:** 直接使用微分的运算公式

$(1)$可得
$$\begin{aligned}
    \mathrm{d}z&=\mathrm{e}^{\frac{y^2}{x}}\mathrm{d}\left(\frac{y^2}{x}\right) \\
    &=\mathrm{e}^{\frac{y^2}{x}}\cdot\frac{2xy\mathrm{d}y-y^2\mathrm{d}x}{x^2} \\
    &=\mathrm{e}^{\frac{y^2}{x}}\left(-\frac{y^2}{x^2}\mathrm{d}x+\frac{2y}{x}\mathrm{d}y\right)
\end{aligned}$$

$(2)$两边同时取微分有
$$\begin{aligned}
    \mathrm{d}(\mathrm{e}^z)&=\mathrm{d}(xyz) \\
    \mathrm{e}^z\mathrm{d}z&=yz\mathrm{d}x+xz\mathrm{d}y+xy\mathrm{d}z \\
    (xyz-xy)\mathrm{d}z&=yz\mathrm{d}x+xz\mathrm{d}y \\
    \mathrm{d}z&=\frac{z}{x(z-1)}\mathrm{d}x+\frac{z}{y(z-1)}\mathrm{d}y
\end{aligned}$$

**注：** 由此题可知，使用微分的运算公式过程更简洁。

> 7. 已知下列式子是某个函数$u$的全微分$\mathrm{d}u$，求$u$的表达式(原函数)：
> 
> $(1)\mathrm{d}u=(x^2+x^3+y)\mathrm{d}x+(1+x)\mathrm{d}y$
> 
> $(2)\mathrm{d}u=\frac{yz\mathrm{d}x+xz\mathrm{d}y+xy\mathrm{d}z}{\sqrt{1+xyz}}$

**解答：** 

$(1)$从$\mathrm{d}u=\frac{\partial u}{\partial x}\mathrm{d}x+\frac{\partial u}{\partial y}\mathrm{d}y$出发可得
$$\begin{cases}
    \frac{\partial u}{\partial x}=x^2+x^3+y \\
    \frac{\partial u}{\partial y}=1+x
\end{cases}$$
对第一式积分可得
$$u=\frac{1}{4}x^4+\frac{1}{3}x^3+xy+\varphi(y)$$
将此式对$y$求偏导，可得
$$\begin{aligned}
    \frac{\partial u}{\partial y}&=1+x \\
    x+\varphi'(y)&=1+x \\
    \varphi'(y)&=1 \\
    \varphi(y)&=y+C
\end{aligned}$$
因此
$$u=\frac{1}{4}x^4+\frac{1}{3}x^3+xy+y+C$$
其中$C$是任意常数。

或直接观察到
$$\begin{aligned}
    \mathrm{d}u&=x^2\mathrm{d}x+x^3\mathrm{d}x+\mathrm{d}y+(y\mathrm{d}x+x\mathrm{d}y) \\
    &=\mathrm{d}\left(\frac{x^3}{3}\right)+\mathrm{d}\left(\frac{x^4}{4}\right)+\mathrm{d}y+\mathrm{d}(xy) \\
    &=\mathrm{d}\left(\frac{x^3}{3}+\frac{x^4}{4}+xy+y\right) \\
    u&=\frac{x^4}{4}+\frac{x^3}{3}+xy+y+C
\end{aligned}$$

$(2)$观察可得
$$\begin{aligned}
    \mathrm{d}u&=\frac{\mathrm{d}(1+xyz)}{\sqrt{1+xyz}} \\
    &=\mathrm{d}\left(2\sqrt{1+xyz}\right) \\
    u&=2\sqrt{1+xyz}+C
\end{aligned}$$
其中$C$是任意常数。

**注：** 从本例，我们可以看到积分的方法只对二元函数适用，再往上就很复杂了。另一个方法称为“凑微分”法，过程很简洁，但是比较需要注意力。

> 8. 设变量$x, y, z$满足关系式$z=f(x+y+z, xyz)$，其中$f$可微。求$\frac{\partial z}{\partial x}, \frac{\partial x}{\partial z}, \frac{\partial x}{\partial y}$。

**方法1:** 首先将$z$看为因变量，$x, y$是相互独立的自变量，则由复合函数求导的链式法则
$$\begin{aligned}
    \frac{\partial z}{\partial x}&=f'_1\cdot(1+\frac{\partial z}{\partial x})+f'_2\cdot(yz+xy\cdot\frac{\partial z}{\partial x}) \\
    (1-f'_1-xyf'_2)\frac{\partial z}{\partial x}&=f'_1+yzf'_2 \\
    \frac{\partial z}{\partial x}&=\frac{f'_1+yzf'_2}{1-f'_1-xyf'_2}
\end{aligned}$$
然后将$x$看为因变量，$y, z$是相互独立的自变量
$$\begin{aligned}
    1&=f'_1\cdot(\frac{\partial x}{\partial z}+1)+f'_2\cdot(xy+yz\cdot\frac{\partial x}{\partial z}) \\
    (f'_1+yzf'_2)\frac{\partial x}{\partial z}&=1-f'_1-xyf'_2 \\
    \frac{\partial x}{\partial z}&=\frac{1-f'_1-xyf'_2}{f'_1+yzf'_2}
\end{aligned}$$
同理有
$$\begin{aligned}
    0&=f'_1\cdot(1+\frac{\partial x}{\partial y})+f'_2\cdot(xz+yz\cdot\frac{\partial x}{\partial y}) \\
    (f'_1+yzf'_2)\frac{\partial x}{\partial y}&=-(f'_1+xzf'_2) \\
    \frac{\partial x}{\partial y}&=-\frac{f'_1+xzf'_2}{f'_1+yzf'_2}
\end{aligned}$$

**方法2:** 
等式两端求微分可得
$$\begin{aligned}
    \mathrm{d}z&=f'_1\cdot\mathrm{d}(x+y+z)+f'_2\cdot\mathrm{d}(xyz) \\
    &=f'_1(\mathrm{d}x+\mathrm{d}y+\mathrm{d}z)+f'_2(yz\mathrm{d}x+xz\mathrm{d}y+xy\mathrm{d}z) \\
    (1-f'_1-xyf'_2)\mathrm{d}z&=(f'_1+yzf'_2)\mathrm{d}x+(f'_1+xzf'_2)\mathrm{d}y
\end{aligned}$$
由此可知
$$\begin{aligned}
    \frac{\partial z}{\partial x}&=\frac{f'_1+yzf'_2}{1-f'_1-xyf'_2} \\
    \frac{\partial x}{\partial z}&=\frac{1-f'_1-xyf'_2}{f'_1+yzf'_2} \\
    \frac{\partial x}{\partial y}&=-\frac{f'_1+xzf'_2}{f'_1+yzf'_2}
\end{aligned}$$

**方法3:** 设$F(x, y, z)=z-f(x+y+z, xyz)$，则
$$\begin{aligned}
    \frac{\partial z}{\partial x}&=-\frac{F_x}{F_z}=-\frac{-f'_1-yzf'_2}{1-f'_1-xyf'_2}=\frac{f'_1+yzf'_2}{1-f'_1-xyf'_2} \\
    \frac{\partial x}{\partial z}&=-\frac{F_z}{F_x}=\frac{1-f'_1-xyf'_2}{f'_1+yzf'_2} \\
    \frac{\partial x}{\partial y}&=-\frac{F_y}{F_x}=-\frac{f'_1+xzf'_2}{f'_1+yzf'_2}
\end{aligned}$$

**注：** 方法3称为公式法，是三个方法中最简洁的，建议多用。方法2其实也还行，方法1就过于复杂了。

> 9. 由方程组$\begin{cases}
    xu-yv=0 \\
    yu+xv=1
> \end{cases}$确定了$u, v$是$x, y$的函数，求$\frac{\partial u}{\partial x},\frac{\partial u}{\partial y},\frac{\partial v}{\partial x},\frac{\partial v}{\partial y}$。

**方法1：** 对方程组内的$x$求偏导数可得
$$\begin{cases}
    u+x\frac{\partial u}{\partial x}-y\frac{\partial v}{\partial x}=0 \\
    y\frac{\partial u}{\partial x}+v+x\frac{\partial v}{\partial x}=1
\end{cases}$$
可以解得
$$\begin{cases}
    \frac{\partial u}{\partial x}=-\frac{xu+yv}{x^2+y^2} \\
    \frac{\partial v}{\partial x}=\frac{yu-xv}{x^2+y^2}
\end{cases}$$
同理可得
$$\begin{cases}
    \frac{\partial u}{\partial y}=\frac{xv-yu}{x^2+y^2} \\
    \frac{\partial v}{\partial y}=-\frac{xu+yv}{x^2+y^2}
\end{cases}$$

**方法2:** 对方程组取微分可得
$$\begin{cases}
    (u\mathrm{d}x+x\mathrm{d}u)-(v\mathrm{d}y+y\mathrm{d}v)=0 \\
    (u\mathrm{d}y+y\mathrm{d}u)+(v\mathrm{d}x+x\mathrm{d}v)=0
\end{cases}$$
$$\begin{cases}
    x\mathrm{d}u-y\mathrm{d}v=-u\mathrm{d}x+v\mathrm{d}y \\
    x\mathrm{d}v+y\mathrm{d}u=-v\mathrm{d}x-u\mathrm{d}y
\end{cases}$$
$$\begin{cases}
    (x^2+y^2)\mathrm{d}u=-(xu+yv)\mathrm{d}x+(xv-yu)\mathrm{d}y \\
    (x^2+y^2)\mathrm{d}v=-(xv-yu)\mathrm{d}x-(xu+yv)\mathrm{d}y
\end{cases}$$
由此可得
$$\begin{aligned}
    \frac{\partial u}{\partial x}&=-\frac{xu+yv}{x^2+y^2} \\
    \frac{\partial u}{\partial y}&=\frac{xv-yu}{x^2+y^2} \\
    \frac{\partial v}{\partial x}&=\frac{yu-xv}{x^2+y^2} \\
    \frac{\partial v}{\partial y}&=-\frac{xu+yv}{x^2+y^2}
\end{aligned}$$

**注：** 方法2，即微分法的运算量更少，建议使用。

> 10. 设$z=(x^2+y^2)\mathrm{e}^{x+y}$，求高阶混合偏导数$\frac{\partial^{m+n}z}{\partial x^m\partial y^n}$。

**解答：** 设$u=x^2\mathrm{e}^x\mathrm{e}^y, v=y^2\mathrm{e}^x\mathrm{e}^y$，则$z=u+v$。对函数$u$而言，有$\frac{\partial^n u}{\partial y^n}=x^2\mathrm{e}^x\mathrm{e}^y$，再由高阶导数的莱布尼茨公式得到
$$\begin{aligned}
    \frac{\partial^{m+n}u}{\partial x^m\partial y^n}&=\frac{\partial^m}{\partial x^m}\left(\frac{\partial^nu}{\partial y^n}\right) \\
    &=\frac{\partial^m}{\partial x^m}(x^2\mathrm{e}^x\mathrm{e}^y) \\
    &=\mathrm{e}^{x+y}[x^2+2mx+m(m-1)]
\end{aligned}$$
同理有
$$\begin{aligned}
    \frac{\partial^{m+n}v}{\partial x^m\partial y^n}&=\frac{\partial^n}{\partial y^n}\left(\frac{\partial^mv}{\partial x^m}\right) \\
    &=\frac{\partial^n}{\partial y^n}(y^2\mathrm{e}^x\mathrm{e}^y) \\
    &=\mathrm{e}^{x+y}[y^2+2ny+y(y-1)]
\end{aligned}$$
最后得到
$$\begin{aligned}
    \frac{\partial^{m+n}z}{\partial x^m\partial y^n}&=\frac{\partial^{m+n}u}{\partial x^m\partial y^n}+\frac{\partial^{m+n}v}{\partial x^m\partial y^n} \\
    &=\mathrm{e}^{x+y}[x^2+y^2+2mx+2ny+m(m-1)+n(n-1)]
\end{aligned}$$

**注：** 这种题目需要合理运用一元函数的求导法则来简化运算。

> 11. 设$u=u(\sqrt{x^2+y^2})$具有连续二阶偏导数，且满足$\frac{\partial^2u}{\partial x^2}+\frac{\partial^2u}{\partial y^2}-\frac{1}{x}\frac{\partial u}{\partial x}+u=x^2+y^2$。试求函数$u$的表达式。

**解答：** 令$r=\sqrt{x^2+y^2}$，可知
$$\begin{aligned}
    \frac{\partial u}{\partial x}&=\frac{\mathrm{d}u}{\mathrm{d}r}\cdot\frac{\partial r}{\partial x} \\
    &=\frac{x}{r}\cdot\frac{\mathrm{d}u}{\mathrm{d}r} \\
    \frac{\partial^2u}{\partial x^2}&=\frac{x^2}{r^2}\cdot\frac{\mathrm{d}^2u}{\mathrm{d}r^2}+\frac{r-x\frac{\partial r}{\partial x}}{r^2}\cdot\frac{\mathrm{d}u}{\mathrm{d}r} \\
    &=\frac{x^2}{r^2}\cdot\frac{\mathrm{d}^2u}{\mathrm{d}r^2}+\frac{1}{r}\cdot\frac{\mathrm{d}u}{\mathrm{d}r}-\frac{x^2}{r^3}\cdot\frac{\mathrm{d}u}{\mathrm{d}r}
\end{aligned}$$
同理可得
$$\begin{aligned}
    \frac{\partial u}{\partial y}&=\frac{\mathrm{d}u}{\mathrm{d}r}\cdot\frac{\partial r}{\partial y} \\
    &=\frac{y}{r}\cdot\frac{\mathrm{d}u}{\mathrm{d}r} \\
    \frac{\partial^2u}{\partial y^2}&=\frac{y^2}{r^2}\cdot\frac{\mathrm{d}^2u}{\mathrm{d}r^2}+\frac{1}{r}\cdot\frac{\mathrm{d}u}{\mathrm{d}r}-\frac{y^2}{r^3}\cdot\frac{\mathrm{d}u}{\mathrm{d}r}
\end{aligned}$$
代入等式可得
$$\begin{aligned}
    \frac{x^2+y^2}{r^2}\cdot\frac{\mathrm{d}^2u}{\mathrm{d}r^2}+\frac{2}{r}\cdot\frac{\mathrm{d}u}{\mathrm{d}r}-\frac{x^2+y^2}{r^3}\cdot\frac{\mathrm{d}u}{\mathrm{d}r}-\frac{1}{x}\cdot\frac{x}{r}\cdot\frac{\mathrm{d}u}{\mathrm{d}r}+u&=x^2+y^2 \\
    \frac{\mathrm{d}^2u}{\mathrm{d}r^2}+u&=r^2
\end{aligned}$$
解这个二阶常系数非齐次微分方程，可得通解为
$$u=C_1\cos r+C_2\sin r+r^2-2$$
其中$C_1, C_2$为任意常数。最终结果为
$$u=C_1\cos(\sqrt{x^2+y^2})+C_2\sin(\sqrt{x^2+y^2})+x^2+y^2-2$$

> 12. 已知方程$u+\mathrm{e}^u=xy$确定了隐函数$u=u(x, y)$，求$\left.\frac{\partial^2u}{\partial x\partial y}\right|_{(1, 1)}$。

**解答：** 设$F(u, x, y)=u+\mathrm{e}^u-xy$，显然$F(0, 1, 1)=0$，且
$$F_u(0, 1, 1)=\left.(1+\mathrm{e}^u)\right|_{u=0}=2\neq 0$$
因此可用隐函数求导法。此时
$$\begin{aligned}
    \frac{\partial u}{\partial x}&=-\frac{F_x}{F_u} \\
    &=\frac{y}{1+\mathrm{e}^u} \\
    \frac{\partial u}{\partial y}&=-\frac{F_y}{F_u} \\
    &=\frac{x}{1+\mathrm{e}^u}
\end{aligned}$$
因此
$$\begin{aligned}
    \left.\frac{\partial^2u}{\partial x\partial y}\right|_{(1, 1)}&=\left.\frac{\partial}{\partial y}\left(\frac{\partial u}{\partial x}\right)\right|_{(1, 1)} \\
    &=\left.\frac{\partial}{\partial y}\left(\frac{y}{1+\mathrm{e}^u}\right)\right|_{(1, 1)} \\
    &=\left.\frac{(1+\mathrm{e}^u)-y\mathrm{e}^u\cdot\frac{\partial u}{\partial y}}{(1+\mathrm{e}^u)^2}\right|_{(1, 1)} \\
    &=\left.\frac{(1+\mathrm{e}^u)^2-xy\mathrm{e}^u}{(1+\mathrm{e}^u)^3}\right|_{(1, 1)} \\
    &=\frac{3}{8}
\end{aligned}$$

> 13. 解二阶偏微分方程$\frac{\partial^2z}{\partial x\partial y}=0$，其中$z=z(x, y)$有连续的二阶偏导数。

**解答：** 方程可写为
$$\frac{\partial}{\partial x}\left(\frac{\partial z}{\partial y}\right)=0$$
由此可知偏导函数$\frac{\partial z}{\partial y}$与$x$无关，可设$\frac{\partial z}{\partial y}=h(y)$，其中$h(y)$是任意的可微函数。此时，$z$可视为$h(y)$的原函数，两边积分得到
$$\begin{aligned}
    z=\int h(y)\mathrm{d}y+\varphi(x)
\end{aligned}$$
由$h(y)$的任意性可知$\int h(y)\mathrm{d}y$也是任意的，可设为$\psi(y)$。因此我们得到了最终结果
$$z=\varphi(x)+\psi(y)$$
其中$\varphi(x), \psi(y)$都是可微函数。

**注：** 对于某些特殊的偏微分方程，例如本题，是可以通过偏导数的基本原理来求解的。一般来说，二阶偏微分方程的通解有两个任意函数。
